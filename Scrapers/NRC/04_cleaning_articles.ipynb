```python
r = {'url': 'https://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141', 'raw_html': b'<!DOCTYPE html>\n<html class=" " lang="nl">\n<head>\n\t\t<meta charset="utf-8">\n\n<!--\n\nAlle auteurs-, naburige en databankrechten die op de inhoud en opmaak van de NRC websites\nen NRC apps rusten, worden door NRC Media BV uitdrukkelijk voorbehouden. De inhoud van de\nNRC websites en apps is uitsluitend voor persoonlijk, niet-commercieel gebruik en het is\nniet toegestaan om gegevens van de website of uit de apps door middel van screen scraping\n(of een andere geautomatiseerde werkwijze) te vergaren.\nZie ook https://www.nrc.nl/leveringsvoorwaarden/\n \nAll copyrights, neighbouring rights and database rights in the content and layout of the\nNRC websites and NRC apps are explicitly reserved by NRC Media BV. The content of the NRC\nwebsites and NRC apps is for personal, non-commercial use only and it is not allowed to\ncollect data from the website or from the apps by means of screen scraping (or any other\nautomated method).\nSee also https://www.nrc.nl/leveringsvoorwaarden/\n\n-->\n\n<meta name="dcterms.rights" content="Copyright NRC Media BV">\n<link rel="license" href="https://www.nrc.nl/leveringsvoorwaarden/">\t\t<title>Minder inkomen \xc3\xa9n weinig spaargeld, het is de werkelijkheid voor een half miljoen werkenden - NRC</title>\n\n\t\t\t<meta name="viewport" content="width=device-width, initial-scale=1">\n\n\n\n\t\t<link rel="preload" href="https://assets.nrc.nl/static/front/fonts/GuardianEgyp-Medium-Web.331856397.woff2" as="font" type="font/woff2" crossorigin>\n\n\t\t<link rel="preload" href="https://assets.nrc.nl/static/front/fonts/GuardianEgyp-Light-Web.21f6f2364.woff2" as="font" type="font/woff2" crossorigin>\n\n\t\t<link rel="preload" href="https://assets.nrc.nl/static/front/fonts/LFT_Etica_Semibold.84c5e94cc.woff2" as="font" type="font/woff2" crossorigin>\n\n\t\t\t<link rel="stylesheet" href="https://assets.nrc.nl/static/front/css/main.0f428ef76.css">\n\n\n\n\n\t\n\n\t\t\t<link rel="stylesheet" href="https://assets.nrc.nl/static/front/css/split/luister-naar.214667307.css">\n\n\n\n\n\n\n\n\t\t\t<link rel="stylesheet" href="https://assets.nrc.nl/static/front/css/split/paywall.-280952311.css">\n\n\n\t<script>\n\t\twindow.nmt = window.nmt || {};\n\t\t(window.nmt.setColorScheme = function(docEl) {\n\t\t\ttry {\n\t\t\t\tdocEl = docEl || document.documentElement;\n\t\t\t\tconst docElClassList = docEl.classList;\n\t\t\t\tconst storedScheme = window.localStorage.getItem(\'nmt_color_scheme\');\n\t\t\t\tconst forceDarkMode = docElClassList.contains(\'force-dark-mode\');\n\n\t\t\t\tif ((forceDarkMode == false && storedScheme == \'light\')\n\t\t\t\t\t|| (forceDarkMode == false && window.matchMedia(\'(prefers-color-scheme: light)\').matches && storedScheme != \'dark\')\n\t\t\t\t) {\n\t\t\t\t\treturn;\n\t\t\t\t}\n\n\t\t\t\tif (forceDarkMode == true\n\t\t\t\t\t|| storedScheme == \'dark\'\n\t\t\t\t\t|| window.matchMedia(\'(prefers-color-scheme: dark)\').matches) {\n\t\t\t\t\tdocElClassList.add(\'dark-mode\');\n\t\t\t\t}\n\t\t\t} catch (err) {\n\t\t\t\tconsole.log(err);\n\t\t\t}\n\t\t})();\n\t</script>\n\n\n<meta name="twitter:site" content="@nrc">\n<meta property="twitter:account_id" content="4503599627452245">\n<meta property="og:site_name" content="NRC">\n<meta property="fb:page_id" content="196118303833398">\n\n<meta property="fb:pages" content="196118303833398"><meta property="fb:pages" content="285390408300968"><meta property="fb:pages" content="238487967764"><meta property="fb:pages" content="1331135880337552"><meta property="og:type" content="article">\n<meta property="og:url" content="https://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141">\n\n\t<meta name="twitter:creator" content="@nrc">\n\n<meta property="og:title" content="Minder inkomen \xc3\xa9n weinig spaargeld, het is de werkelijkheid voor een half miljoen werkenden">\n<meta property="og:description" content="CPB-onderzoek: Vooral flexwerkers verloren inkomen in coronajaar 2020. Veel van hen hadden weinig spaargeld om deze tegenslag op te vangen, concludeert het CPB nu.">\n<meta name="description" content="CPB-onderzoek: Vooral flexwerkers verloren inkomen in coronajaar 2020. Veel van hen hadden weinig spaargeld om deze tegenslag op te vangen, concludeert het\xe2\x80\xa6">\n\n<meta name="twitter:card" content="summary_large_image">\n<meta property="og:image" content="https://images.nrc.nl/rsJp5-SbHACE7AknCKH05lMJpvc=/1200x627/smart/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg">\n\n\n\t<meta property="og:image:width" content="1200">\n\t<meta property="og:image:height" content="627">\n\n\t<meta property="og:image:alt" content="In de horeca werden werknemers hard geraakt door  virusuitbraak">\n\n\t<meta name="googlebot" content="noarchive">\n\n\t<meta name="parsely-type" content="post">\n\t<link rel="shortcut icon" href="https://assets.nrc.nl/static/front/icons/favicon.ico">\n\t<link rel="mask-icon" href="https://assets.nrc.nl/static/front/icons/favicon.svg" color="#d30910"><link rel="apple-touch-icon" href="https://assets.nrc.nl/static/front/icons/apple-touch-icon-180x180.png">\n\t<link rel="canonical" href="https://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141">\n\n</head>\n<body class="artikel\n\tarticle-view article-view--nieuwsbericht has-nmt-layout-sidebar allow-de-sidebar has-article-share has-featured-image--below-header in-dossier-coronavirus\n\t\n\t\n\t"\n\t\tdata-article-updated-at="2022-03-01T14:37:29+02:00"\n\tdata-nmo-content-id="1200"\n\n\n\tdata-page-type="article"\n\t\tdata-allow-ads="1"\n\tdata-article-id="4095141"\n\tdata-article-urn="urn:X-nl-nrc-article:nrc-nl:gn4:45785917"\n\tdata-article-relative-url="/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141"\n\tdata-article-type="nieuwsbericht"\n\t\tdata-dossier="coronavirus"\n\tdata-search-api="/search/api/v1/article/search"\n\tdata-search-banner-text-abonnement="Abonnement op NRC Handelsblad of NRC in de ochtend - Vanaf \xe2\x82\xac 2,00 per week."\n\tdata-user-is-logged-in="0"\n\tdata-canonical-base-url="https://www.nrc.nl"\n\tdata-papi-base-url="https://www.nrc.nl/papi/api"\n\tdata-static-path="https://assets.nrc.nl/static/front/"\n\tdata-paywall-limit="10"\n\tdata-paywall-article-expire-days="30"\n\t>\n\t\n<div class="shared-svgs" hidden>\n<svg>\n\t<defs>\n\t\t<symbol id="logo-nrc" viewBox="0 0 491 147">\n\t\t\t<g transform="translate(-166 -395)">\n\t\t\t\t<path d="M18,247l17-4V154l-17-4V130l53-6,4,2v14h2c14-10,28-16,46-16,23,0,37,10,37,35v85l12,3v19H105V247l14-3V173c0-14-7-18-20-18-8,0-15,2-22,5v85l13,2v19H18Zm169,0,12-3V155l-17-5V130l53-6,4,2v20h1c7-12,20-22,36-22,2,0,6,1,7,1v36c-3-1-8-1-12-1-13,0-23,2-31,6v76l20,4v21H186V247Zm101-50c0-52,32-74,76-74,16,0,32,2,41,6v41H380l-4-20c-3-1-6-2-14-2-19,0-31,13-31,46,0,36,16,47,41,47,15,0,27-3,34-6v22c-8,6-23,13-46,13-48,0-72-28-72-73" transform="translate(148 272)"/>\n\t\t\t\t<path d="M737,196l-38-71h30l53,71-53,72H699" transform="translate(-125 274)" fill="#cd1719"/>\n\t\t\t</g>\n\t\t</symbol>\n\t</defs>\n</svg><svg>\n\t<defs>\n\t\t<symbol id="header-icon--search" viewBox="0 0 640.6 770.4">\n\t\t\t<path d="M640.6,716.7L446.7,456c1.6-1.5,3.1-3,4.7-4.6c49.9-49.9,77.4-116.3,77.4-187s-27.5-137-77.4-187\n\tC401.5,27.5,335.1,0,264.4,0s-137,27.5-187,77.4C27.5,127.4,0,193.8,0,264.4s27.5,137,77.4,187c49.9,49.9,116.3,77.4,187,77.4 c37.6,0,74.1-7.8,107.5-22.7l196.5,264.2L640.6,716.7z M90,264.4C90,168.2,168.2,90,264.4,90s174.4,78.2,174.4,174.4 s-78.2,174.4-174.4,174.4S90,360.6,90,264.4z"/>\n\t\t</symbol>\n\t\t<symbol id="header-icon--close" viewBox="0 0 32 32">\n\t\t\t<path d="M32 27.733l-11.733-11.733 11.733-11.733-4.267-4.267-11.733 11.733-11.733-11.733-4.267 4.267 11.733 11.733-11.733 11.733 4.267 4.267 11.733-11.733 11.733 11.733z"/>\n\t\t</symbol>\n\t\t<symbol id="header-icon--user" viewBox="0 0 22.07 22">\n\t\t\t<path d="M14.39 12.6c-2.08-.77-1.25-2.15-.9-2.7l.2-.22C15.65 7.68 17.3 0 11 0 4.77 0 6.37 7.68 8.3 9.69l.2.2c.35.56 1.25 1.94-.9 2.7C4.58 13.14 0 15.01 0 22h22.07c0-6.99-4.64-8.86-7.68-9.4z"/>\n\t\t</symbol>\n\t\t<symbol id="nav-icon--arrow" viewBox="0 0 32 32">\n\t\t\t<path d="M25.5 14L10.4.7a3.4 3.4 0 0 0-4.3 0 2.4 2.4 0 0 0 0 3.8l13 11.4L6 27.2a2.4 2.4 0 0 0 0 3.8 3.3 3.3 0 0 0 4.3 0l15-13.3a2.4 2.4 0 0 0 0-3.7z"/>\n\t\t</symbol>\n\t</defs>\n</svg></div>\n\n\n<nav class="nav-sidebar" tabindex="-1" data-make-links-searchable hidden>\n\t<header class="nav-sidebar__header">\n\n\t\t<button type="button" class="nav-sidebar__header__button nav-sidebar__header__button--back" data-back-button aria-label="Terug">\n\t\t\t<svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t</button>\n\n\t\t<button type="button" class="nav-sidebar__header__button nav-sidebar__header__button--close" data-toggle="nav-sidebar">\n\t\t\tSluit menu <svg><use xlink:href="#header-icon--close"></use></svg>\n\t\t</button>\n\t</header>\n\n\t<div class="nav-sidebar__col" data-nav-col="first" data-state="active">\n\t\t<ul class="nav-sidebar__col__items">\n\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/binnenland/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="binnenland">\n\t\t\t\t\t\tBinnenland\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/buitenland/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="buitenland">\n\t\t\t\t\t\tBuitenland\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/economie/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="economie">\n\t\t\t\t\t\tEconomie\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/den-haag/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="denhaag">\n\t\t\t\t\t\tDen Haag\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/cultuur/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="cultuur">\n\t\t\t\t\t\tCultuur\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/wetenschap/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="wetenschap">\n\t\t\t\t\t\tWetenschap\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/sport/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="sport">\n\t\t\t\t\t\tSport\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/opinie/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="opinie">\n\t\t\t\t\t\tOpinie\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/columnisten/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="columnisten">\n\t\t\t\t\t\tColumnisten\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/tech-en-media/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="tech-en-media">\n\t\t\t\t\t\tTech en Media\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/boeken/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="boeken">\n\t\t\t\t\t\tBoeken\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/leven/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="leven">\n\t\t\t\t\t\tLeven\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/inbeeld/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="inbeeld">\n\t\t\t\t\t\tIn Beeld\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/carriere/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="carriere">\n\t\t\t\t\t\tCarri\xc3\xa8re\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/klimaat/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="klimaat">\n\t\t\t\t\t\tKlimaat\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/recht/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="recht">\n\t\t\t\t\t\tRecht\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/podcast/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="podcast">\n\t\t\t\t\t\tPodcast\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/archief/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="archief">\n\t\t\t\t\t\tArchief\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/tag/luister-naar/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="luister-naar">\n\t\t\t\t\t\tLuister naar\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/rubriek/commentaar/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="commentaar">\n\t\t\t\t\t\tCommentaar\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/rubriek/de-ombudsman/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="ombudsman">\n\t\t\t\t\t\tOmbudsman\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/puzzels/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="puzzels">\n\t\t\t\t\t\tPuzzels\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/index/magazine/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="magazine">\n\t\t\t\t\t\tMagazine\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<button type="button" class="nav-sidebar__button" data-show-nav-col="account">\n\t\t\t\t\t\tService <svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t\t\t\t</button>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="https://nrcwebwinkel.nl/?utm_source=NRC_header&amp;utm_medium=nrc.nl&amp;utm_content=tekstlink&amp;utm_campaign=headernrc"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="webwinkel">\n\t\t\t\t\t\tWebwinkel\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="https://nrcadverteren.nl/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="familieberichten">\n\t\t\t\t\t\tFamilieberichten\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t\t\t<li>\n\t\t\t\t\t<a href="/over-ons/"\n\t\t\t\t\t\tclass="\n\t\t\t\t\t\t\tnav-sidebar__button\n\t\t\t\t\t\t\tnav-sidebar__button--link\n\t\t\t\t\t\t"\n\t\t\t\t\t\tdata-category="menu"\n\t\t\t\t\t\tdata-slug="over-nrc">\n\t\t\t\t\t\tOver NRC\n\t\t\t\t\t</a>\n\t\t\t\t</li>\t\t</ul>\n\t</div>\n\n\t\t\t<div class="nav-sidebar__col" data-nav-col="account" tabindex="-1" data-state>\n\t\t\t\t<h3 class="nav-sidebar__col__header">\n\t\t\t\t\tService\n\t\t\t\t</h3>\n\n\t\t\t\t<ul class="nav-sidebar__col__items nav-sidebar__col__items--categories" data-nav-col-items>\n\t\t\t\t\t\t\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/bezorging" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Bezorgservice</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/onderbreking" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Onderbrekingsservice</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/verhuizing" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Verhuisservice</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/gegevens" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Mijn gegevens</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/abonnement" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Mijn abonnementen</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/digitaal" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">FAQ Digitaal Lezen</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="/nieuwsbrieven/" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Nieuwsbrieven</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://service.nrc.nl/faq" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Veelgestelde vragen</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="/mijn-nrc/leeslijst" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Mijn leeslijst</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="/de/" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Digitale krant</span>\n\t\t\t</a>\n\t\t</li>\n\t\t<li class="nav-sidebar__col__item">\n\t\t\t<a href="https://login.nrc.nl/wachtwoord-vergeten" class="nav-sidebar__button nav-sidebar__button--link">\n\t\t\t\t<span class="nav-sidebar__button__text">Wachtwoord vergeten</span>\n\t\t\t</a>\n\t\t</li>\n\n\t\t\t\t</ul>\n\t\t\t</div>\n</nav>\n<header class="header" data-site-header data-strip-from-headless>\n\t\t<div data-banner-holder="werving-top"></div>\n\n\t\t\t<div class="banner banner--leaderboard "\n\t\tdata-strip-from-headless>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="leaderboard"\n\t\t\t\tdata-req-visitor-state="logged-out"\n\t\t></div>\n\t</div>\n\n\n\t\t<div class="header__main" data-make-links-searchable>\n\t\t\t<div class="header__inner-container"\n\t\t\t\tdata-append="scroll-progress-indicator">\n\n\t\t\t\t<button type="button" class="header__nav-toggle" data-toggle="nav-sidebar">\n<svg width="18" height="14" viewBox="0 0 18 14"><path stroke-width="2.3" d="M0,1.4h18M0,7h18M0,12.6h18"/></svg>\t\t\t\t\t<span>Menu</span>\n\t\t\t\t</button>\n\n\t\t\t\t<a href="/" class="header__logo" aria-label="NRC">\n\t\t\t\t\t<svg preserveAspectRatio="xMidYMid meet" width="110" height="55">\n\t\t\t\t\t\t<use xlink:href="#logo-nrc"></use>\n\t\t\t\t\t\t<desc>NRC</desc>\n\t\t\t\t\t</svg>\n\t\t\t\t</a>\n\n\t\t\t\t<div class="header__buttons">\n\t\t\t\t\t<a href="/mijn-nrc/nieuws/"\n\t\t\t\t\t\tclass="header__button header__button--volgen"\n\t\t\t\t\t\tdata-volgen-add-indicator-dot\n\t\t\t\t\t\tdata-req-visitor-state="logged-in">\n<svg viewBox="0 0 24.4 20.3" width="27"><path d="M12.11 18.3c1.3 0 2.4-1 2.7-2.3h-5.3c.2 1.2 1.3 2.3 2.6 2.3m7.3-2.9c-1.2-1.2-2.1-1.4-2.1-2.8v-4c0-2.3-1.6-4.6-4-5.2 0 0-.1-1.2-1.3-1.2s-1.3 1.2-1.3 1.2c-2.4.6-4 2.9-4 5.2v3.9c0 1.4-.9 1.7-2.1 2.9h14.8m-7.3 4.9c-1.8 0-3.5-1.1-4.3-2.9H.01l3.3-3.4c.4-.4.7-.7 1-.9l.5-.5v-4c0-1.6.6-3.1 1.5-4.4.8-1 1.8-1.8 3-2.3.6-1 1.6-1.6 2.8-1.6 1.2 0 2.2.6 2.8 1.6 2.7 1.1 4.5 3.8 4.5 6.8v4.1l.4.4c.3.3.7.6 1.1 1l3.3 3.4h-7.8a5.1 5.1 0 0 1-4.3 2.7z"/></svg>\t\t\t\t\t\t<span>Mijn nieuws</span>\n\t\t\t\t\t</a>\n\t\t\t\t\t<a href="/index/podcast/"\n\t\t\t\t\t\tclass="header__button header__button--podcasts">\n<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 24 24" width="24" height="24">\n\t<path d="M19.06 11.74h0a3.4 3.4 0 013.4 3.4v3.35a3.4 3.4 0 01-3.4 3.4h-.79V8.42a6.31 6.31 0 10-12.61 0v13.46h-.72a3.4 3.4 0 01-3.4-3.4v-3.34a3.4 3.4 0 013.4-3.4h0" fill="none" stroke-miterlimit="10" stroke-width="2"/>\n</svg>\t\t\t\t\t\t<span>Podcasts</span>\n\t\t\t\t\t</a>\n\t\t\t\t\t<a href="/de/"\n\t\t\t\t\t\tclass="header__button header__button--de"\n\t\t\t\t\t\tdata-de-replace-with-info-icon-on-mobile\n\t\t\t\t\t\tdata-de-add-indicator-dot>\n<svg width="18" preserveAspectRatio="xMidYMid meet" viewBox="0 0 24 32">\n\t<path d="M0 0v32h24V0H0zm21.4 29.3H2.6V2.6h18.7v26.7zM17.3 20H6.7v1.3h10.6V20zm0 4H6.7v1.3h10.6V24zM6.7 14.5l1.3-.3V7.1l-1.3-.2V5.4h4.2l3.1 6.1h.7V7.1l-1.3-.3V5.4h3.9v1.5l-1.3.2V16h-2.7L9.4 8.9h-.1v5.3l1.4.2V16h-4v-1.5z"/>\n</svg>\t\t\t\t\t\t<span>Digitale krant</span>\n\t\t\t\t\t</a>\n\t\t\t\t\t<button type="button" class="header__button header__button--search" data-toggle="search-bar" aria-label="Zoeken">\n\t\t\t\t\t\t<svg width="16" height="32" preserveAspectRatio="xMidYMid meet">\n\t\t\t\t\t\t\t<use xlink:href="#header-icon--search"></use>\n\t\t\t\t\t\t</svg>\n\t\t\t\t\t</button>\n\t\t\t\t\t<button type="button" class="header__button header__button--profile" data-toggle="profile-menu" aria-label="Mijn NRC">\n\t\t\t\t\t\t<div class="header__avatar" data-avatar-initial data-req-visitor-state="logged-in" aria-hidden="true">\n\t\t\t\t\t\t\t<svg width="22" height="22" preserveAspectRatio="xMidYMid meet">\n\t\t\t\t\t\t\t\t<use xlink:href="#header-icon--user"></use>\n\t\t\t\t\t\t\t</svg>\n\t\t\t\t\t\t</div>\n\t\t\t\t\t\t<svg width="22" height="22" preserveAspectRatio="xMidYMid meet" data-req-visitor-state="logged-out">\n\t\t\t\t\t\t\t<use xlink:href="#header-icon--user"></use>\n\t\t\t\t\t\t</svg>\n\t\t\t\t\t</button>\n\t\t\t\t</div>\n<div class="profile-menu" data-profile-menu hidden>\n\t<button type="button" class="profile-menu__header" data-toggle="profile-menu" data-header aria-label="Sluit Mijn NRC">\n\t\t<div class="profile-menu__header__avatar" data-avatar-initial aria-hidden="true">\n\t\t\t<svg width="22" preserveAspectRatio="xMidYMid meet">\n\t\t\t\t<use xlink:href="#header-icon--user"></use>\n\t\t\t</svg>\n\t\t</div>\n\t\t<span class="profile-menu__header__text" data-header-text>\n\t\t\tMijn NRC\n\t\t</span>\n\t</button>\n\n\t<button type="button" class="profile-menu__header profile-menu__header--back-btn" data-back-btn data-show-nav-col="first">\n\t\t<svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t<span class="profile-menu__header__text" data-header-text>\n\t\t\tTerug\n\t\t</span>\n\t</button>\n\n\t<div class="profile-menu__cols">\n\t\t<div class="profile-menu__col" data-nav-col="first" data-state="active">\n\t\t\t<div class="profile-menu__message profile-menu__message--error" data-nav-error>\n\t\t\t\t<div class="profile-menu__message__holder" data-nav-error-messages></div>\n\t\t\t\t<a class="profile-menu__message__button profile-menu__message__button--error" href="https://service.nrc.nl/">\n\t\t\t\t\tLees meer\n\t\t\t\t</a>\n\t\t\t</div>\n\n\t\t\t<div class="profile-menu__col__items">\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/login/"\n\t\t\t\t\tdata-login-link\n\t\t\t\t\tdata-req-visitor-state="logged-out">\n\t\t\t\t\tInloggen\n\t\t\t\t</a>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/mijn-nrc/nieuws/"\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tMijn nieuws\n\t\t\t\t</a>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/mijn-nrc/leeslijst"\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tMijn leeslijst\n\t\t\t\t</a>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/mijn-nrc/recent-gelezen/"\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tMijn geschiedenis\n\t\t\t\t</a>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/de/"\n\t\t\t\t\tdata-req-visitor-state="logged-out-mobile">\n\t\t\t\t\tDigitale krant\n\t\t\t\t</a>\n\n\t\t\t\t<div class="profile-menu__separator" data-req-visitor-state="logged-in"></div>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/profiel/onderwerpen/"\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tVolgonderwerpen\n\t\t\t\t</a>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/profiel/nieuwsbrieven/"\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tNieuwsbrieven\n\t\t\t\t</a>\n\n\t\t\t\t<div data-req-visitor-state="logged-in"\n\t\t\t\t\tdata-req-subscriber-state="non-subscriber"\n\t\t\t\t\tclass="profile-menu__non-subscriber-links">\n\t\t\t\t\t<div class="profile-menu__separator"></div>\n\n\t\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\t\thref="/login/subscriber/?return_to=/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141">\n\t\t\t\t\t\tIk ben abonnee\n\t\t\t\t\t</a>\n\n\t\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\t\thref="https://abonnementen.nrc.nl/nrc-nl/?utm_intern=nrc-web-profielmenu">\n\t\t\t\t\t\tNeem een abonnement\n\t\t\t\t\t</a>\n\t\t\t\t</div>\n\n\t\t\t\t<div class="profile-menu__separator"></div>\n\n\t\t\t\t<button type="button" class="profile-menu__button" data-req-visitor-state="logged-in" data-show-nav-col="instellingen">\n\t\t\t\t\tInstellingen\n\t\t\t\t\t<svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t\t\t</button>\n\n\t\t\t\t<button type="button" class="profile-menu__button" data-show-nav-col="color-scheme-options" hidden>\n\t\t\t\t\tWeergaveopties\n\t\t\t\t\t<svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t\t\t</button>\n\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/logout/"\n\t\t\t\t\tdata-logout-link\n\t\t\t\t\tdata-req-visitor-state="logged-in">\n\t\t\t\t\tUitloggen\n\t\t\t\t</a>\n\n\t\t\t\t<div class="profile-menu__separator" data-req-visitor-state="logged-in"></div>\n\n\t\t\t\t<button class="profile-menu__fixed-message"\n\t\t\t\t\tdata-fixed-message="1202"\n\t\t\t\t\tdata-show-nav-col="service">\n\t\t\t\t\t<h4 class="profile-menu__fixed-message__header">Hoe kunt u ons bereiken?</h4>\n\n\t\t\t\t\t<p>Vanwege het coronavirus werken onze medewerkers thuis.</p>\n\t\t\t\t</button>\n\n\t\t\t\t<button type="button" class="profile-menu__button" data-show-nav-col="service">\n\t\t\t\t\tService\n\t\t\t\t\t<svg><use xlink:href="#nav-icon--arrow"></use></svg>\n\t\t\t\t</button>\n\t\t\t</div>\n\t\t</div>\n\n\t\t<div class="profile-menu__col" data-nav-col="instellingen" tabindex="-1" data-state>\n\t\t\t<div class="profile-menu__col__items">\n\t\t\t\t<a class="profile-menu__button" href="/profiel/instellingen/privacy/">Privacy-instellingen</a>\n\t\t\t\t<a class="profile-menu__button" href="/profiel/instellingen/notificaties/">Notificaties</a>\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/profiel/instellingen/abonnement-delen/"\n\t\t\t\t\tdata-family-settings-link-for="main"\n\t\t\t\t\thidden>\n\t\t\t\t\tAbonnement delen\n\t\t\t\t</a>\n\t\t\t\t<a class="profile-menu__button"\n\t\t\t\t\thref="/profiel/instellingen/abonnement-delen/"\n\t\t\t\t\tdata-family-settings-link-for="sub"\n\t\t\t\t\thidden>\n\t\t\t\t\tGedeeld abonnement\n\t\t\t\t</a>\n\t\t\t</div>\n\t\t</div>\n\n\t\t<div class="profile-menu__col" data-nav-col="color-scheme-options" tabindex="-1" data-state>\n\t\t\t<div class="profile-menu__col__items">\n\t\t\t\t<div class="profile-menu__color-scheme-options" data-color-scheme-options>\n\n\t\t\t\t\t<input type="radio" name="color-scheme-option" id="color-scheme-option-light" value="light" />\n\t\t\t\t\t<label for="color-scheme-option-light">\n\t\t\t\t\t\tLichte weergave\n\t\t\t\t\t</label>\n\n\t\t\t\t\t<input type="radio" name="color-scheme-option" id="color-scheme-option-dark" value="dark" />\n\t\t\t\t\t<label for="color-scheme-option-dark">\n\t\t\t\t\t\tDonkere weergave\n\t\t\t\t\t</label>\n\n\t\t\t\t\t<input type="radio" name="color-scheme-option" id="color-scheme-option-system" value="system" />\n\t\t\t\t\t<label for="color-scheme-option-system" hidden>\n\t\t\t\t\t\tVolg instellingen van mijn apparaat <span data-text="current-system-scheme"></span>\n\t\t\t\t\t</label>\n\t\t\t\t</div>\n\t\t\t</div>\n\t\t</div>\n\n\t\t<div class="profile-menu__col" data-nav-col="service" tabindex="-1" data-state>\n\t\t\t<div class="profile-menu__col__items">\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/bezorging">Bezorgservice</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/onderbreking">Vakantieservice</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/verhuizing">Verhuisservice</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/gegevens">Mijn gegevens</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/abonnement">Mijn abonnementen</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/digitaal">FAQ Digitaal Lezen</a>\n\t\t\t\t<a class="profile-menu__button" href="https://service.nrc.nl/faq">Veelgestelde vragen</a>\n\t\t\t\t<a class="profile-menu__button" href="https://login.nrc.nl/wachtwoord-vergeten">Wachtwoord vergeten</a>\n\t\t\t</div>\n\t\t</div>\n\t</div>\n</div>\t\t\t</div>\n<div class="header__search" data-header-search hidden>\n\t<form name="searchform" action="/search/" method="get" class="header__search__form">\n\t\t<input type="text" name="q" class="header__search__input" placeholder="Zoekterm" autocomplete="off" data-search-input>\n\n\t\t<button class="header__search__button header__search__button--submit" title="Zoeken">\n\t\t\t<svg width="16" height="32" preserveAspectRatio="xMidYMid meet">\n\t\t\t\t<use xlink:href="#header-icon--search"></use>\n\t\t\t</svg>\n\t\t</button>\n\n\t\t<button type="button" class="header__search__button header__search__button--close" data-toggle="search-bar" title="Sluiten">\n\t\t\t<svg width="16" height="32" preserveAspectRatio="xMidYMid meet">\n\t\t\t\t<use xlink:href="#header-icon--close"></use>\n\t\t\t</svg>\n\t\t</button>\n\t</form>\n</div>\n\n<div data-search-suggestions class="search-suggestions"></div>\n\t\t</div>\n\n</header>\n\n\n\n\t<div class="header__de">\n\t\t<div class="header__de__inner">\n\t\t\t<div class="header__de__inner-container">\n\n<a href="/de/" class="de__btn back" data-de-back-button>\n<svg width="8px" height="15px">\n    <path stroke-width="0.5" d="M6.187,5.485 L6.174,5.502 L1.243,0.000 L0.003,1.423 L5.144,7.200 L0.044,13.578 L1.286,15.000 L8.000,7.508 L6.187,5.485 Z"></path>\n</svg>\t<span>Terug naar de krant</span>\n</a>\n\n<a class="de__btn list" data-sidebar-toggle="list">\n\t<span>Lijst</span>\n<svg width="18" height="14" viewBox="0 0 18 14"><path stroke-width="2.3" d="M0,1.4h18M0,7h18M0,12.6h18"/></svg></a>\n\t\t\t</div>\n\t\t</div>\n\t</div>\n\n<div data-user-confirmation-screen-holder data-strip-from-headless></div>\n\n\t<main>\n\t<script>\n\t\twindow.nmt = window.nmt || {};\n\t\twindow.nmt.bottom = window.nmt.bottom || [];\n\t</script>\n\n<div class="paywall hidden" data-paywall>\n\t<div class="paywall__container">\n\t\t<div class="paywall__header">\n\t\t\t<a href="/login/?return_to=%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141%3Ft%3D1646637674" data-appview-href="nrc://login" data-action-login\n\t\t\t\tclass="paywall__header__button">\n\t\t\t\tInloggen\n\t\t\t</a>\n\t\t</div>\n\t\t<div class="paywall__content">\n\t\t\t<div class="paywall__div">\n\t\t\t\t<div class="paywall__view paywall__main" data-paywall-render-main>\n\t\t\t\t\t<!-- render main hier -->\n\t\t\t\t</div>\n\t\t\t\t<div class="paywall__side paywall__view--side" data-paywall-render-sidebar>\n\t\t\t\t\t<!-- render sidebar here  -->\n\t\t\t\t</div>\n\t\t\t</div>\n\t\t</div>\n\t</div>\n</div>\n\n\t\t<div class="nmt-layout nmt-layout--sidebar-align-right">\n\n\t\t\t<div class="nmt-layout__main content-container" data-de-article-btn-container>\n\t\t\t\t<article class="article-outer-container">\n\t\t\t\t\t<div class="inner">\n\n\n<div class="article__header-and-content">\n\n\t<p class="print-layout-warning">N.B. Het kan zijn dat elementen ontbreken aan deze printversie.</p>\n\t\n\n\t<div class="article-header-container intro-inner">\n\n<div class="center-block intro-col article__header">\n\n<div class="luister-naar" data-ui="floating-progressbar">\n\t<video\n\t\tsrc="https://static.nrc.nl/tts/4095141-20220301-143742.mp3"\n\t\tpreload="metadata"\n\t\ttitle="Minder inkomen \xc3\xa9n weinig spaargeld, het is de werkelijkheid voor een half miljoen werkenden"\n\t\tdata-poster="https://images.nrc.nl/rsJp5-SbHACE7AknCKH05lMJpvc=/1200x627/smart/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg"\n\t\tplaysinline\n\t\thidden\n\t></video>\n\t<button type="button" class="luister-naar__start-button" data-ui="start-button" aria-label="Lees dit artikel voor">\n<svg viewBox="0 0 16.43 13.94" width="16.43" height="13.94">\n\t<path d="M2.63 4.06H0v5.86h2.63l5 3.68s.91.77.91 0V.28c0-.62-.8 0-.8 0z"/>\n\t<g data-state="not-muted">\n\t\t<path d="M11.35 2.91a.68.68 0 00-.95 0 .67.67 0 000 .91 4.4 4.4 0 010 6.22.66.66 0 000 .94.68.68 0 00.95 0 5.75 5.75 0 000-8.12z"/>\n\t\t<path d="M13.16.81a.67.67 0 00-.95 1 7.39 7.39 0 010 10.46.67.67 0 00.95.95 8.73 8.73 0 000-12.41z"/>\n\t</g>\n\t<g data-state="muted">\n\t\t<path fill="none" stroke-linecap="round" stroke-miterlimit="10" stroke-width="1.5px" d="M10.91 9.35l4.77-4.77M15.68 9.35l-4.77-4.77"/>\n\t</g>\n  </svg>\n\t\t<h4 class="luister-naar__header">Luister naar</h4>\n\t\t<span class="luister-naar__duration" data-text="total-time"></span>\n\t</button>\n\t<div class="luister-naar__controls" data-ui="controls" hidden>\n\t\t<button type="button" data-btn="play-pause" data-state="paused" disabled aria-label="Pauzeren">\n<svg width="48" height="48" viewBox="0 0 71 71">\n\t<path data-for-state="paused" d="M24.3 23.58a2.89 2.89 0 012.89-2.89 2.94 2.94 0 011.44.39L49.26 33a2.88 2.88 0 010 5L28.63 49.92a2.88 2.88 0 01-3.94-1.06 2.83 2.83 0 01-.39-1.44z"/>\n\t<path data-for-state="playing" stroke-linecap="round" stroke-miterlimit="10" stroke-width="7px" d="M27.74 23.93v23.15M43.26 23.93v23.15"/>\n</svg>\t\t\t<div class="spinner"><span class="spinner-part"></span></div>\n\t\t</button>\n\n\t\t<div class="luister-naar__duration-slider">\n\t\t\t<div class="luister-naar__range-slider" data-ui="range-slider">\n\t\t\t\t<input data-ui="duration-slider" type="range" min="0" max="100" step="0.01" value="0" tabindex="-1" disabled />\n\t\t\t\t<div class="luister-naar__range-slider__progressbar" data-progressbar="duration"></div>\n\t\t\t\t<div class="luister-naar__range-slider__range" data-rangebar></div>\n\t\t\t</div>\n\t\t</div>\n\n\t\t<div class="luister-naar__playback-rate" tabindex="0" data-ui="playback-rate" aria-label="Pas afspeelsnelheid aan">\n\t\t\t1x\n\t\t</div>\n\n\t\t<button type="button" data-btn="close" aria-label="Sluiten">\n\t\t\t<svg viewBox="0 0 34.02 34.02">\n\t\t\t\t<path fill="none" stroke="white" stroke-linecap="round" stroke-width="2px" d="M12.34 21.51l9-9M12.34 12.51l9 9"/>\n\t\t\t</svg>\n\t\t</button>\n\n\t\t\t<a href="#feedbackform" data-toggle="feedbackform">Opmerkingen?</a>\n\t\t\t<a href="/tag/luister-naar/" data-link="luister-meer">Luister meer...</a>\n\t</div>\n</div>\n\n\n\t\t<h1 data-flowtype="headline">Minder inkomen \xc3\xa9n weinig spaargeld, het is de werkelijkheid voor een half miljoen werkenden</h1>\n\n\n\n\t\t<div class="intro article__intro" data-flowtype="intro"><p><span class="keyword">CPB-onderzoek</span> Vooral flexwerkers verloren inkomen in coronajaar 2020. Veel van hen hadden weinig spaargeld om deze tegenslag op te vangen, concludeert het CPB nu.</p></div>\n\n\n<section class="article__byline">\n\t\t<div class="article__byline__inner">\n\t\t\t<div class="article__byline__list">\n\n\t\t\t\t\t<div class="article__byline__list__link article__byline__list__link__author">\n\t\t\t\t\t\t<span class="article__byline__icon"><svg width="14" height="14" viewBox="0 0 300 300"><path d="M253 123l-77-77 46-46 77 77-46 46zm-92-61l77 77s-35 16-46 77c-62 62-123 62-123 62s-24 36-46 15l93-94c55 12 50-39 37-52s-62-21-52 37L7 277c-21-21 15-46 15-46s0-62 62-123c51-5 77-46 77-46z"/></svg></span>\n\t\t\t\t\t\t<ul class="article__byline__text unstyled">\n<li><a href="/search/?author_filter=Christiaan%20Pelgrim" rel="author">Christiaan Pelgrim</a></li>\t\t\t\t\t\t</ul>\n\t\t\t\t\t</div>\n\n\t\t\t\t\t<a class="article__byline__list__link"\n\t\t\t\t\t\thref="/nieuws/2022/02/28/">\n\t\t\t\t\t\t<span class="article__byline__icon article__byline__icon__clock"><svg width="14" height="14" viewBox="0 0 60 60"><g fill="none"><circle cx="50%" cy="50%" r="24.8" stroke-width="6"/><path d="M30,17 30,30.5 40,38.4" stroke-width="4.4" stroke-linecap="round"/></g></svg></span>\n\t\t\t\t\t\t<time class="article__byline__text" datetime="2022-02-28T21:45:14+02:00">\n\t\t\t\t\t\t\t28 februari 2022\n\t\t\t\t\t\t</time>\n\t\t\t\t\t</a>\n\n\t\t\t\t\t<div class="article__byline__list__link">\n\t\t\t\t\t\t<span class="article__byline__icon">\n<svg width="14" height="14" viewBox="0 0 32 32">\n\t<path d="M16 6.038v-2.038h4v-2c0-1.105-0.895-2-2-2h-6c-1.105 0-2 0.895-2 2v2h4v2.038c-6.712 0.511-12 6.119-12 12.962 0 7.18 5.82 13 13 13s13-5.82 13-13c0-6.843-5.288-12.451-12-12.962zM22.071 26.071c-1.889 1.889-4.4 2.929-7.071 2.929s-5.182-1.040-7.071-2.929c-1.889-1.889-2.929-4.4-2.929-7.071s1.040-5.182 2.929-7.071c1.814-1.814 4.201-2.844 6.754-2.923l-0.677 9.813c-0.058 0.822 0.389 1.181 0.995 1.181s1.053-0.36 0.995-1.181l-0.677-9.813c2.552 0.079 4.94 1.11 6.754 2.923 1.889 1.889 2.929 4.4 2.929 7.071s-1.040 5.182-2.929 7.071z"></path>\n</svg>\t\t\t\t\t\t</span>\n\t\t\t\t\t\t<time class="article__byline__text" duration="P117S">\n\t\t\t\t\t\t\tLeestijd 2 minuten\n\t\t\t\t\t\t</time>\n\t\t\t\t\t</div>\n\t\t\t</div>\n\t\t</div>\n\n\t\t<div class="article__byline__social">\n\t\t\t\n\n\n\n\t\t<div class="share share--in-byline share--compact" \n\t\t\tdata-share-widget\n\t\t\tdata-strip-from-headless>\n\t\t\t<div class="share__inner"\n\t\t\t\t\tdata-de-add-article-download-button\n>\n<a href="https://wa.me/?text=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141%3Futm_source%3DNRC%26utm_medium%3Dwhatsapp" title="Deel dit artikel met een WhatsApp-contact" class="share__button share__whatsapp " data-type="whatsapp">\n<svg viewBox="0 0 440 440" width="44" height="44"><path d="M232 90c-65 0-118 52-118 117 0 32 17 61 17 61l-21 62 65-20s24 14 57 14c65 0 118-52 118-117S297 90 232 90zm55 163c-7 8-22 11-29 9s-30-8-46-22c-15-14-28-29-33-40s-5-17-4-20c0-3 2-17 11-23 0 0 3-2 5-2h9s2 0 4 3c1 3 9 21 10 22s2 5 0 8-8 9-8 9-2 2 0 5 8 12 16 20c8 7 18 12 23 14s6-1 8-3 8-10 8-10 2-3 6-1 23 11 23 11 2 0 2 3c1 1 2 9-5 17z"/></svg>\t\t\t\t\t</a><a href="https://www.facebook.com/sharer/sharer.php?u=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op Facebook" class="share__button share__facebook share--pop-me-up" data-type="facebook">\n<svg width="44" height="44" viewBox="0 0 900 900"><path d="M495 693h-90V469h-75v-87h75v-64c0-37 11-65 32-86 20-19 46-29 80-29 19 0 41 1 67 3v78h-47c-28 0-42 13-42 38v60h86l-11 87h-75v224z"/></svg>\t\t\t\t\t</a><a href="mailto:?body=Op%20nrc.nl%20las%20ik%20het%20volgende%20interessante%20artikel%3A%20https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141&amp;subject=Interessant%20artikel%20op%20nrc.nl" title="Mail dit artikel naar een vriend(in)" class="share__button share__mail " data-type="mail">\n<svg width="44" height="44" viewBox="-52 -74 250 250"><path d="M140 86L93 41l47-37v82zM5 0h130L70 53 5 0zM0 86V4l46 37L0 86zm68-28c1 0 1 1 2 1s1 0 2-1l17-14 47 46H5l47-45 16 13z"/></svg>\t\t\t\t\t</a><a href="https://twitter.com/share?text=Minder%20inkomen%20%C3%A9n%20weinig%20spaargeld%2C%20het%20is%20de%20werkelijkheid%20voor%20een%20half%20miljoen%20werkenden&amp;url=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op Twitter" class="share__button share__twitter share--pop-me-up" data-type="twitter">\n<svg viewBox="0 0 262 262" width="44" height="44"><path d="M201 91s-5 9-14 14c2 39-27 82-78 83-26 0-42-13-43-13s20 4 41-11c-21-1-26-19-26-20s6 2 13 0c-23-6-22-28-22-28s6 4 12 4c-21-16-9-37-8-37s19 27 57 29c-6-30 29-44 47-25 9-2 18-7 18-7s-3 10-13 15c8-1 16-4 16-4z"/></svg>\t\t\t\t\t</a><a href="https://www.linkedin.com/cws/share?lang=nl_NL&amp;url=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op LinkedIn" class="share__button share__linkedin share--pop-me-up" data-type="linkedin">\n<svg width="44" height="44" viewBox="0 0 440 440"><path d="M322 330H118c-10 0-18-8-18-17V107c0-9 8-17 18-17h204c10 0 18 8 18 17v206c0 9-8 17-18 17zm-186-39h37V183h-37v108zm19-161c-13 0-21 8-21 19 0 10 8 19 20 19h1c12 0 20-9 20-19 0-11-8-19-20-19zm149 99c0-34-18-49-42-49-19 0-28 10-33 18v-15h-36v108h36v-61c0-3 0-6 1-8 3-7 9-14 19-14 13 0 18 10 18 25v58h37v-62z"/></svg>\t\t\t\t\t</a><button href="" title="Kopieer artikel-url naar Klembord" class="share__button share__clipboard " data-type="clipboard">\n<svg viewBox="-6 -6 36 36" width="24" height="24">\n\t<path d="M17 9H7V7h10m0 6H7v-2h10m-3 6H7v-2h7M12 3a1 1 0 011 1 1 1 0 01-1 1 1 1 0 01-1-1 1 1 0 011-1m7 0h-4.18A3.01 3.01 0 0012 1c-1.3 0-2.4.84-2.82 2H5a2 2 0 00-2 2v14a2 2 0 002 2h14a2 2 0 002-2V5a2 2 0 00-2-2z"/>\n</svg>\t\t\t\t\t</button>\n\t\t\t\t\t<button type="button" class="share__button share__button--expand" data-expand-share\n\t\t\t\t\t\ttitle="Meer opties"\n\t\t\t\t\t\tdata-expanded-title="Minder opties">\n<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 15 3" width="15" height="3" class="share__svg share__svg--dots">\n  <circle cx="1.5" cy="1.5" r="1.5"/>\n  <circle cx="7.5" cy="1.5" r="1.5"/>\n  <circle cx="13.5" cy="1.5" r="1.5"/>\n</svg><svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 14.2 14.2" class="share__svg share__svg--close">\n\t<path d="M14.2 1.6L12.6 0 7.1 5.5 1.6 0 0 1.6l5.5 5.5L0 12.6l1.6 1.6 5.5-5.4 5.5 5.4 1.6-1.6-5.4-5.5z"/>\n</svg>\t\t\t\t\t</button>\n\n\t\t\t\t\t<button type="button" class="share__button share__button--leeslijst leeslijst__article-button" data-leeslijst-button="add" data-article-id="4095141" data-title-add="Toevoegen aan leeslijst" data-title-remove="Verwijder uit leeslijst" title="Toevoegen aan leeslijst">\n<svg viewBox="0 0 9 15" class="share__svg share__svg--leeslijst">\n   <path d="M0,0h9v15l-4.5-4L0,15V0z"/>\n </svg>\t\t\t\t\t</button>\n\n\t\t\t</div>\n\t\t</div>\n\n\t\t</div>\n</section></div>\t</div>\n\n\n<figure class="\n\tarticle__featured-image\n\tarticle__featured-image--landscape\n\tarticle__featured-image--below-header\n"\n\tdata-featured-image\n\tdata-fixedimagepoint="center">\n\n\t\t\t\n\n\n\n\n\n\n\t<div class="responsive-img-div "\n\t\tdata-open-in-lightbox\n\t\tdata-aspect-ratio="0.667334000667334"\n\t\tdata-lightbox-highres="https://images.nrc.nl/HzbVBy8v1eP2D2RSbO81JdcPllI=/1920x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg"\n\t\tdata-lightbox-highres-webp="https://images.nrc.nl/vQCRBPH2gpVudO7Se-tAv6wIGnY=/1920x/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg"\n\t\tdata-lightbox-thumbnail="https://images.nrc.nl/pN1aSGBc2pKvOr5rXD6DkbEFSgQ=/160x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg">\n\t\t\t\t\t<div class="placeholder"\n\t\t\t\tstyle="padding-bottom: 66.7334000667334%"></div>\n\n\n\t\t<picture>\n\t\t\t<source srcset="\t\thttps://images.nrc.nl/vQCRBPH2gpVudO7Se-tAv6wIGnY=/1920x/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 1920w,\t\thttps://images.nrc.nl/45t08PBLEWy4C78e66uaVVMrTKA=/1280x/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 1280w,\t\thttps://images.nrc.nl/fL-jA29llX18FxkORHVSAR1jiVE=/640x/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 640w,\t\thttps://images.nrc.nl/mdSyhwGneVw_2Kk2xY36ySnmKOk=/320x/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 320w" type="image/webp">\n\t\t\t<img srcset="\t\thttps://images.nrc.nl/HzbVBy8v1eP2D2RSbO81JdcPllI=/1920x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 1920w,\t\thttps://images.nrc.nl/BIo-YXvqs8YWZHSGLNvZO2AHUMY=/1280x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 1280w,\t\thttps://images.nrc.nl/M0j46Oh1nufxicEbiqS_F-8hGn0=/640x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 640w,\t\thttps://images.nrc.nl/nKIOsx8Sdi1vJObsaJLFtcOVbtg=/320x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg 320w" sizes="\t\t(min-width: 1408px) 988px,\n\t\t(min-width: 980px) calc(100vw - 336px - 2*12px),\n\t\t(min-width: 660px) calc(100vw - 4*12px),\n\t\t(min-width: 500px) calc(100vw - 2*12px),\n\t\t100vw\n" src="https://images.nrc.nl/BIo-YXvqs8YWZHSGLNvZO2AHUMY=/1280x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg" decoding="async" alt="In de horeca werden werknemers hard geraakt door virusuitbraak">\n\t\t</picture>\n\t</div>\n\n\n\n\t\t<figcaption class="article__featured-image__caption" data-is-lightbox-caption>\n\t\t\t\t<span class="caption-text article__featured-image__caption__text">In de horeca werden werknemers hard geraakt door  virusuitbraak</span>\n\t\t\t\t<span class="caption-producer article__featured-image__caption__producer has-inline-svg">\n\t\t\t\t\tFoto Olivier Middendorp\n<svg width="20" height="10" viewBox="0 0 240 165"><path d="M222 165H17c-10 0-18-8-18-17V37c0-10 8-17 18-17h8V9c0-5 6-9 10-9h35c5 0 8 5 8 10v10h144c10 0 18 8 18 17v111c0 10-8 17-18 17zM127 34c-31 0-57 26-57 57 0 32 26 58 57 58s57-26 57-58c0-31-26-57-57-57zm0 97c-22 0-39-18-39-40s18-40 39-40c22 0 39 18 39 40s-17 40-39 40z"/></svg>\t\t\t\t</span>\n\t\t</figcaption>\n</figure>\n\t<div class="article-container main ">\n\n\n\n\n\n\t\t<div class="content article__content">\n\t\t\t<p>Werknemers die in 2020 de zwaarste inkomensdaling doormaakten, hadden ook de kleinste financi\xc3\xable buffers om dat verlies op te vangen. Dat concludeert het Centraal Planbureau (CPB) deze dinsdag in een gedetailleerde studie van salaris- en vermogensgegevens.</p><p>Ongeveer een half miljoen werknemers kregen in het eerste coronajaar te maken met een combinatie van fors inkomensverlies \xc3\xa9n het ontbreken van spaargeld. Zo\xe2\x80\x99n 1 miljoen werknemers \xe2\x80\x93 vooral flexkrachten \xe2\x80\x93 maakten in 2020 een sterke inkomensdaling door, van minstens 10 procent, schrijft het Planbureau. De helft van hen had minder dan 10.000 euro aan spaargeld achter de hand. En hoe groter het inkomensverlies, hoe minder spaartegoed werknemers gemiddeld hadden.</p><a class="lees-ook" href="//www.nrc.nl/nieuws/2021/12/28/grotere-kloof-tussen-laag-en-hoogopgeleide-jongeren-door-corona-a4072540">\n<picture class="lees-ook__image">\n<source srcset="https://images.nrc.nl/qFpE0gUd2P6dwCWtEsUs7RoUjh8=/640x384/smart/filters:no_upscale():format(webp)/s3/static.nrc.nl/images/gn4/stripped/data80141887-b5666a.jpg" type="image/webp"/>\n<img alt="" src="https://images.nrc.nl/zrcB6rYum92uVOL-VTsyrzHud9k=/640x384/smart/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data80141887-b5666a.jpg"/>\n</picture>\n<span class="lees-ook__content">\n\t\tLees ook: <strong>Grotere kloof tussen laag- en hoogopgeleide jongeren door corona</strong>\n</span>\n</a><p>Hoewel veel van hen konden terugvallen op het inkomen van een partner, kon een op de drie werknemers met inkomensverlies dat niet. Hun partner heeft geen of een laag inkomen, of ze hebben geen partner.</p><p>\xe2\x80\x9eDeze cijfers zeggen iets over de kans dat mensen in de problemen zijn gekomen\xe2\x80\x9d, zegt CPB-onderzoeker Diederik Dicou. \xe2\x80\x9eWij kunnen bijvoorbeeld niet zien of mensen in de schulden zijn beland. Maar als je een grote teruggang in inkomen en weinig buffers ziet, dan weet je dat er re\xc3\xable problemen zijn veroorzaakt.\xe2\x80\x9d</p><p>Bij deze cijfers past een kanttekening: het CPB kijkt alleen naar bruto-inkomens, en laat het effect van belasting, uitkeringen en toeslagen buiten beschouwing.</p><p>Ook in \xe2\x80\x98normalere\xe2\x80\x99 jaren zijn er werknemers met meer dan 10 procent inkomensverlies, bijvoorbeeld door ontslag. Het aantal mensen dat hier in 2020 mee te maken kreeg lag zo\xe2\x80\x99n 15 procent hoger dan in 2019. </p><h2 class="gn4-crosshead">Hard geraakt</h2><p>Vooral flexwerkers werden hard geraakt door de economische klap na de virusuitbraak. In de horeca-, recreatie- en cultuursector werden veel flexkrachten naar huis gestuurd. Dat ziet het CPB terug in de gemiddelde inkomensontwikkeling van uitzend- en oproepkrachten. Die was in april 2020 \xe2\x80\x93 het dieptepunt \xe2\x80\x93 11 procentpunt achtergebleven bij het jaar ervoor. Bij de volledige groep werknemers zag het CPB een veel kleinere achterstand, nog geen 2 procentpunt.</p><p>In de tweede helft van 2020 verbeterden de inkomens van beide groepen weer, dankzij snel economisch herstel, maar het gat tussen \xe2\x80\x98vast\xe2\x80\x99 en \xe2\x80\x98flex\xe2\x80\x99 bleef groot. </p><p>\xe2\x80\x9eDe tragiek van deze crisis\xe2\x80\x9d, zegt Dicou, \xe2\x80\x9eis dat die het hardst is aangekomen in de sectoren waar veel flexibele arbeidskrachten werken. Dat maakt de klap extra hard.\xe2\x80\x9d </p><a class="lees-ook" href="//www.nrc.nl/nieuws/2020/05/08/de-coronacrisis-is-een-onvoorstelbare-gamechanger-voor-de-arbeidsmarkt-a3999151">\n<picture class="lees-ook__image">\n<source srcset="https://images.nrc.nl/8ozsimBX1p6YzUZu_0hpZsW0a5Q=/640x384/smart/filters:no_upscale():format(webp)/s3/static.nrc.nl/bvhw/files/2020/05/data57982802-6c5dac.jpg" type="image/webp"/>\n<img alt="" src="https://images.nrc.nl/I4YT_X0f6djmOOHdSsUeVeXyF58=/640x384/smart/filters:no_upscale()/s3/static.nrc.nl/bvhw/files/2020/05/data57982802-6c5dac.jpg"/>\n</picture>\n<span class="lees-ook__content">\n\t\tLees ook: <strong>De coronacrisis is een \xe2\x80\x98onvoorstelbare gamechanger\xe2\x80\x99 voor de arbeidsmarkt</strong> (2020)\n\t</span>\n</a><p>Daar komt bij dat de steunregelingen van de overheid, met name loonsubsidie NOW, vooral gericht waren op het in dienst houden van werkenden met een vast contract. Dicou: \xe2\x80\x9eZij hebben per saldo geen extreme schok ervaren.\xe2\x80\x9d</p><p>En dat terwijl flexwerkers gemiddeld minder spaargeld hebben \xc3\xa9n minder inkomensbescherming: zij hebben vaak minder lang recht op de wat ruimere WW-uitkering. Daardoor vallen zij sneller terug op de sobere bijstandsuitkering.</p><h2 class="gn4-crosshead">\xe2\x80\x98Kloof verkleinen\xe2\x80\x99</h2><p>Voor het CPB is daarom duidelijk dat \xe2\x80\x9estructurele hervorming\xe2\x80\x9d van de arbeidsmarkt noodzakelijk is, bijvoorbeeld door de kloof tussen vast en flexibel werk te verkleinen. Ook zou het kabinet kwetsbare huishoudens kunnen stimuleren meer financi\xc3\xable buffers op te bouwen. </p><p>Het CPB  wijst op verschillende regels die sparen \xe2\x80\x9eontmoedigen\xe2\x80\x9d, zoals de \xe2\x80\x9estrenge vermogensgrenzen\xe2\x80\x9d voor mensen die een bijstandsuitkering of kwijtschelding van gemeentelijke belastingen aanvragen. </p><p>Ook zelfstandig ondernemers werden relatief hard geraakt in 2020, maar zij hadden gemiddeld meer financi\xc3\xable reserves om die klap op te vangen dan flexibele werknemers, zag het CPB. Dat is niet verwonderlijk. \xe2\x80\x9eZij hebben die buffers ook hard nodig\xe2\x80\x9d, zegt onderzoeker Dicou, \xe2\x80\x9eomdat ze bijvoorbeeld vaak geen pensioen hebben. Als ze op hun vermogen moeten interen, kan dat in de toekomst problemen veroorzaken.\xe2\x80\x9d</p>\n<aside class="newslettersignup belowarticlesignup">\n\t<p class="call-to-action-title">\n\t\t<span>Nieuwsbrief</span>\n\t\tNRC Vandaag\n\t</p>\n\t<p class="\n\t\tcall-to-action-description\n\t\t">\n\t\tElke ochtend een overzicht van onze beste stukken en al het belangrijke andere nieuws\n\t</p>\n\n\t<form method="post" target="_top" action="/newsletters/inline-subscribe" data-newsletter-name="NRC Vandaag">\n\t\t<input type="hidden" name="return_to" value="http://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141#msg=subscribed-via-inline-form" />\n\t\t<input type="hidden" name="OPTIN" value="nrcvandaag" />\n\t\t<input type="hidden" name="SUBSCRIBE_SOURCE" value="belowarticlesignup" />\n\t\t<input type="email" name="MAIL" placeholder="Jouw e-mailadres" title="Vul je e-mailadres in" autocorrect="off" autocapitalize="off" required />\n\t\t<button>Inschrijven</button>\n\t</form>\n</aside>\n\n\n\n\t\t<div class="article__published-in"\n\t\t\tdata-valid-permissions-for-show="nrch-digitale-editie-zesdaags,nrcnext-digitale-editie"\n\t\t\t>\n\t\t\tEen versie van\n\t\t\t\tdit artikel\n\t\t\tverscheen ook in\n\t\t\t<a href="/handelsblad/2022/03/01/#306" data-published-in-link>\n\t\t\t\tNRC Handelsblad\n\t\t\t\tvan 1 maart 2022\n\t\t\t</a>\n\t\t</div>\n\t\t<div class="article__published-in"\n\t\t\tdata-valid-permissions-for-show="nrch-digitale-editie-zesdaags,nrcnext-digitale-editie"\n\t\t\t>\n\t\t\tEen versie van\n\t\t\t\tdit artikel\n\t\t\tverscheen ook in\n\t\t\t<a href="/ochtend/2022/03/01/#306" data-published-in-link>\n\t\t\t\tNRC in de ochtend\n\t\t\t\tvan 1 maart 2022\n\t\t\t</a>\n\t\t</div>\n\n\t\t</div>\n\n\n\n<footer class="article__footer">\n\n\n\t\t\t\t<section class="metabox metabox--in-footer"\n\t\t\tdata-metabox="footer"\n\t\t\tdata-initial-metabox-target="onderwerpen"\n\t\t>\n\t\t\t\t<div class="metabox__group" data-metabox-group data-strip-from-headless>\n\t\t\t\t\t<button type="button" class="metabox__list__link" data-toggle="collecties">\n\t\t\t\t\t\t\t<span class="metabox__icon metabox__icon--leeslijst">\n<svg viewBox="0 0 9 15" class="share__svg share__svg--leeslijst">\n   <path d="M0,0h9v15l-4.5-4L0,15V0z"/>\n </svg>\t\t\t\t\t\t\t</span>\n\t\t\t\t\t\t\t<span class="metabox__text">Leeslijst</span>\n\t\t\t\t\t</button>\n\n\t\t\t\t\t\t<button type="button" class="metabox__list__link metabox__list__link--share" data-toggle="share">\n\t\t\t\t\t\t\t<span class="metabox__icon metabox__icon--share">\n<svg viewBox="0 0 19.6 10.45"><path d="M19.6 5.23L13.07 0v2.61C1.3 2.61 0 9.15 0 10.45c2.61-5.22 11.76-2.61 13.07-2.61v2.61z"/></svg>\t\t\t\t\t\t\t</span>\n\t\t\t\t\t\t\t<span class="metabox__text">Artikel delen</span>\n\t\t\t\t\t\t</button>\n\t\t\t\t</div>\n\t\t\t<div class="metabox__group" data-metabox-group>\n\t\t\t\t\t<a href="#feedbackform" class="metabox__list__link" data-toggle="feedbackform" data-strip-from-headless>\n\t\t\t\t\t\t<span class="metabox__icon">\n<svg width="14" height="9" viewBox="0 0 140 90"><path d="M140 86L93 41l47-37v82zM5 0h130L70 53 5 0zM0 86V4l46 37L0 86zm68-28c1 0 1 1 2 1s1 0 2-1l17-14 47 46H5l47-45 16 13z"/></svg>\t\t\t\t\t\t</span>\n\t\t\t\t\t\t<span class="metabox__text">Mail de redactie</span>\n\t\t\t\t\t</a>\n\n\t\t\t\t\t<button type="button" class="metabox__list__link" data-toggle="onderwerpen">\n\t\t\t\t\t\t<span class="metabox__icon">\n<svg version="1.1" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px">\n\t<path d="M6.6,15c1,0,1.9-0.8,1.9-1.9H4.7C4.7,14.2,5.5,15,6.6,15z"/>\n\t<path d="M11.3,10.2c0-3.3-0.1-6.2-3-7.9C8.3,0.8,8,0,6.5,0C5,0,4.7,0.8,4.7,2.3c-2.9,1.7-3,4.6-3,7.9H0v2h13v-2H11.3z"/>\n</svg>\t\t\t\t\t\t</span>\n\t\t\t\t\t\t<span class="metabox__text">Volgonderwerpen</span>\n\t\t\t\t\t</button>\n\t\t\t</div>\n\t\t\t<div class="metabox__line" data-line>\n\t\t\t\t<div class="metabox__line__item" data-line-item></div>\n\t\t\t</div>\n\t\t\t<div class="metabox__target-items" data-target-items>\n\t\t\t\t\t<div class="metabox__target" data-target="feedbackform"\n\t\t\t\t\t\t\tdata-strip-from-headless\n\t\t\t\t\t\t>\n\n<form data-feedback-form\n\tid="feedbackform"\n\tclass="feedback-form feedback-form--below-article"\n\taction="/contact/"\n\tmethod="post">\n\n\t<input type="text" name="nrcform_man_or_bot" class="honing" placeholder="Niet invullen, alleen voor bots" title="Alleen voor bots">\n\n\t\t<input type="hidden" name="nrcform_article_id" value="4095141">\n\t\t<header class="feedback-form__header">\n\t\t\t<a class="feedback-form__header__close" href="#" data-toggle="feedbackform">\n<svg height="16" width="16" viewBox="0 0 256 256"><path stroke-linecap="round" stroke-width="36" d="M50,50L206,206M50,206L206,50"/></svg>\t\t\t</a>\n\t\t\t<h3 class="feedback-form__header__title">\n\t\t\t\tOpmerkingen? Mail ons\n\t\t\t</h3>\n\t\t\t<p class="feedback-form__header__intro">\n\t\t\t\tHeeft u een tip over dit onderwerp, ziet u een spelfout of feitelijke onjuistheid?\n\t\t\t\tWe stellen het zeer op prijs als u ons daarover een bericht stuurt.\n\t\t\t\tU kunt ons ook <a href="/anonieme-tips/">anoniem een tip geven</a>.\n\t\t\t</p>\n\t\t</header>\n\n\t<div class="feedback-form__main">\n\t\t<fieldset class="bestemming">\n\t\t\t\t<label for="nrcform_destination">Betreft:</label>\n\t\t\t\t<fieldset class="radio-list">\n\t\t\t\t\t\t\t<label class="radio-wrapper">\n\t\t\t\t\t\t\t\t<input type="radio" name="nrcform_destination" value="Taalfout of feitelijke onjuistheid" required>\n\t\t\t\t\t\t\t\tTaalfout of feitelijke onjuistheid\n\t\t\t\t\t\t\t</label>\n\t\t\t\t\t\t\t<label class="radio-wrapper">\n\t\t\t\t\t\t\t\t<input type="radio" name="nrcform_destination" value="Opmerking over het artikel" required>\n\t\t\t\t\t\t\t\tOpmerking over het artikel\n\t\t\t\t\t\t\t</label>\n\t\t\t\t\t\t<label class="radio-wrapper">\n\t\t\t\t\t\t\t<input type="radio" name="nrcform_destination" value="Opmerking over het voorgelezen artikel" required>\n\t\t\t\t\t\t\tOpmerking over het voorgelezen artikel\n\t\t\t\t\t\t</label>\n\t\t\t\t</fieldset>\n\t\t</fieldset>\n\n\t\t<fieldset class="reactie">\n\t\t\t<div>\n\t\t\t\t<label for="content">Bericht:</label>\n\t\t\t\t<textarea name="nrcform_message"\n\t\t\t\t\tid="content" cols="50" rows="7" title="Schrijf hier uw tip of opmerking" required></textarea>\n\t\t\t</div>\n\t\t</fieldset>\n\n\t\t<fieldset class="author_name">\n\t\t\t<label for="author_name">Naam:</label>\n\t\t\t<input class="field" type="text" id="author_name" name="nrcform_name"\n\t\t\t\tautocapitalize="off" autocorrect="off" title="Vul hier je naam in" required>\n\t\t</fieldset>\n\n\t\t<fieldset class="author_email">\n\t\t\t<label for="author_email">E-mailadres:</label>\n\t\t\t<input class="field" type="email" id="author_email" name="nrcform_email"\n\t\t\t\tautocapitalize="off" autocorrect="off" title="Vul hier je e-mailadres in" required>\n\t\t</fieldset>\n\n\t\t<input type="hidden" id="cdest" name="nrcform_cdest">\n\n\t\t<fieldset>\n\t\t\t<label></label>\n\t\t\t<div>\n\t\t\t\t<button\n\t\t\t\t\tclass="feedback-form__submit-button"\n\t\t\t\t\tdata-feedback-submit-button\n\t\t\t\t\tdata-label-orig="Verzenden"\n\t\t\t\t\tdata-label-sending="Uw bericht wordt verzonden..."\n\t\t\t\t\tdata-label-sent="Uw bericht is verzonden"\n\t\t\t\t\ttype="submit"\n\t\t\t\t\tname="feedbacksubmit">Verzenden</button>\n\t\t\t</div>\n\t\t</fieldset>\n\n\t\t<fieldset class="feedback-form__result-msg-holder">\n\t\t\t<p class="feedback-form__result-msg" data-feedback-result-msg></p>\n\t\t</fieldset>\n\t</div>\n</form>\t\t\t\t\t</div>\n\t\t\t\t\t<div class="metabox__target" data-target="collecties" data-strip-from-headless>\n\t\t\t\t\t\t<button type="button"\n\t\tclass="metabox__list__link"\n\t\tdata-leeslijst-button="add"\n\t\tdata-article-id="4095141"\n\t\tdata-title-add="Opslaan in leeslijst"\n\t\tdata-title-remove="Opgeslagen"\n\t\ttitle="Opslaan in leeslijst">\n\t\t<span class="metabox__icon metabox__icon--leeslijst">\n<svg viewBox="0 0 9 15" class="share__svg share__svg--leeslijst">\n   <path d="M0,0h9v15l-4.5-4L0,15V0z"/>\n </svg>\t\t</span>\n\t\t<span class="metabox__text" data-leeslijst-link-text>Opslaan in leeslijst</span>\n\t</button>\n\t\t\t\t\t\t<div class="nmt-item__tag-collection" data-tag-collection data-tag-enabled="false">\n\t\t\t\t\t\t\t<div class="nmt-item__tag__info">Toevoegen aan</div>\n\n\t\t\t\t\t\t\t<div class="nmt-item__tag__created" data-created-tags>\n\n\t\t\t\t\t\t\t\t\t<div class="nmt-item__tag__label" data-action-add-collection><span>+</span> verzameling</div>\n\n\t\t\t\t\t\t\t</div>\n\n\n\t\t\t\t\t\t</div>\n\t\t\t\t\t</div>\n\t\t\t\t\t<div class="metabox__target" data-target="share" data-strip-from-headless>\n\t\t\t\t\t\t\n\n\n\n\t\t<div class="share" \n\t\t\tdata-share-widget\n\t\t\tdata-strip-from-headless>\n\t\t\t<div class="share__inner share__inner--expanded"\n>\n<a href="https://wa.me/?text=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141%3Futm_source%3DNRC%26utm_medium%3Dwhatsapp" title="Deel dit artikel met een WhatsApp-contact" class="share__button share__whatsapp " data-type="whatsapp">\n<svg viewBox="0 0 440 440" width="44" height="44"><path d="M232 90c-65 0-118 52-118 117 0 32 17 61 17 61l-21 62 65-20s24 14 57 14c65 0 118-52 118-117S297 90 232 90zm55 163c-7 8-22 11-29 9s-30-8-46-22c-15-14-28-29-33-40s-5-17-4-20c0-3 2-17 11-23 0 0 3-2 5-2h9s2 0 4 3c1 3 9 21 10 22s2 5 0 8-8 9-8 9-2 2 0 5 8 12 16 20c8 7 18 12 23 14s6-1 8-3 8-10 8-10 2-3 6-1 23 11 23 11 2 0 2 3c1 1 2 9-5 17z"/></svg>\t\t\t\t\t</a><a href="https://www.facebook.com/sharer/sharer.php?u=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op Facebook" class="share__button share__facebook share--pop-me-up" data-type="facebook">\n<svg width="44" height="44" viewBox="0 0 900 900"><path d="M495 693h-90V469h-75v-87h75v-64c0-37 11-65 32-86 20-19 46-29 80-29 19 0 41 1 67 3v78h-47c-28 0-42 13-42 38v60h86l-11 87h-75v224z"/></svg>\t\t\t\t\t</a><a href="mailto:?body=Op%20nrc.nl%20las%20ik%20het%20volgende%20interessante%20artikel%3A%20https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141&amp;subject=Interessant%20artikel%20op%20nrc.nl" title="Mail dit artikel naar een vriend(in)" class="share__button share__mail " data-type="mail">\n<svg width="44" height="44" viewBox="-52 -74 250 250"><path d="M140 86L93 41l47-37v82zM5 0h130L70 53 5 0zM0 86V4l46 37L0 86zm68-28c1 0 1 1 2 1s1 0 2-1l17-14 47 46H5l47-45 16 13z"/></svg>\t\t\t\t\t</a><a href="https://twitter.com/share?text=Minder%20inkomen%20%C3%A9n%20weinig%20spaargeld%2C%20het%20is%20de%20werkelijkheid%20voor%20een%20half%20miljoen%20werkenden&amp;url=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op Twitter" class="share__button share__twitter share--pop-me-up" data-type="twitter">\n<svg viewBox="0 0 262 262" width="44" height="44"><path d="M201 91s-5 9-14 14c2 39-27 82-78 83-26 0-42-13-43-13s20 4 41-11c-21-1-26-19-26-20s6 2 13 0c-23-6-22-28-22-28s6 4 12 4c-21-16-9-37-8-37s19 27 57 29c-6-30 29-44 47-25 9-2 18-7 18-7s-3 10-13 15c8-1 16-4 16-4z"/></svg>\t\t\t\t\t</a><a href="https://www.linkedin.com/cws/share?lang=nl_NL&amp;url=https%3A%2F%2Fwww.nrc.nl%2Fnieuws%2F2022%2F02%2F28%2Finkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141" title="Deel dit artikel op LinkedIn" class="share__button share__linkedin share--pop-me-up" data-type="linkedin">\n<svg width="44" height="44" viewBox="0 0 440 440"><path d="M322 330H118c-10 0-18-8-18-17V107c0-9 8-17 18-17h204c10 0 18 8 18 17v206c0 9-8 17-18 17zm-186-39h37V183h-37v108zm19-161c-13 0-21 8-21 19 0 10 8 19 20 19h1c12 0 20-9 20-19 0-11-8-19-20-19zm149 99c0-34-18-49-42-49-19 0-28 10-33 18v-15h-36v108h36v-61c0-3 0-6 1-8 3-7 9-14 19-14 13 0 18 10 18 25v58h37v-62z"/></svg>\t\t\t\t\t</a><button href="" title="Kopieer artikel-url naar Klembord" class="share__button share__clipboard " data-type="clipboard">\n<svg viewBox="-6 -6 36 36" width="24" height="24">\n\t<path d="M17 9H7V7h10m0 6H7v-2h10m-3 6H7v-2h7M12 3a1 1 0 011 1 1 1 0 01-1 1 1 1 0 01-1-1 1 1 0 011-1m7 0h-4.18A3.01 3.01 0 0012 1c-1.3 0-2.4.84-2.82 2H5a2 2 0 00-2 2v14a2 2 0 002 2h14a2 2 0 002-2V5a2 2 0 00-2-2z"/>\n</svg>\t\t\t\t\t</button>\n\n\n\t\t\t</div>\n\t\t</div>\n\n\t\t\t\t\t</div>\n\n\t\t\t\t\t<div class="metabox__target" data-target="onderwerpen">\n\t\t\t\t\t\t<div class="article__footer__carousel-container">\n\t\t\t\t\t\t\t<div class="article__carousel" data-article-slider>\n\n\t\t\t\t\t\t\t</div>\n\t\t\t\t\t\t</div>\n\t\t\t\t\t</div>\n\n\t\t\t</div>\n\t\t</section>\n\n\t\t\t<div class="banner banner--belowarticle-video "\n\t\tdata-strip-from-headless>\n\t\t\t<div class="banner__label"></div>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="belowarticle-video"\n\t\t\t\tdata-req-visitor-state="logged-in"\n\t\t></div>\n\t</div>\n\n</footer>\n\n\t</div><!--.article-container-->\n</div>\t\t\t\t\t</div>\n\t\t\t\t</article>\n\t\t\t</div>\n\n\t\t\t\t<section class="nmt-layout__sidebar sidebar" data-strip-from-headless>\n\t\t\t\t\t\t\n\t\t\t\t<div class="banner banner--rectangle banner--rectangle-next-to-article"\n\t\tdata-strip-from-headless>\n\t\t\t<div class="banner__label"></div>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="rectangle"\n\t\t\t\tdata-allow-live-refill="1"\n\t\t></div>\n\t</div>\n\n\n\t\t\t\t<div class="banner banner--belowarticlefooter-mobile "\n\t\tdata-strip-from-headless>\n\t\t\t<div class="banner__label"></div>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="belowarticlefooter-mobile"\n\t\t></div>\n\t</div>\n\n\n\t\t\t\n\t<div class="more-in-dossier"\n\t\tdata-more-in-sidebar="more-in-dossier"\n\t\tdata-more-in-sidebar-url="/data/more/dossier/coronavirus"\n\t\t\tdata-allow-sticky\n\t\t>\n\t\t<div class="more-in-dossier__inner" data-sticky-elem>\n\t\t\t<a href="/index/uitbraak-coronavirus/" class="more-in-dossier__heading">\n\t\t\t\t<h6 class="more-in-dossier__heading__flag">Dossier</h6>\n\t\t\t\t<h6 class="more-in-dossier__heading__headline">Coronavirus</h6>\n\t\t\t</a>\n\t\t\t<div class="flexfloat-wrapper">\n\t\t\t\t<div class="flexfloat flexfloat--nr-of-columns-1 flexfloat--odd-nr-of-items flexfloat--disallow-nmt-item-themes" data-more-in-sidebar-content>\n\t\t\t\t</div>\n\t\t\t</div>\n\t\t\t<a href="/index/uitbraak-coronavirus/" class="more-in-dossier__more">\n\t\t\t\tMeer Coronavirus\n\t\t\t</a>\n\t\t</div>\n\t</div>\n\n\n\t\t\t\t<section\n\t\tdata-strip-from-headless\n\t\tclass="most-read-holder in-sidebar"\n\t\tdata-most-read\n\t\tdata-most-read-fragment-uri="/data/fragments/most-read"\n\t\t\tdata-sticky="true"\n\t\t>\n\t</section>\n\n\n\n\t\t\t\t</section>\n\t\t</div>\n\n\t\n\t\t<div data-emailwall-holder data-on-demand-fragment-uri="/data/fragments/emailwall?v=3"></div>\n\t</main>\n\n\t\t<section class="article-context">\n\t\t\t<div class="nmt-layout article-context__inner">\n\t\t\t\t<div class="nmt-layout__main">\n\t\t<section class="related related--empty" data-related-widget\n\tdata-mlt="/search/api/v1/article/mlt-decay/4095141/[size]"\n\tdata-mlt-parsely="/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141/data/related"\n\t>\n\t<div>\n\t\t<h4 class="related__heading">\n\t\t\tLees ook deze artikelen\n\t\t</h4>\n\n\t\t<ul class="related__items flexfloat" data-related-items></ul>\n\t</div>\n</section><section class="related related--banner-holder">\n\t<div>\n\n\t<div class="banner banner--belowarticle "\n\t\tdata-strip-from-headless>\n\t\t\t<div class="banner__label"></div>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="belowarticle"\n\t\t></div>\n\t</div>\n\n\n\t<div class="banner banner--belowarticle-mobile "\n\t\tdata-strip-from-headless>\n\t\t\t<div class="banner__label"></div>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="belowarticle-mobile"\n\t\t></div>\n\t</div>\n\t</div>\n</section>\n\t\t\t\t\t\t\t\t\t<section\n\t\tdata-strip-from-headless\n\t\tclass="most-read-holder "\n\t\tdata-most-read\n\t\tdata-most-read-fragment-uri="/data/fragments/most-read"\n\t\t>\n\t</section>\n\n\n\t\t\t\t</div>\n\t\t\t</div>\n\t\t</section>\n\n\n\t\t<footer class="footer" data-strip-from-headless data-make-links-searchable>\n<section class="footer__container">\n\t<div class="footer__logo__container">\n\t\t<a href="/" class="footer__logo bg-nrc-logo" aria-label="NRC">\n\t\t\t<svg preserveAspectRatio="xMidYMid meet" width="120" height="36">\n\t\t\t\t<use xlink:href="#logo-nrc"></use>\n\t\t\t</svg>\n\t\t</a>\n\t</div>\n\t<div class="flexfloat">\n\t\t<div class="flexfloat__item flexfloat__item--span-1 footer__item">\n\t\t\t<ul>\n\t\t\t\t<li><a href="/over-ons/">Over NRC</a></li>\n\t\t\t\t<li><a href="/over-ons/">Over ons</a></li>\n\t\t\t\t<li><a href="/vacatures/" title="Werken bij NRC">Werken bij</a></li>\n\t\t\t\t<li><a href="/auteursrecht-databankrecht/">Auteursrecht</a></li>\n\t\t\t\t<li><a href="/privacy/" title="Privacy en NRC">Privacy</a></li>\n\t\t\t\t<li><a href="/leveringsvoorwaarden/">Leverings&shy;voorwaarden</a></li>\n\t\t\t\t<li><a href="https://nrccode.nrc.nl/">NRC-Code</a></li>\n\t\t\t\t<li><a href="/app/">Onze app</a></li>\n\t\t\t\t<li><a href="/index/archief/">Archief</a></li>\n\t\t\t</ul>\n\t\t</div>\n\t\t<div class="flexfloat__item flexfloat__item--span-1 footer__item">\n\t\t\t<ul>\n\t\t\t\t<li><a href="/mijn-nrc/">Mijn NRC</a></li>\n\t\t\t\t<li><a href="https://abonnementen.nrc.nl/?utm_intern=nrc-web-footer">Neem een abonnement</a></li>\n\t\t\t\t<li><a href="https://www.nrc.nl/login/">Inloggen</a></li>\n\t\t\t\t<li><a href="https://login.nrc.nl/create?service=https://www.nrc.nl/login/">Account aanmaken</a></li>\n\t\t\t\t<li><a href="/de/">Digitale krant</a></li>\n\t\t\t\t<li class="footer__item--mijn-leeslijst"><a href="/mijn-nrc/leeslijst/">Mijn leeslijst</a></li>\n\t\t\t\t<li><a href="https://login.nrc.nl/login?service=https://service.nrc.nl/index.php%3Fmodule%3Dabonnement">Mijn abonnementen</a></li>\n\t\t\t\t<li><a href="https://service.nrc.nl/">Service &amp; bezorging</a></li>\n\t\t\t\t<li><a href="/nieuwsbrieven/">Nieuwsbrieven</a></li>\n\n\t\t\t</ul>\n\t\t</div>\n\t\t<div class="flexfloat__item flexfloat__item--span-1 footer__item">\n\t\t\t<ul>\n\t\t\t\t<li><a href="/contact/" title="Contact met de redactie">Contact</a></li>\n\t\t\t\t<li><a href="/contact/" title="Contact met de redactie">Redactie</a></li>\n\t\t\t\t<li><a href="/opinie-contact/" title="Contact met de opinieredactie">Opinieredactie</a></li>\n\t\t\t\t<li><a href="/contact/ombudsman/" title="Contact met de ombudsman">De ombudsman</a></li>\n\t\t\t\t<li><a href="/colofon/">Colofon</a></li>\n\t\t\t\t<li><a href="https://www.nrcmedia.nl/adverteren/?utm_campaign=footer&amp;utm_source=nrc.nl">AdSales</a></li>\n\t\t\t\t<li><a href="https://service.nrc.nl/">Klantenservice</a></li>\n\t\t\t\t<li><a href="https://nrcadverteren.nl/">Familieberichten</a></li>\n\t\t\t</ul>\n\t\t</div>\n\t\t<div class="flexfloat__item flexfloat__item--span-1 footer__item">\n\t\t\t<ul>\n\t\t\t\t<li>NRC Websites</li>\n\t\t\t\t<li><a href="https://www.nrcmedia.nl/">NRC Media</a></li>\n\t\t\t\t<li><a href="https://www.nrc.nl/index/carriere/">NRC Carri&egrave;re</a></li>\n\t\t\t\t<li><a href="https://nrclive.nl/">NRC Live</a></li>\n\t\t\t\t<li><a href="https://nrcwebwinkel.nl?/utm_source=NRC_footer&amp;utm_medium=nrc.nl&amp;utm_content=tekstlink&amp;utm_campaign=footernrc">NRC Webwinkel</a></li>\n\t\t\t\t<li><a href="https://www.nrclezersfonds.nl/">NRC Lezersfonds</a></li>\n\t\t\t</ul>\n\t\t</div>\n\t</div>\n</section>\n<section class="footer__functions">\n\t<aside class="socialbuttons">\n\t\t<a href="/rss/" class="inner rss" title="NRC.nl RSS-feed">\n<svg viewBox="0 0 80 80" width="44" height="44"><path d="M39 61h-8c0-11-9-20-19-19v-8c15 0 28 13 28 27zm6 0c0-18-15-33-33-33v-8c23 0 42 19 41 41z"/><circle cx="17.5" cy="55.5" r="5.7"/></svg>\t\t</a>\n\t\t<a href="https://twitter.com/nrc" class="inner twitter" title="NRC op Twitter">\n<svg viewBox="0 0 262 262" width="44" height="44"><path d="M201 91s-5 9-14 14c2 39-27 82-78 83-26 0-42-13-43-13s20 4 41-11c-21-1-26-19-26-20s6 2 13 0c-23-6-22-28-22-28s6 4 12 4c-21-16-9-37-8-37s19 27 57 29c-6-30 29-44 47-25 9-2 18-7 18-7s-3 10-13 15c8-1 16-4 16-4z"/></svg>\t\t</a>\n\t\t<a href="https://www.facebook.com/NRC" class="inner facebook" title="NRC op Facebook">\n<svg width="44" height="44" viewBox="0 0 900 900"><path d="M495 693h-90V469h-75v-87h75v-64c0-37 11-65 32-86 20-19 46-29 80-29 19 0 41 1 67 3v78h-47c-28 0-42 13-42 38v60h86l-11 87h-75v224z"/></svg>\t\t</a>\n\t\t<a href="https://www.instagram.com/nrcnl/" class="inner instagram" title="NRC op Instagram">\n<svg width="44" height="44" viewbox="0 0 880 880">\n  <path d="M440 174c-68 0-76 0-103 1s-45 5-61 12c-16 6-30 15-44 29a123 123 0 0 0-29 44c-6 16-10 34-11 60-1 27-2 35-2 104 0 68 0 76 2 103 1 27 5 45 11 60 6 16 15 30 29 45a123 123 0 0 0 44 28c16 6 34 10 61 12 27 1 35 2 103 2s76 0 103-2c27-1 45-5 61-12a123 123 0 0 0 44-28 123 123 0 0 0 29-45c6-16 10-34 11-60 1-27 2-35 2-103s0-76-2-104c-1-27-5-45-11-60a123 123 0 0 0-29-44 123 123 0 0 0-44-29c-16-6-34-10-61-12-27-1-35-2-103-1zm0 45c67 0 75 0 101 1 24 1 38 5 46 9 12 5 20 10 29 18 9 9 14 17 19 29 3 9 8 22 8 47 1 26 1 34 2 101 0 67 0 75-1 101-1 24-5 38-9 46-5 12-10 20-19 29a77 77 0 0 1-29 18c-9 3-22 7-46 9-26 1-34 1-101 2-67 0-75 0-101-2-24-1-38-5-46-9-12-5-20-10-29-18a77 77 0 0 1-19-29c-3-9-8-22-8-46-1-26-1-34-2-101 0-67 0-75 2-101 1-24 5-38 8-47 5-12 10-20 19-29a77 77 0 0 1 29-18c9-3 22-8 46-9 26-1 34-1 101-1zm0 288a83 83 0 1 1 0-167 83 83 0 0 1 0 167zm0-212a128 128 0 1 0 0 257 128 128 0 0 0 0-257zm163-5a30 30 0 1 1-60 0 30 30 0 0 1 60 0z"/>\n</svg>\t\t</a>\n\t\t<a href="https://www.linkedin.com/company/nrc-nl/" class="inner linkedin" title="NRC op LinkedIn">\n<svg width="44" height="44" viewBox="0 0 440 440"><path d="M322 330H118c-10 0-18-8-18-17V107c0-9 8-17 18-17h204c10 0 18 8 18 17v206c0 9-8 17-18 17zm-186-39h37V183h-37v108zm19-161c-13 0-21 8-21 19 0 10 8 19 20 19h1c12 0 20-9 20-19 0-11-8-19-20-19zm149 99c0-34-18-49-42-49-19 0-28 10-33 18v-15h-36v108h36v-61c0-3 0-6 1-8 3-7 9-14 19-14 13 0 18 10 18 25v58h37v-62z"/></svg>\t\t</a>\n\t</aside>\n\t<a href="#" title="Naar bovenkant pagina">\n\t\t<aside class="top"></aside>\n\t</a>\n</section>\t\t</footer>\n\n\n\t\t<div class="banner banner--floorad "\n\t\tdata-strip-from-headless>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="floorad"\n\t\t\t\tdata-req-visitor-state="logged-out"\n\t\t></div>\n\t</div>\n\n\t\t<div class="banner banner--werving "\n\t\tdata-strip-from-headless>\n\t\t<div class="banner__output"\n\t\t\t\tdata-banner-position="werving"\n\t\t\t\tdata-banner-instant-load="1"\n\t\t></div>\n\t</div>\n\n\n\t\n\t<script type="application/ld+json" data-json-ld-for-pagemetadata>\n\t\t{"@context": "http://schema.org", "@type": "NewsArticle", "headline": "Minder inkomen \\u00e9n weinig spaargeld, het is de werkelijkheid voor een half miljoen werkenden", "description": "CPB-onderzoek: Vooral flexwerkers verloren inkomen in coronajaar 2020. Veel van hen hadden weinig spaargeld om deze tegenslag op te vangen, concludeert het CPB nu.", "dateCreated": "2022-02-28T21:45:14+02:00", "datePublished": "2022-02-28T21:45:14+02:00", "dateModified": "2022-02-28T21:45:14+02:00", "url": "https://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141", "mainEntityOfPage": "https://www.nrc.nl/nieuws/2022/02/28/inkomensval-en-weinig-spaargeld-voor-half-miljoen-werkenden-in-2020-a4095141", "author": [{"@type": "Person", "name": "Christiaan Pelgrim", "url": "https://www.nrc.nl/search/?author_filter=Christiaan%20Pelgrim"}], "publisher": {"@type": "Organization", "name": "NRC", "logo": {"@type": "ImageObject", "url": "https://www.nrc.nl/static/front/img/nrc-organization-logo.png", "width": 183, "height": 60}}, "articleSection": "economie", "keywords": ["type:nieuwsbericht", "inkomensval", "centraal planbureau", "flexwerk", "sociale ongelijkheid", "sociale zekerheid", "flexwerkers", "ontslagen", "arbeidsmarkt", "inkomensterugval", "luister naar", "terugval", "kloof", "economie", "cpb onderzoek", "coronacrisis", "cpb", "coronavirus"], "isAccessibleForFree": false, "isPartOf": {"@type": ["CreativeWork", "Product"], "name": "NRC", "productID": "nrc.nl:basic"}, "hasPart": [{"@type": "WebPageElement", "isAccessibleForFree": false, "cssSelector": ".article__content"}], "creator": ["Christiaan Pelgrim"], "image": {"@type": "ImageObject", "url": "https://images.nrc.nl/BIo-YXvqs8YWZHSGLNvZO2AHUMY=/1280x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg", "width": 5994, "height": 4000}, "thumbnailUrl": "https://images.nrc.nl/nKIOsx8Sdi1vJObsaJLFtcOVbtg=/320x/filters:no_upscale()/s3/static.nrc.nl/images/gn4/stripped/data82700949-d8bbbc.jpg", "timestamp": {"@type": "BlockchainTransaction", "identifier": "31e6df4f186f540cf1a5be9892ed08084825ba5298c02eadf25d29a0847d2b23", "hash": "799749f46fbe51e9fd110594a24fb53a891d3fac8f406ab1d6fcb3c1ebf42017", "hashLink": "https://www.nrc.nl/api/wordproof/hashinput?id=4577170", "recordedIn": {"@type": "Blockchain", "name": "eosio_main"}}}\n\t</script>\n\n\n\n\n<script>\n\twindow.nmt = window.nmt || {};\n\tnmt.splitCSSManifest = {\'debug.css\': \'css/split/debug.-1245891145.css\', \'carriere-search-box.css\': \'css/split/carriere-search-box.-1251622270.css\', \'search-suggestions.css\': \'css/split/search-suggestions.417517412.css\', \'digitale-editie-header.css\': \'css/split/digitale-editie-header.-1292816922.css\', \'carriere-carousel.css\': \'css/split/carriere-carousel.-1411666895.css\', \'emailwall.css\': \'css/split/emailwall.-748892452.css\', \'in-rubric-kwestie-in-kaart.css\': \'css/split/in-rubric-kwestie-in-kaart.1342405153.css\', \'video.css\': \'css/split/video.355411904.css\', \'inbeeld-fullpage.css\': \'css/split/inbeeld-fullpage.2122733106.css\', \'article-archived.css\': \'css/split/article-archived.-647732372.css\', \'in-rubric-maandagprofiel.css\': \'css/split/in-rubric-maandagprofiel.335185914.css\', \'crossword-article.css\': \'css/split/crossword-article.-512924605.css\', \'puzzle-article.css\': \'css/split/puzzle-article.1092360121.css\', \'hersenkrakers-puzzle-article.css\': \'css/split/hersenkrakers-puzzle-article.-1017913432.css\', \'podcast-player-article.css\': \'css/split/podcast-player-article.-975588690.css\', \'widget-regeerakkoord.css\': \'css/split/widget-regeerakkoord.-676414431.css\', \'liveblog-slowblog.css\': \'css/split/liveblog-slowblog.-2071737157.css\', \'over-ons.css\': \'css/split/over-ons.1441722853.css\', \'hersenkrakers.css\': \'css/split/hersenkrakers.-2069400170.css\', \'talk-holder.css\': \'css/split/talk-holder.1498666403.css\', \'foto-van-de-dag.css\': \'css/split/foto-van-de-dag.1147767795.css\', \'necrologieen.css\': \'css/split/necrologieen.1253507019.css\', \'onze_app.css\': \'css/split/onze_app.-1871446397.css\', \'next-afteller.css\': \'css/split/next-afteller.244263538.css\', \'quiz.css\': \'css/split/quiz.-110532525.css\', \'survey.css\': \'css/split/survey.808970437.css\', \'dataform.css\': \'css/split/dataform.1694878449.css\', \'nieuwsbrief-aanmeldpagina.css\': \'css/split/nieuwsbrief-aanmeldpagina.-370041140.css\', \'instant-feedback-widget.css\': \'css/split/instant-feedback-widget.433723702.css\', \'content-timeline.css\': \'css/split/content-timeline.-1460492270.css\', \'beslisboom.css\': \'css/split/beslisboom.-2016997515.css\', \'luister-naar.css\': \'css/split/luister-naar.214667307.css\', \'dossier-themes.css\': \'css/split/dossier-themes.-1098630785.css\', \'weblog.css\': \'css/split/weblog.1184150787.css\', \'special-kindernamen.css\': \'css/split/special-kindernamen.-612870079.css\', \'special-holleeder-overview.css\': \'css/split/special-holleeder-overview.-1720315160.css\', \'special-lintjesregen.css\': \'css/split/special-lintjesregen.280180777.css\', \'special-lostboys.css\': \'css/split/special-lostboys.-1195423080.css\', \'special-bevrijding-nederland.css\': \'css/split/special-bevrijding-nederland.1441551219.css\', \'special-l-flat.css\': \'css/split/special-l-flat.1545061049.css\', \'special-video-slideshow.css\': \'css/split/special-video-slideshow.-865343200.css\', \'special-jaar1968.css\': \'css/split/special-jaar1968.-1461097637.css\', \'special-regeerakkoord.css\': \'css/split/special-regeerakkoord.1031467206.css\', \'special-inbeeld2019.css\': \'css/split/special-inbeeld2019.-147144998.css\', \'special-inbeeld-nachtmerries.css\': \'css/split/special-inbeeld-nachtmerries.1539182926.css\', \'special-econogram.css\': \'css/split/special-econogram.-75369209.css\', \'special-parallax.css\': \'css/split/special-parallax.51751439.css\', \'special-usp-page.css\': \'css/split/special-usp-page.2043476670.css\', \'special-image-map.css\': \'css/split/special-image-map.48794537.css\', \'special-partijvergelijker.css\': \'css/split/special-partijvergelijker.1116504468.css\', \'native-ad-base.css\': \'css/split/native-ad-base.-1535187962.css\', \'native-ad-overview.css\': \'css/split/native-ad-overview.1200449730.css\', \'native-ad-article.css\': \'css/split/native-ad-article.953407079.css\', \'advertorial-header.css\': \'css/split/advertorial-header.146936986.css\', \'digitale-editie.css\': \'css/split/digitale-editie.-519574663.css\', \'colors-guide.css\': \'css/split/colors-guide.823283329.css\', \'txt.css\': \'css/split/txt.152730371.css\', \'search-page.css\': \'css/split/search-page.1972569516.css\', \'bigboard.css\': \'css/split/bigboard.-70395287.css\', \'blocky.css\': \'css/split/blocky.634479716.css\', \'podcastclub-header.css\': \'css/split/podcastclub-header.-1639718795.css\', \'podcastclub.css\': \'css/split/podcastclub.1131984251.css\', \'wilde-tuin.css\': \'css/split/wilde-tuin.1816970018.css\', \'campaign-articles.css\': \'css/split/campaign-articles.-1980105606.css\', \'item-filters.css\': \'css/split/item-filters.-1223708317.css\', \'family-members-accept-invite-page.css\': \'css/split/family-members-accept-invite-page.-2117914014.css\', \'family-members-explainer.css\': \'css/split/family-members-explainer.1969100436.css\', \'mijnnrc.css\': \'css/split/mijnnrc.1226595786.css\', \'newsletters.css\': \'css/split/newsletters.305306476.css\', \'topic-promotion-page.css\': \'css/split/topic-promotion-page.-1140168942.css\', \'compact-grid.css\': \'css/split/compact-grid.84881274.css\', \'archive-landing.css\': \'css/split/archive-landing.-1803808823.css\', \'vacatures.css\': \'css/split/vacatures.-405707691.css\', \'paywall.css\': \'css/split/paywall.-280952311.css\'};\n\n\twebpackManifest = {\'vendor\': \'js/chunks/vendor.c7ec1c7ee.js\', \'main\': \'js/chunks/main.00a1f7139.js\', \'ArchiveLanding\': \'js/chunks/ArchiveLanding.126eabf7f.js\', \'Beslisboom\': \'js/chunks/Beslisboom.b82899430.js\', \'BevrijdingNederland\': \'js/chunks/BevrijdingNederland.eb6c75300.js\', \'BigBoard\': \'js/chunks/BigBoard.a2c34ec72.js\', \'CarriereCarousel\': \'js/chunks/CarriereCarousel.f5c83f545.js\', \'ChessSolutionToggler\': \'js/chunks/ChessSolutionToggler.ccdfde606.js\', \'CrosswordInserter\': \'js/chunks/CrosswordInserter.1d3fb5902.js\', \'DataForm\': \'js/chunks/DataForm.658cb01cf.js\', \'DeArticleSidebar\': \'js/chunks/DeArticleSidebar.295615f39.js\', \'Econogram\': \'js/chunks/Econogram.2733da857.js\', \'FotoVanDeDag\': \'js/chunks/FotoVanDeDag.eed93433d.js\', \'GlossaryAnnotator\': \'js/chunks/GlossaryAnnotator.50f0a3691.js\', \'Hersenkrakers\': \'js/chunks/Hersenkrakers.20a8913bb.js\', \'ImageMap\': \'js/chunks/ImageMap.2480d62f3.js\', \'InBeeld2019\': \'js/chunks/InBeeld2019.e36e2a312.js\', \'InBeeldFullpageScroller\': \'js/chunks/InBeeldFullpageScroller.b93ea7a3b.js\', \'InBeeldFullpageScrollerMerkel\': \'js/chunks/InBeeldFullpageScrollerMerkel.0c8cb5529.js\', \'InBeeldFullpageScrollerObama\': \'js/chunks/InBeeldFullpageScrollerObama.9b3e10eeb.js\', \'InBeeldFullpageScrollerRutte\': \'js/chunks/InBeeldFullpageScrollerRutte.268b74984.js\', \'InstantFeedbackWidget\': \'js/chunks/InstantFeedbackWidget.c4a4fc463.js\', \'Liveblog\': \'js/chunks/Liveblog.ebcfd294f.js\', \'LuisterNaar\': \'js/chunks/LuisterNaar.617ed5c94.js\', \'MijnNRC\': \'js/chunks/MijnNRC.750713c54.js\', \'Newsletters\': \'js/chunks/Newsletters.cda0a78e4.js\', \'NextAfteller\': \'js/chunks/NextAfteller.9c352699c.js\', \'NotificationPromotionPage\': \'js/chunks/NotificationPromotionPage.be3c1c5c7.js\', \'OverOns\': \'js/chunks/OverOns.08b5ce73f.js\', \'PodcastPlayerArticle\': \'js/chunks/PodcastPlayerArticle.1b8e97df4.js\', \'Profiel\': \'js/chunks/Profiel.65d51c6d7.js\', \'Quiz\': \'js/chunks/Quiz.3ab4649d4.js\', \'SearchPage\': \'js/chunks/SearchPage.d0a8dbff2.js\', \'SearchSuggestions\': \'js/chunks/SearchSuggestions.bac86d34c.js\', \'SudokuInserter\': \'js/chunks/SudokuInserter.6e92c7e65.js\', \'Survey\': \'js/chunks/Survey.efc73e9a3.js\', \'TopicPromotionPage\': \'js/chunks/TopicPromotionPage.4674fe7a8.js\', \'ServiceEmails\': \'js/chunks/ServiceEmails.43a52921b.js\', \'VacaturesPage\': \'js/chunks/VacaturesPage.703a08e31.js\', \'VideoPlayer\': \'js/chunks/VideoPlayer.bd57eb2f3.js\', \'WidgetRegeerakkoord\': \'js/chunks/WidgetRegeerakkoord.bff9da582.js\', \'WildeTuin\': \'js/chunks/WildeTuin.3d4384ee8.js\', \'de\': \'js/chunks/de.6ed56c030.js\', \'firebase\': \'js/chunks/firebase.2c97a960a.js\', \'hammerjs\': \'js/chunks/hammerjs.1fa4dbdee.js\', \'itemFilters\': \'js/chunks/itemFilters.2546da46a.js\', \'InBeeldNachtmerries\': \'js/chunks/InBeeldNachtmerries.a7f9b6a95.js\', \'Jaar1968\': \'js/chunks/Jaar1968.6c22f6851.js\', \'Kindernamen\': \'js/chunks/Kindernamen.f86ac2bd5.js\', \'LFlat\': \'js/chunks/LFlat.b0036a914.js\', \'Lintjesregen\': \'js/chunks/Lintjesregen.232f1e8ad.js\', \'Lostboys\': \'js/chunks/Lostboys.af4feab48.js\', \'ParallaxBlocks\': \'js/chunks/ParallaxBlocks.ec66cb797.js\', \'Partijvergelijker\': \'js/chunks/Partijvergelijker.de5077603.js\', \'Regeerakkoord\': \'js/chunks/Regeerakkoord.1dc1e302a.js\', \'UspPage\': \'js/chunks/UspPage.9032db4e9.js\', \'VideoSlideshow\': \'js/chunks/VideoSlideshow.72c3956dd.js\'};\n\n\t!function(e){var n=window.webpackJsonp;window.webpackJsonp=function(t,c,i){for(var a,u,s,f=0,l=[];f<t.length;f++)u=t[f],r[u]&&l.push(r[u][0]),r[u]=0;for(a in c)Object.prototype.hasOwnProperty.call(c,a)&&(e[a]=c[a]);for(n&&n(t,c,i);l.length;)l.shift()();if(i)for(f=0;f<i.length;f++)s=o(o.s=i[f]);return s};var t={},r={runtime:0};function o(n){if(t[n])return t[n].exports;var r=t[n]={i:n,l:!1,exports:{}};return e[n].call(r.exports,r,r.exports,o),r.l=!0,r.exports}o.e=function(e){var n=r[e];if(0===n)return new Promise(function(e){e()});if(n)return n[2];var t=new Promise(function(t,o){n=r[e]=[t,o]});n[2]=t;var c=document.getElementsByTagName("head")[0],i=document.createElement("script");i.type="text/javascript",i.charset="utf-8",i.async=!0,i.timeout=12e4,o.nc&&i.setAttribute("nonce",o.nc),i.src=o.p+window.webpackManifest[e];var a=setTimeout(u,12e4);function u(){i.onerror=i.onload=null,clearTimeout(a);var n=r[e];0!==n&&(n&&n[1](new Error("Loading chunk "+e+" failed.")),r[e]=void 0)}return i.onerror=i.onload=u,c.appendChild(i),t},o.m=e,o.c=t,o.d=function(e,n,t){o.o(e,n)||Object.defineProperty(e,n,{configurable:!1,enumerable:!0,get:t})},o.n=function(e){var n=e&&e.__esModule?function(){return e.default}:function(){return e};return o.d(n,"a",n),n},o.o=function(e,n){return Object.prototype.hasOwnProperty.call(e,n)},o.p="https://assets.nrc.nl/static/front/",o.oe=function(e){throw console.error(e),e}}([]);\n</script>\n\n<script src="https://assets.nrc.nl/static/front/js/vendor.c7ec1c7ee.js" defer></script>\n<script src="https://assets.nrc.nl/static/front/js/main.00a1f7139.js" defer></script>\n<script src="https://content2.nrc.nl/tag/tag.js" defer></script>\n</body>\n</html>', 'in_cache_date': datetime.datetime(2022, 3, 7, 8, 39, 47, 91886), 'parsed': 0}
```
