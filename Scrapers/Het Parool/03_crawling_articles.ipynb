```python
from bs4 import BeautifulSoup
import requests as req
from datetime import datetime
```

```python
latest_article_url = 'http://parool.nl/amsterdam/advocaat-gijzelnemer-apple-store-overleden~bc7cc889/'
```

```python
def fetchArticle(url):
    document = {}
    
    try:
        r = req.get(url)
        r.raise_for_status()
    
    except req.exceptions.HTTPError as err:
        # TODO: add error handling
        pass
    
    else:        
        document['url'] = url
        document['raw_html'] = r.content
        document['in_cache_date'] = datetime.utcnow()
        document['parsed'] = 0
        
    return document
```

```python
article = fetchArticle(latest_article_url)
print(article)
```
