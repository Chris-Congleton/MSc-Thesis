```python
r = {'url': 'https://www.volkskrant.nl/nieuws-achtergrond/jacht-op-kosovaren-stelt-navo-voor-dilemma~b34416f7/', 'raw_html': b'<!DOCTYPE html>\n<html lang="nl-NL" class="is-article-page  "> <head> <meta charset="utf-8"> <meta http-equiv="x-ua-compatible" content="ie=edge"> <meta name="viewport" content="width=device-width,initial-scale=1,viewport-fit=cover">     <title>Jacht op Kosovaren stelt NAVO voor dilemma | De Volkskrant</title>     <meta property="article:section" content="Nieuws &amp; Achtergrond">  <meta property="article:published_time" content="1999-03-29T00:00+02:00">  <meta property="article:author" content="Van onze buitenlandredactie Anet Bleich">   <meta property="article:tag" content="politiek">  <meta property="article:tag" content="internationale betrekkingen">  <meta property="article:tag" content="conflicten, oorlog en vrede">  <meta property="article:tag" content="internationale organisaties">  <meta property="article:tag" content="Amerika">  <meta property="article:tag" content="Belgrado">  <meta property="article:tag" content="Bosni\xc3\xab">  <meta property="article:tag" content="Europa">     <meta name="description" content="Servi\xc3\xab graaft zich in. In Kosovo woedt een moorddadige terreur tegen de Albanese burgers. Wat kan de NAVO doen?.."> <meta name="twitter:card" content="summary_large_image"> <meta name="twitter:title" content="Jacht op Kosovaren stelt NAVO voor dilemma"> <meta name="twitter:description" content="Servi\xc3\xab graaft zich in. In Kosovo woedt een moorddadige terreur tegen de Albanese burgers. Wat kan de NAVO doen?.."> <meta name="twitter:image"> <meta property="og:type" content="article"> <meta property="og:site_name" content="de Volkskrant"> <meta property="og:locale" content="nl_NL"> <meta property="og:title" content="Jacht op Kosovaren stelt NAVO voor dilemma"> <meta property="og:description" content="Servi\xc3\xab graaft zich in. In Kosovo woedt een moorddadige terreur tegen de Albanese burgers. Wat kan de NAVO doen?.."> <meta property="og:url" content="https://www.volkskrant.nl/gs-b34416f7"> <meta property="og:image">  <link rel="preconnect" href="https://myprivacy.dpgmedia.net"> <link rel="preconnect" href="https://advertising-cdn.dpgmedia.cloud"> <link rel="preconnect" href="https://www.googletagmanager.com"> <link rel="dns-prefetch" href="https://myprivacy.dpgmedia.net"> <link rel="dns-prefetch" href="https://advertising-cdn.dpgmedia.cloud"> <link rel="dns-prefetch" href="https://www.googletagmanager.com"> <script> \n    if (typeof dtrum === \'object\') {\n        dtrum.enable();\n    }\n\n    window.APP = {\n        config: {\n            connection: {},\n            dfp: JSON.parse(null),\n            sessionCookieName: "VK_SESSION",\n        },\n        helpers: {\n            waitUntilTrue: function waitUntilTrue(fn, successCb, errorCb, delayBetweenIterations, maxIterations, _currIteration) {\n                if (!_currIteration) {\n                    setTimeout(function () {\n                        window.APP.helpers.waitUntilTrue(fn, successCb, errorCb, delayBetweenIterations, maxIterations, 1)\n                    }, 0);\n                    return;\n                }\n\n                if (fn()) {\n                    if (typeof successCb === \'function\') {\n                        successCb();\n                    }\n                } else if (_currIteration < maxIterations) {\n                    setTimeout(function () {\n                        window.APP.helpers.waitUntilTrue(fn, successCb, errorCb, delayBetweenIterations, maxIterations, _currIteration + 1)\n                    }, delayBetweenIterations);\n                } else {\n                    if (typeof errorCb === \'function\') {\n                        errorCb();\n                    }\n                }\n            },\n            debounce: function(func, wait) {\n                let timeout;\n\n                return function executedFunction() {\n                    const context = this,\n                        args = arguments;\n\n                    clearTimeout(timeout);\n                    timeout = setTimeout(function() {\n                        timeout = null;\n                        func.apply(context, args);\n                    }, wait);\n                };\n            },\n            runWhenInBreakpoint: function(breakpoints, fn) {\n                function _getActiveBreakpoint() {\n                    if (1320 <= window.innerWidth) {\n                        return \'xlarge\';\n                    }\n\n                    if (960 <= window.innerWidth) {\n                        return \'large\';\n                    }\n\n                    if (640 <= window.innerWidth) {\n                        return \'medium\';\n                    }\n\n                    return \'small\';\n                }\n\n                const onResize = window.APP.helpers.debounce(function() {\n                    if (breakpoints[_getActiveBreakpoint()]) {\n                        fn();\n                    }\n                    window.removeEventListener(\'resize\', onResize);\n                }, 100);\n\n                if (breakpoints[_getActiveBreakpoint()]) {\n                    fn();\n                } else {\n                    window.addEventListener(\'resize\', onResize);\n                }\n            }\n        },\n        modules: {\n            advertising: {\n                initialized: false,\n                adSlots: null,\n                paywallContentId: "b34416f7",\n                isAdvertorialArticle: false,\n            },\n            temptation: {\n                executed: false,\n                showPaywall: false\n            }\n        },\n        supported: {\n            idleCallback: "requestIdleCallback" in window\n        }\n    };\n\n    window.dataLayer = [JSON.parse("{\\"screen_info\\":{\\"content_id\\":\\"34416f7\\",\\"content_access\\":\\"free\\",\\"content_category\\":\\"Nieuws \\u0026 Achtergrond\\",\\"content_type\\":\\"normal\\",\\"content_pubdate\\":922658400000,\\"content_length\\":\\"0-50\\",\\"content_tags\\":\\"|politiek|internationale betrekkingen|conflicten, oorlog en vrede|internationale organisaties|Amerika|Belgrado|Bosni\\u00EB|Europa|\\",\\"content_author\\":\\"Van onze buitenlandredactie Anet Bleich\\",\\"page_type\\":\\"article\\",\\"page_category\\":\\"Nieuws \\u0026 Achtergrond\\",\\"page_navpath\\":\\"Nieuws \\u0026 Achtergrond\\",\\"page_brand_name\\":\\"de Volkskrant\\",\\"page_platform_type\\":\\"website\\",\\"page_template\\":\\"default\\",\\"page_identifier\\":\\"live_feed\\"},\\"content\\":{\\"wiki\\":{\\"label\\":\\"|Amerika|Belgrado|Bosni\\u00EB|Europa|Joegoslavi\\u00EB|Kosovo|Moskou|NAVO|Rambouillet|Rusland|Servi\\u00EB|Srebrenica|\\"},\\"iptc\\":{\\"id\\":\\"|11000000|16000000|20000638|20000644|\\",\\"label\\":\\"|politiek|conflicten, oorlog en vrede|internationale betrekkingen|internationale organisaties|\\"},\\"ner\\":{\\"loc\\":\\"|Amerika|Belgrado|Bosni\\u00EB|Europa|Joegoslavi\\u00EB|Kosovo|Moskou|Rambouillet|Rusland|Servi\\u00EB|Srebrenica|\\",\\"org\\":\\"|NAVO|Nemtsov|\\",\\"per\\":\\"|Fjodorov|Gajdar|\\"}}}")];\n\n    // ask loginHelpers to prepare userInfo\n    window.APP.helpers.waitUntilTrue(\n        function() {\n            // temptation call refreshes the access-token when needed; we need to wait for temptation to be executed\n            return window.APP.loginHelpers && window.APP.modules.temptation.executed;\n        },\n        function() {\n            window.APP.loginHelpers.prepareUserInfo();\n        },\n        function() {\n        },\n        100, 100);\n\n    window.APP.helpers.waitUntilTrue(\n        function() {\n            return window.APP.loginHelpers && window.APP.loginHelpers.isUserInfoPrepared();\n        },\n        function() {\n            const userId = window.APP.loginHelpers.getUserId();\n\n            if (userId !== null) {\n                window.dataLayer[0][\'gigya-id\'] = userId;\n                window.dataLayer.push({\n                    "event": "gigya-id",\n                    "gigya-id": userId\n                });\n            }\n        },\n        function() {\n        },\n        100, 100);\n\n\n    window.APP.helpers.waitUntilTrue(\n        function() {\n            return window.APP.loginHelpers;\n        },\n        function() {\n            window.APP.loginHelpers.removePWSCookie(\'"humo"\' === \'"volkskrant"\');\n        },\n        function() {\n            console.error("could not remove corrupt PWS cookie");\n        },\n        100, 100);\n\n    if (Boolean(navigator.platform) && /MacIntel|Macintosh|iPad|iPhone|iPod/.test(navigator.platform)) {\n        let html = document.querySelector(\'html\');\n\n        if (html.classList) {\n            html.classList.add(\'js-is-ios\');\n        }\n\n        window.APP.config.platform = navigator.platform;\n    }\n\n    if (Boolean(navigator.platform) && /Android|Linux armv8l/.test(navigator.platform)) {\n        let html = document.querySelector(\'html\');\n\n        if (html.classList) {\n            html.classList.add(\'js-is-android\');\n        }\n\n        window.APP.config.platform = navigator.platform;\n    }\n\n    if (Boolean(navigator.connection) && Boolean(navigator.connection.effectiveType)) {\n        let html = document.querySelector(\'html\');\n\n        if (html.classList) {\n            html.classList.add(\'js-is-\' + navigator.connection.effectiveType);\n        }\n\n        window.APP.config.connection.effectiveType = navigator.connection.effectiveType;\n    } </script>  <script>window._privacy=window._privacy||[]</script>  <link rel="stylesheet" href="https://www.volkskrant.nl/sel-static/shared-assets/css/volkskrant-typography-release-13.css">  <link rel="stylesheet" href="/static/main-f05703ffcf.css">     <link rel="stylesheet" href="https://www.volkskrant.nl/sel-static/teasers/volkskrant-main-8543d7afc6.css" media="print" onload=\'this.media="all"\'> <noscript><link rel="stylesheet" href="https://www.volkskrant.nl/sel-static/teasers/volkskrant-main-8543d7afc6.css"></noscript>      <script async src="https://advertising-cdn.dpgmedia.cloud/web-advertising/13/0/4/advert-xandr.js" type="application/javascript"></script><script async data-advert-src="https://acdn.adnxs.com/ast/ast.js" data-advert-vendors="32"></script><script async data-advert-src="https://cdn.brandmetrics.com/survey/script/ff6ac435ab3d450881ba95c0e58401db.js" data-advert-vendors="422"></script><script async src="https://advertising-cdn.dpgmedia.cloud/web-advertising/prebid.js"></script><script type="application/javascript">(function(){if("undefined"==typeof performance.mark){performance.mark=function(){}}performance.mark("advert - inline-script - start");function a(a,c,settings){var d=a[c];if("undefined"==typeof d){console.error("[ADVERT] Unknown slot \'".concat(c,"\'"));return null}if("string"==typeof d||null===d){return d}var e=b(settings);return"undefined"==typeof d[e]?d.default:d[e]}function b(settings){var a=window.innerWidth,b=settings.xandr.breakpoints;return Object.keys(b).reduce(function(c,d){if(b[d]<=a){if(null===c||b[d]>b[c]){c=d}}return c},null)} function e(a){if("function"==typeof Storage){return window.localStorage.getItem(a)||null}else if(navigator.cookieEnabled){return g(a)}return null}function f(a){var b=e(a);return b?decodeURIComponent(b).split(","):[]}function g(a){var b=document.cookie.match("(?:^|;\\\\s*)".concat(a,"=([^;]*)"));return b&&decodeURI(b[1])||null}function isFirefoxBrowser(){return"undefined"!=typeof InstallTrigger}function isSafariBrowser(){return(!!window.ApplePaySetupFeature||!!window.safari)&&_agentHas("Safari")&&!_agentHas("Chrome")&&!_agentHas("CriOS")}function isEdgeBrowser(){return _agentHas("edg")}function _agentHas(a){return-1<navigator.userAgent.toLowerCase().search(a.toLowerCase())}window.advert=window.advert||{};window.advert.cmd=window.advert.cmd||[];  function h(a,b){if("function"==typeof window.__tcfapi){a();return}setTimeout(function(){return h(a,++b)},100*b)}var i;function j(a){performance.mark("advert - inline-script - privacy - start");h(function(){performance.mark("advert - inline-script - privacy - api available");if(i){window.__tcfapi("removeEventListener",2,i)}i=function(b,c){if(c&&-1!==["tcloaded","useractioncomplete"].indexOf(b.eventStatus)){performance.mark("advert - inline-script - privacy - tcdata");a(b)}};window.__tcfapi("addEventListener",2,i)},1)}function k(b,settings,d){ j(function(a){if(0===[1,2,7,10].filter(function(b){return!a.purpose.consents[b]}).length){n(a);m(settings,a);l(settings)}});window.getAdvertSlotByName=function(a){return window.advert.getSlot(a)};window.getAdvertSlotById=function(a){return window.advert.getSlot(b[a])};window.loadAdvertSlot=function(e,f){window.advert.cmd.push(["loaded",function(){var c=a(b,e,settings);if(c){window.advert.loadSlot(c,f||e)}}])};}function l(settings){window.advert.cmd.push(function(){     performance.mark("advert - inline-script - paywall - start");function a(){return new Promise(function(a){var b=setTimeout(function(){console.warn("[ADVERT] Timed out waiting for paywall");a(!1)},5e3),c=setInterval(function(){if(window.APP&&window.APP.modules&&window.APP.modules.advertising&&window.APP.modules.temptation&&window.APP.modules.temptation.executed){clearInterval(c);clearTimeout(b);a(window.APP.modules.advertising.paywallContentId&&!window.APP.modules.advertising.isAdvertorialArticle&&window.APP.modules.temptation.showPaywall)}},10);if(window.APP&&window.APP.modules&&window.APP.modules.advertising&&window.APP.modules.temptation&&window.APP.modules.temptation.executed){clearInterval(c);clearTimeout(b);a(window.APP.modules.advertising.paywallContentId&&!window.APP.modules.advertising.isAdvertorialArticle&&window.APP.modules.temptation.showPaywall)}})}try{performance.mark("advert - inline-script - paywall - start");a().then(function(a){performance.mark("advert - inline-script - paywall - end");if(!a){window.advert.init(settings,window.advertHooks||{});}})}catch(a){console.warn("[ADVERT] Failed to load paywall. Continuing\\u2026");window.advert.init(settings,window.advertHooks||{})}})} function applyGigyaId(a){var b=window.dataLayer&&window.dataLayer.find(function(a){return null!==a["gigya-id"]});if(b){a.gigyaid=b["gigya-id"]}else{a.gigyaid=!1}}  function m(settings,a){var b=a.purpose.consents,c=a.vendor.consents,d=b[3]&&b[4],h=settings.xandr.targeting,i=!isSafariBrowser()&&!isFirefoxBrowser()&&!isEdgeBrowser();h.targeted=d;h.thirdpartycookie=i?1:0;if(d){h.spid=e("_sp_id")}else{delete h.spid} if(d&&c[412]){h.cxid=e("_cX_G");h.cxsg=f("cxSegments");h.cxsg_m=f("cxSegments_m")}else{delete h.cxid;delete h.cxsg;delete h.cxsg_m} if(window.openDataLayer){h.pageviewid=window.openDataLayer.pageViewId}  if(d){applyGigyaId(h)}if(d){settings.xandr.user=g("authId")}else{delete settings.xandr.user}}function n(a){Array.from(document.querySelectorAll("[data-advert-src]")).forEach(function(b){var c=b.getAttribute("data-advert-vendors");if(c&&c.split(",").some(function(b){return!a.vendor.consents[b]})){return}b.setAttribute("src",b.getAttribute("data-advert-src"));b.removeAttribute("data-advert-src");b.removeAttribute("data-advert-vendors")})}window.loadAdvertSettings=function(a,settings,b){if(!a||!settings){console.error("[ADVERT] Trying to load without slot mappings or settings",a,settings);return}k(a,settings,b)};if(document.currentScript){document.currentScript.remove()}performance.mark("advert - inline-script - end")})();</script><style>[data-advert-slot-state]{display:-webkit-box;display:-ms-flexbox;display:flex;-webkit-box-align:center;-ms-flex-align:center;align-items:center;-webkit-box-pack:center;-ms-flex-pack:center;justify-content:center}\n[data-advert-slot-empty],[data-advert-slot-state=\'created\']{display:none}\narticle [data-advert-slot-state="rendered"]:not([data-advert-slot-empty]){display:-webkit-box;display:-ms-flexbox;display:flex;-webkit-box-align:center;-ms-flex-align:center;align-items:center;-webkit-box-pack:center;-ms-flex-pack:center;justify-content:center;padding:20px 0;border:1px solid rgba(0,0,0,0.03);background-color:rgba(0,0,0,0.03)}</style><script type="application/javascript">(function(a,settings,b){performance.mark("advert - load-settings - start");if(!window.loadAdvertSettings){console.error("[ADVERT] Tried to load advert settings before main initialisation.");return}window.loadAdvertSettings(a,settings,b);if(document.currentScript){document.currentScript.remove()}performance.mark("advert - load-settings - end")})( {"branded2":"branded--2","dfp-sky1":"sky--1","sky1":"sky--1","mid2":"mid--2","article_sky_1":{"large":"sky--1","default":null,"xlarge":"sky--1"},"top1":"top--1","layer1":null,"mid1":"mid--1"} , {"xandr":{"memberId":7463,"disablePSA":true,"enableSafeFrame":false,"breakpoints":{"small":0,"medium":640,"large":960,"xlarge":1320},"targeting":{"pag":"detail","gamAdUnit":"vk/vk.site-%screenSize%","cat":["nieuws-achtergrond"],"cid":"34416f7","articleType":"normal","pay":false,"inread":true,"top":["europa","internationale-organisaties","politiek","amerika","bosnie","internationale-betrekkingen","conflicten-oorlog-en-vrede","belgrado"],"maxTemperature":11,"minTemperature":6,"temperature":9,"weatherIcon":"B","weatherType":"bewolkt","bsuit":"NScreened"},"slots":[{"name":"top--1","placement":{"small":"volkskrant_wsc_other_detail_h1","medium":"volkskrant_wmc_other_detail_h1","large":"volkskrant_wmc_other_detail_h1","xlarge":"volkskrant_wmc_other_detail_h1"},"sizeMapping":"top1","lazyLoad":false,"targeting":{"pos":"top1"}},{"name":"mid--1","placement":{"small":"volkskrant_wsc_other_detail_r4","medium":"volkskrant_wmc_other_detail_r2","large":"volkskrant_wmc_other_detail_r2","xlarge":"volkskrant_wmc_other_detail_r2"},"sizeMapping":"mid1","lazyLoad":true,"lazyLoadThreshold":{"small":300,"default":null},"targeting":{"pos":"mid1"}},{"name":"sky--1","placement":{"small":"volkskrant_wsc_other_detail_r1","medium":"volkskrant_wmc_other_detail_r1","large":"volkskrant_wmc_other_detail_r1","xlarge":"volkskrant_wmc_other_detail_r1"},"sizeMapping":"sky1","lazyLoad":true,"lazyLoadThreshold":{"default":null,"medium":9999,"large":9999,"xlarge":9999},"targeting":{"pos":"sky1"}},{"name":"mid--2","placement":{"small":"volkskrant_wsc_other_detail_r5","medium":"volkskrant_wmc_other_detail_r3","large":"volkskrant_wmc_other_detail_r3","xlarge":"volkskrant_wmc_other_detail_r3"},"sizeMapping":"mid2","lazyLoad":true,"lazyLoadThreshold":{"small":600,"default":null},"targeting":{"pos":"mid2"}},{"name":"branded--2","placement":{"small":"volkskrant_wsc_other_detail_b2","medium":"volkskrant_wmc_other_detail_b2","large":"volkskrant_wmc_other_detail_b2","xlarge":"volkskrant_wmc_other_detail_b2"},"sizeMapping":"branded2","lazyLoad":true,"targeting":{"pos":"branded2"}}],"sizeMapping":{"top1":{"small":[],"medium":[[728,90],[768,100],[1800,200]],"large":[[728,90],[970,250],[970,251],[1800,200]],"xlarge":[[970,250],[728,90],[970,90],[970,251],[1800,200]]},"mid1":{"small":[[320,250],[320,240],[300,250],[320,50],[320,100],[300,600],[320,241],[320,101]],"medium":[[300,250],[300,252],[728,90],[336,280]],"large":[[300,250],[300,252],[970,250],[728,90],[336,280]],"xlarge":[[300,250],[300,252],[970,250],[728,90],[336,280]]},"sky1":{"small":[[320,250],[320,240],[300,250],[320,50],[320,100],[280,280],[320,700],[360,675],[320,242],[360,700]],"medium":[[300,251],[300,600],[300,250],[160,600],[120,600],[300,601],[300,602]],"large":[[300,251],[300,600],[300,250],[160,600],[120,600],[300,601],[300,602]],"xlarge":[[300,251],[300,600],[300,250],[160,600],[120,600],[300,601],[300,602]]},"mid2":{"small":[[320,250],[320,240],[300,250],[320,50],[320,100],[300,600],[320,243]],"medium":[[300,250],[336,280],[352,280],[300,253],[336,283],[600,280]],"large":[[300,250],[336,280],[352,280],[300,253],[336,283],[600,280]],"xlarge":[[300,250],[336,280],[352,280],[336,283],[300,253],[600,280]]},"branded2":{"small":[],"medium":[[300,285],[300,280]],"large":[[300,285],[300,280]],"xlarge":[[300,285],[300,280]]}},"loadSlotBatchCount":{"small":8,"medium":6,"large":6,"xlarge":6},"lazyLoadThreshold":600},"prebid":{"timeout":2000,"gdpr":{},"additionalConfig":{"criteo":{"fastBidVersion":"latest"},"floors":{"data":{"schema":{"fields":["bidder","mediaType","size"]},"skipRate":0,"currency":"EUR","values":{"*|banner|1800x200":12,"*|banner|320x100":0.75,"gps|banner|*":0.4,"*|banner|320x400":6,"*|banner|320x50":0.75,"*|banner|320x240":2,"*|banner|300x250":2,"*|banner|728x90":2,"*|banner|320x480":8,"criteo|banner|970x250":6,"*|banner|120x600":2,"*|banner|300x600":7,"*|banner|970x250":7,"*|banner|160x600":2,"*|banner|16x9":8,"*|banner|1800x1000":12,"*|banner|400x225":8,"*|banner|970x1000":12,"*|banner|336x280":2,"*|banner|320x500":6,"*|banner|970x90":2,"*|video|*":8}}},"improvedigital":{"usePrebidSizes":true},"enableSendAllBids":true,"priceGranularity":{"buckets":[{"increment":0.01,"min":0,"max":20,"precision":2},{"increment":0.1,"min":20,"max":30,"precision":2},{"increment":1,"min":30,"max":31,"precision":2}]},"currency":{"adServerCurrency":"EUR","granularityMultiplier":1,"defaultRates":{"USD":{"EUR":0.885}}},"rubicon":{"singleRequest":true},"userSync":{"filterSettings":{"iframe":{"bidders":"*","filter":"include"}},"userIds":[{"name":"criteo"},{"name":"pubCommonId","storage":{"type":"cookie","name":"_pubcid","expires":365}},{"name":"pubProvidedId","params":{"eids":[{"uids":[{"ext":{"stype":"ppuid"},"atype":1,"id":"%%PWID%%"}],"source":"%%domain%%"}]}},{"name":"id5Id","params":{"partner":677},"storage":{"name":"id5id","expires":30,"type":"html5","refreshInSeconds":28800}}]}},"slots":{"sky--1":{"sizes":{"small":[[320,240],[300,250],[320,50],[320,100]],"medium":[[300,250],[300,600],[160,600],[120,600]],"large":[[300,250],[300,600],[160,600],[120,600]],"xlarge":[[300,600],[300,250],[160,600],[120,600]]},"bidders":{"small":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195214","zoneId":"954316"},"bidder":"rubicon"},{"params":{"siteId":"270649"},"bidder":"ix"},{"params":{"placementId":"22125158"},"bidder":"improvedigital"},{"params":{"placementId":"22271050"},"bidder":"imprweborama"},{"params":{"placementId":"18652809"},"bidder":"gps"}],"default":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"1205703"},"bidder":"imprweborama"},{"params":{"placementId":"18652691"},"bidder":"gps"}]}},"mid--2":{"sizes":{"small":[[320,240],[300,250],[320,50],[320,100],[300,600]],"medium":[[300,250],[336,280]],"large":[[300,250],[336,280]],"xlarge":[[300,250],[336,280]]},"bidders":{"small":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195214","zoneId":"954316"},"bidder":"rubicon"},{"params":{"siteId":"270649"},"bidder":"ix"},{"params":{"placementId":"22125158"},"bidder":"improvedigital"},{"params":{"placementId":"18652812"},"bidder":"gps"}],"default":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"18652695"},"bidder":"gps"}]}},"top--1":{"sizes":{"large":[[728,90],[970,250]],"xlarge":[[728,90],[970,90],[970,250],[1800,1000],[970,1000]],"medium":[[728,90]]},"bidders":{"small":[],"default":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"1205703"},"bidder":"imprweborama"},{"params":{"placementId":"18652690"},"bidder":"gps"}],"medium":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"18652690"},"bidder":"gps"}],"xlarge":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"1205703"},"bidder":"imprweborama"},{"params":{"placementId":"18652690"},"bidder":"gps"}]}},"mid--1":{"sizes":{"small":[[320,240],[300,250],[320,50],[320,100],[300,600]],"medium":[[300,250],[728,90],[336,280]],"large":[[300,250],[970,250],[728,90],[336,280]],"xlarge":[[300,250],[970,250],[728,90],[336,280]]},"bidders":{"small":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195214","zoneId":"954316"},"bidder":"rubicon"},{"params":{"siteId":"270649"},"bidder":"ix"},{"params":{"placementId":"22125158"},"bidder":"improvedigital"},{"params":{"placementId":"18652811"},"bidder":"gps"}],"default":[{"params":{"publisherId":"156353","currency":"EUR","adSlot":"volkskrant"},"bidder":"pubmatic"},{"params":{"networkId":"6816"},"bidder":"criteo"},{"params":{"siteId":"195192","zoneId":"954292"},"bidder":"rubicon"},{"params":{"siteId":"371290"},"bidder":"ix"},{"params":{"placementId":"22125157"},"bidder":"improvedigital"},{"params":{"placementId":"18652693"},"bidder":"gps"}]}}},"bidders":{"pubmatic":{"targetingFields":["pag","pos","positie","top","ksg","cxsg","cxid","cat"]},"criteo":{"targetingFields":[]},"rubicon":{"accountId":"12108","targetingFields":["pag","pos","positie","top","ksg","cxsg","cxid","cat"]},"improvedigital":{"targetingFields":["pag","pos","positie","top","cat"]},"imprweborama":{"targetingFields":["pag","pos","positie","top","cat"]},"gps":{"targetingFields":["iabCategories","bsuit","top","iabTopics","pag"]}}}} ,{});</script>  <script>window.dataLayer=window.dataLayer||[],window.dataLayer.push({"gtm.start":(new Date).getTime(),event:"gtm.js"})</script> <script src="https://t.volkskrant.nl/gtm.js?id=GTM-TPHZL7D" async="async"></script>    <script type="text/javascript"> window.cmpProperties = {\n            privacyManagerId: 148844,\n            baseUrl: \'https://www.volkskrant.nl\',\n            cmpCname: \'https://cmp.volkskrant.nl\',\n            featureFlags: {\n                unified: true\n            }\n        } </script>    <script async type="text/javascript" src="https://myprivacy.dpgmedia.net/consent.js"></script>   <script>function showPrivacyModal(){window._privacy.openModal()}</script>    <script> window.APP.helpers.waitUntilTrue(\n            function () {\n                return Boolean(window.temptation && window.temptation.init && window.APP.loginHelpers);\n            },\n            function () {\n                window.temptation.init("b34416f7",\n                    false,\n                    {"articleTitle":"Jacht op Kosovaren stelt NAVO voor dilemma","articleType":"normal","author":"Van onze buitenlandredactie Anet Bleich"});\n            },\n            function () {\n                window.APP.modules.temptation.executed = true;\n                console.error(\'Could not init paywall\');\n            },\n            100,\n            100\n        ); </script>  <script type="text/javascript" src="https://polyfill.io/v2/polyfill.min.js?features=Array.from,IntersectionObserver" nomodule async="async"></script> <script src="/static/main-356027f9bc.js" defer="defer"></script>  <script src="https://www.volkskrant.nl/sel-static/teasers/main-5c832f4c87.js" defer="defer"></script>  <link rel="canonical" href="https://www.volkskrant.nl/nieuws-achtergrond/jacht-op-kosovaren-stelt-navo-voor-dilemma~b34416f7/"> <link rel="apple-touch-icon-precomposed" sizes="192x192" href="/static/apple-touch-icon-192x192.png"> <link rel="apple-touch-icon-precomposed" sizes="180x180" href="/static/apple-touch-icon-180x180.png"> <link rel="apple-touch-icon-precomposed" sizes="167x167" href="/static/apple-touch-icon-167x167.png"> <link rel="apple-touch-icon-precomposed" sizes="152x152" href="/static/apple-touch-icon-152x152.png"> <link rel="apple-touch-icon-precomposed" sizes="144x144" href="/static/apple-touch-icon-144x144.png"> <link rel="apple-touch-icon-precomposed" sizes="128x128" href="/static/apple-touch-icon-128x128.png"> <link rel="apple-touch-icon-precomposed" sizes="120x120" href="/static/apple-touch-icon-120x120.png"> <link rel="apple-touch-icon-precomposed" sizes="114x114" href="/static/apple-touch-icon-114x114.png"> <link rel="apple-touch-icon-precomposed" sizes="76x76" href="/static/apple-touch-icon-76x76.png"> <link rel="apple-touch-icon-precomposed" sizes="72x72" href="/static/apple-touch-icon-72x72.png"> <link rel="apple-touch-icon-precomposed" sizes="60x60" href="/static/apple-touch-icon-60x60.png"> <link rel="icon" type="image/png" href="/static/favicon-196x196.png" sizes="196x196"> <link rel="icon" type="image/png" href="/static/favicon-128x128.png" sizes="128x128"> <link rel="icon" type="image/png" href="/static/favicon-96x96.png" sizes="96x96"> <link rel="icon" type="image/png" href="/static/favicon-32x32.png" sizes="32x32"> <link rel="icon" type="image/png" href="/static/favicon-16x16.png" sizes="16x16"> \n  <link rel="stylesheet" href="/article-static/article-vk-b3139c7b4e.css"> \n \n \n  <script type="application/ld+json">{"headline":"Jacht op Kosovaren stelt NAVO voor dilemma","url":"https://www.volkskrant.nl/nieuws-achtergrond/jacht-op-kosovaren-stelt-navo-voor-dilemma~b34416f7/","articleSection":"Nieuws & Achtergrond","dateCreated":"1999-03-29T00:00:00+0200","datePublished":"1999-03-29T00:00:00+0200","copyrightYear":"1999","articleBody":"","description":"Servi\xc3\xab graaft zich in. In Kosovo woedt een moorddadige terreur tegen de Albanese burgers. Wat kan de NAVO doen?..","mainEntityOfPage":"https://www.volkskrant.nl/nieuws-achtergrond/jacht-op-kosovaren-stelt-navo-voor-dilemma~b34416f7/","speakable":{"xpath":["/html/head/title","/html/head/meta[@name=\'description\']/@content"],"@type":"SpeakableSpecification"},"potentialAction":{"name":"Abonneren","object":{"name":"Volkskrant","@type":"Newspaper"},"target":{"urlTemplate":"https://www.volkskrant.nl/abonnementen/abonnementen/vk","@type":"EntryPoint"},"@type":"SubscribeAction"},"@type":"NewsArticle","@context":"https://schema.org","@id":"https://www.volkskrant.nl/nieuws-achtergrond/jacht-op-kosovaren-stelt-navo-voor-dilemma~b34416f7/#Article"}</script> \n  </head> <body>   <div class="fjs-subscribe-bar" data-temptation-position="PAGE_TOP"></div> <div class="ad--top fjs-hide-on-paywall" id="top1" data-id="top1"> <script>window.APP.helpers.runWhenInBreakpoint({xlarge:!0,large:!0,medium:!0,small:!1},function(){"function"==typeof window.loadAdvertSlot&&window.loadAdvertSlot("top1")})</script> </div>  <header class="app-header-article fjs-app-header-article"> <div class="app-header-article-container"> <div class="app-header-article__back"> <a class="app-header-article__back-wrapper fjs-back-button" href="/#nieuws-achtergrond" data-gtm="header / back / nieuws &amp; achtergrond"> <svg class="app-header-article__back-icon" width="16" height="22"> <path d="M9.5 18a.5.5 0 0 1-.35-.15l-8.5-8.5a.5.5 0 0 1 0-.7l8.5-8.5a.5.5 0 0 1 .7.7L1.71 9l8.14 8.15a.5.5 0 0 1-.35.85z"/> </svg> </a> </div> <div class="app-header-article__brand"> <a class="app-brand" href="/"> <svg width="234" height="30" viewBox="0 0 234 30" fill="none" xmlns="http://www.w3.org/2000/svg" class="app-brand-logo"> <path fill-rule="evenodd" clip-rule="evenodd" d="M70.8012 15.4634C76.593 15.4634 80.3693 18.2141 80.3693 22.413C80.3693 26.7563 76.6783 29.4785 70.8012 29.4785C64.9522 29.4785 61.2327 26.7563 61.2327 22.471C61.2327 18.1853 64.9522 15.4634 70.8012 15.4634ZM229.23 10.7432L229.599 11.0328V16.1002H233.944V17.2589H229.599V23.9477C229.599 24.7877 229.542 25.7429 229.741 26.3512C229.912 26.9013 230.337 27.2488 231.048 27.2488C232.925 27.2488 233.032 25.6393 233.035 24.0377L233.035 23.6581H234V24.0924C234 25.6853 233.915 26.4091 233.574 27.075C232.779 28.6677 230.991 29.4785 228.236 29.4785C226.163 29.4785 224.744 29.044 223.864 28.0309C223.296 27.3938 222.984 26.4958 222.899 25.4536C222.843 24.8893 222.815 24.4346 222.814 23.1531L222.814 17.2589H220.485V16.3028L229.23 10.7432ZM31.3735 15.521C36.8817 15.521 39.9766 18.0982 40.0616 22.7318H28.9037V23.3977C28.9037 24.7004 28.9605 25.5983 29.1308 26.1487C29.4998 27.4517 30.806 28.2331 32.7647 28.2331C35.1498 28.2331 37.4779 27.0171 39.2383 24.9032L39.8912 25.6274C38.0461 28.1464 35.1216 29.4785 31.5157 29.4785C25.8656 29.4785 22.0893 26.6987 22.0893 22.5289C22.0893 18.2141 25.6667 15.521 31.3735 15.521ZM19.0516 0L19.0518 26.2953C19.0598 27.6889 19.3616 27.8859 21.2093 27.8859H21.919V28.9576H12.5494V27.7125C11.3004 28.8123 9.62525 29.4206 7.29691 29.4206C3.66273 29.4206 0 26.7563 0 22.471C0 18.1565 3.60594 15.521 7.29691 15.521C9.17094 15.521 11.0164 16.1002 12.2372 16.9978L12.2371 2.49295C12.2288 1.23986 11.9265 1.07134 10.0793 1.07134H8.83021V0H19.0516ZM126.024 15.608C127.671 15.608 129.119 15.8685 130.822 16.3898C131.39 16.5632 131.532 16.5923 131.759 16.5923C132.157 16.5923 132.299 16.4186 132.384 16.0423H133.321V20.328H132.384V20.0096C132.384 18.9961 132.242 18.6195 131.759 18.1853C131.049 17.5482 129.687 17.0848 128.465 17.0848C127.358 17.0848 126.762 17.4902 126.762 18.1853C126.762 18.8512 127.16 19.1987 128.835 19.8649L130.538 20.5308C133.293 21.631 134.258 22.6735 134.258 24.5845C134.258 27.5384 131.532 29.3627 127.046 29.3627C124.917 29.3627 123.128 29.0152 120.857 28.2331C120.601 28.1464 120.487 28.0885 120.317 28.0885C119.948 28.0885 119.834 28.3493 119.806 28.9576H118.841V24.0057H119.806V24.1215C119.806 25.1637 119.976 25.772 120.402 26.2645C121.197 27.1909 122.787 27.9147 124.292 27.9147C125.428 27.9147 126.024 27.5096 126.024 26.8434C126.024 26.2357 125.598 25.8299 124.462 25.3084L123.469 24.8453C120.317 23.4265 119.096 22.1526 119.096 20.1833C119.096 17.3453 121.737 15.608 126.024 15.608ZM188.175 15.521C190.929 15.521 192.916 16.216 193.995 17.6061C194.364 18.0694 194.62 18.5904 194.762 19.1117C194.932 19.7197 194.96 20.2124 194.96 21.8918V26.2357C194.96 27.4808 195.188 27.8859 196.465 27.8859H197.033V28.9576H188.373V27.7413C186.897 28.8417 185.023 29.3339 183.121 29.3339C180.225 29.3339 177.812 27.9438 177.812 25.0773C177.812 21.8918 180.878 20.7043 185.336 20.7043C186.358 20.7043 187.096 20.7334 188.147 20.8204V19.8649C188.147 18.967 188.118 18.5904 187.976 18.2141C187.692 17.4608 186.812 17.056 185.477 17.056C184.115 17.056 182.695 17.4608 181.985 18.04C181.474 18.4458 181.275 18.9961 181.275 20.0384V20.3571H180.338V16.0714H181.19C181.219 16.5632 181.332 16.7082 181.673 16.7082C181.9 16.7082 182.582 16.4768 182.809 16.4186C184.74 15.8685 186.556 15.521 188.175 15.521ZM144.025 0V22.0364L147.063 19.7779C147.431 19.4883 148.028 19.0537 148.312 18.7929C148.795 18.359 148.965 18.0982 148.965 17.7798C148.965 17.5191 148.795 17.3453 148.596 17.2589C148.375 17.1738 148.234 17.144 147.31 17.1425L146.268 17.1424V16.0714H156.205V17.1424C152.968 17.2298 150.895 17.7798 149.419 18.8803L155.637 26.8434C156.092 27.4226 156.404 27.7125 156.688 27.7989C157 27.8859 157.256 27.9147 157.568 27.9147C157.795 27.9438 157.993 27.9147 157.993 27.9147V28.9864H148.795L144.025 22.6444V26.2357C144.025 27.663 144.328 27.9079 146.133 27.9146L146.58 27.9147V28.9864H134.343V27.9147H135.053C136.901 27.9147 137.202 27.718 137.21 26.3241L137.21 2.54812C137.21 1.2634 136.935 1.07653 135.132 1.07145L133.803 1.07134V0H144.025ZM103.991 0V22.0364L107.029 19.7779C107.399 19.4883 107.995 19.0537 108.279 18.7929C108.761 18.359 108.932 18.0982 108.932 17.7798C108.932 17.5191 108.761 17.3453 108.563 17.2589C108.341 17.1738 108.201 17.144 107.276 17.1425L106.234 17.1424V16.0714H116.172V17.1424C112.935 17.2298 110.862 17.7798 109.386 18.8803L115.604 26.8434C116.058 27.4226 116.371 27.7125 116.655 27.7989C116.967 27.8859 117.222 27.9147 117.534 27.9147C117.762 27.9438 117.961 27.9147 117.961 27.9147V28.9864H108.761L103.991 22.6444V26.2357C103.991 27.663 104.295 27.9079 106.099 27.9146L106.547 27.9147V28.9864H94.3096V27.9147H95.0196C96.8933 27.9147 97.1773 27.7125 97.1773 26.2645V2.54812C97.1773 1.2634 96.9014 1.07653 95.0989 1.07145L93.7702 1.07134V0H103.991ZM173.212 15.521C175.739 15.521 177.244 17.1715 177.244 19.6332C177.244 21.9785 175.796 23.5423 173.61 23.5423C171.508 23.5423 170.032 22.1526 170.032 20.2124C170.032 19.7488 170.089 19.3437 170.316 18.9088C170.429 18.6774 170.401 18.6483 170.401 18.5904C170.401 18.5037 170.288 18.4458 170.231 18.4458C170.004 18.4458 169.72 18.7065 169.407 19.1987C169.095 19.6909 168.811 20.3859 168.669 20.9654C168.527 21.6022 168.442 22.5289 168.442 23.7452V26.3803C168.442 27.6645 168.746 27.8798 170.547 27.8858L171.452 27.8859V28.9576H159.016V27.8859H159.555C161.429 27.8859 161.713 27.7125 161.713 26.4091V18.6483C161.713 17.3453 161.429 17.1715 159.555 17.1715H158.704V16.1002H168.215V19.4595C169.18 16.9978 170.94 15.521 173.212 15.521ZM90.42 0V26.3803C90.42 27.6934 90.7236 27.8807 92.5269 27.8858L93.3162 27.8859V28.9576H80.7383V27.8859H81.448C83.322 27.8859 83.606 27.7125 83.606 26.4091V2.54812C83.606 1.2634 83.3301 1.07653 81.5273 1.07145L80.1989 1.07134V0H90.42ZM212.763 15.6948C214.637 15.6948 216.113 16.274 217.05 17.3162C217.561 17.8954 217.845 18.6195 217.958 19.4304C218.013 19.9647 218.042 20.2261 218.043 21.2758L218.044 26.3221C218.044 27.6929 218.32 27.8808 220.123 27.8858L220.656 27.8859V28.9576H209.696V28.0017H209.867C210.434 28.0017 210.775 27.9147 210.974 27.7125C211.168 27.5143 211.226 27.0399 211.229 25.8577L211.229 20.878C211.229 19.3149 211.201 18.8803 211.059 18.5616C210.86 18.0694 210.378 17.8378 209.668 17.8378C208.93 17.8378 208.22 18.1565 207.794 18.6774C207.368 19.2275 207.283 19.6332 207.283 21.2835V25.772C207.283 26.3512 207.311 26.9013 207.339 27.1909C207.425 27.7701 207.737 28.0017 208.646 28.0017H208.93V28.9576H197.97V27.8859H198.509C200.383 27.8859 200.667 27.7125 200.667 26.3221V18.7357C200.667 17.3453 200.383 17.1715 198.509 17.1715H197.658V16.1002H207.056V17.9536C208.589 16.4477 210.52 15.6948 212.763 15.6948ZM52.8571 0V1.07134H52.3178C51.182 1.07134 50.9834 1.10014 50.5859 1.15838C49.9044 1.2739 49.5351 1.67933 49.5351 2.28732C49.5351 2.80859 49.7623 3.4457 50.4152 4.95159L56.9456 19.8937L63.0217 5.9071C63.4478 4.92279 64.1575 3.32986 64.1575 2.4614C64.1575 1.40348 63.4934 1.07963 61.0179 1.0715L60.6934 1.07134V0H69.5521V1.07134H69.2682C68.2457 1.07134 67.962 1.10014 67.5927 1.18718C66.9966 1.36094 66.7127 1.73725 66.1448 3.04027L54.6176 28.9576H53.1125L41.4149 3.04027C41.131 2.43228 40.8756 1.91101 40.6484 1.65053C40.2791 1.21598 39.7966 1.07134 38.4337 1.07134H37.4686V0H52.8571ZM70.8012 16.5632C69.6936 16.5632 68.8703 17.056 68.5578 17.8669C68.3592 18.3881 68.3307 18.9382 68.3307 20.8489L68.3307 24.1953C68.3317 25.956 68.3606 26.5218 68.5368 27.018L68.5578 27.075C68.8703 27.8859 69.6936 28.3781 70.8012 28.3781C71.9084 28.3781 72.7317 27.8859 73.0442 27.075C73.2355 26.573 73.2687 26.0437 73.2709 24.2955V20.6458C73.2688 18.9642 73.238 18.4109 73.0652 17.9239L73.0442 17.8669C72.7317 17.056 71.9084 16.5632 70.8012 16.5632ZM9.53991 17.1424C7.26194 17.1424 7.21295 19.0279 7.2119 21.0081V23.9335C7.21295 25.9137 7.26194 27.7989 9.53991 27.7989C10.2782 27.7989 10.9879 27.5096 11.4707 27.075C12.0954 26.525 12.294 25.772 12.294 24.0345V20.9071C12.294 19.1696 12.0954 18.417 11.4707 17.8669C10.9879 17.4323 10.2782 17.1424 9.53991 17.1424ZM187.421 21.8627L187.295 21.8627C185.534 21.8627 184.881 22.5577 184.881 24.3817V24.99C184.881 26.7563 185.279 27.4226 186.386 27.4226C186.954 27.4226 187.465 27.1909 187.777 26.8146C188.004 26.525 188.09 26.2066 188.118 25.5983C188.147 25.1064 188.147 24.6428 188.147 24.3244V21.8918C187.702 21.8664 187.665 21.8631 187.421 21.8627ZM31.2885 16.6215C30.3516 16.6215 29.6134 16.969 29.2727 17.5482C28.9605 18.0694 28.9037 18.5616 28.9037 20.3859V21.5734H33.503L33.5029 20.2288C33.5016 18.8288 33.4702 18.3465 33.2758 17.809C33.0204 17.056 32.3107 16.6215 31.2885 16.6215Z" fill="black"/> </svg> </a>  </div> <div class="app-header-article__menu">  <div class="menu__share-toggle fjs-share-toggle" data-gtm="share / top / toggle"> <svg class="menu__share-toggle-icon menu__share-toggle-icon--article" width="17" height="18"> <title>Deel dit artikel</title> <path d="M13.5 10.5c-.8 0-1.56.4-2.03 1.05L4.91 8.62c.11-.38.12-.78.02-1.17l6.55-2.98a2.54 2.54 0 1 0-.41-.91L4.52 6.54a2.5 2.5 0 1 0-.05 2.98l6.6 2.94a2.5 2.5 0 1 0 2.43-1.96zm0-9a1.5 1.5 0 1 1 0 3 1.5 1.5 0 0 1 0-3zm-11 8a1.5 1.5 0 1 1 0-3 1.5 1.5 0 0 1 0 3zm11 5a1.5 1.5 0 1 1 0-3 1.5 1.5 0 0 1 0 3z"/> </svg> <svg class="menu__share-toggle-icon menu__share-toggle-icon--ios" width="17" height="22"> <g fill="#000" fill-rule="nonzero"> <path d="M15.5 6h-5a.5.5 0 1 0 0 1H15v14H1V7h4.5a.5.5 0 0 0 0-1h-5a.5.5 0 0 0-.5.5v15a.5.5 0 0 0 .5.5h15a.5.5 0 0 0 .5-.5v-15a.5.5 0 0 0-.5-.5z"/> <path d="M4.825 3.88L7.5 1.586V13.22a.5.5 0 1 0 1 0V1.587l2.675 2.292a.5.5 0 1 0 .65-.758l-3.5-3a.498.498 0 0 0-.65 0l-3.5 3a.5.5 0 1 0 .65.758z"/> </g> </svg> </div> <div class="menu__share-modal menu__share-modal--hidden fjs-share-modal"> <div class="share-modal__wrapper"> <a class="share-full-icon share-full-icon--facebook" target="_blank" rel="noopener noreferrer" href="https://www.facebook.com/sharer/sharer.php?u=https://www.volkskrant.nl/fs-b34416f7" data-gtm="share / top / facebook"> <svg viewbox="0 0 36 36"> <g> <circle fill="#45619E" cx="18" cy="18" r="18"/> <path fill="#fff" d="M15 15.3h1.392v-1.307c0-.575.016-1.463.45-2.014.456-.587 1.083-.98 2.16-.98 1.758 0 2.498.24 2.498.24l-.348 1.99s-.58-.16-1.122-.16c-.542 0-1.027.19-1.027.71v1.52h2.22l-.154 1.942H19V24h-2.61v-6.756H15V15.3"/> </g> </svg> </a> <a class="share-full-icon share-full-icon--twitter" target="_blank" rel="noopener noreferrer" href="https://twitter.com/intent/tweet/?url=https://www.volkskrant.nl/ts-b34416f7&amp;via=volkskrant&amp;text=Jacht op Kosovaren stelt NAVO voor dilemma" data-gtm="share / top / twitter"> <svg viewbox="0 0 36 36"> <defs> <path id="a" d="M.003.04h12.624v11.027H.003"/> </defs> <g> <circle fill="#1BABE9" cx="18" cy="18" r="18"/> <g transform="translate(12 13)"> <path fill="#fff" d="M12.627 1.346c-.464.222-.963.372-1.488.44.53-.346.94-.89 1.14-1.542-.507.32-1.06.55-1.65.676C10.16.38 9.49.04 8.74.04c-1.43 0-2.59 1.247-2.59 2.785 0 .218.023.43.067.635C4.062 3.343 2.153 2.235.88.55c-.224.412-.352.89-.352 1.4 0 .966.458 1.818 1.153 2.318-.42-.015-.82-.14-1.17-.35v.036c0 1.35.895 2.475 2.08 2.73-.217.064-.446.098-.682.098-.167 0-.33-.017-.488-.05.33 1.106 1.286 1.91 2.42 1.934-.89.747-2.007 1.192-3.22 1.192-.21 0-.416-.013-.62-.04 1.147.79 2.51 1.252 3.97 1.252 4.767 0 7.373-4.243 7.373-7.922 0-.12-.002-.24-.007-.36.506-.393.945-.883 1.292-1.442" mask="url(#b)"/> </g> </g> </svg> </a> <a class="share-full-icon share-full-icon--mail" href="mailto:?subject=Jacht op Kosovaren stelt NAVO voor dilemma&amp;body=https://www.volkskrant.nl/es-b34416f7" data-gtm="share / top / mail"> <svg viewbox="0 0 36 36"> <g> <path d="M18 36c9.94 0 18-8.06 18-18S27.94 0 18 0 0 8.06 0 18s8.06 18 18 18z"/> <path fill="#fff" d="M18.1 19.872l-1.608-1.423-4.6 3.98c.17.156.394.254.643.254h11.13c.248 0 .473-.1.64-.256l-4.597-3.985L18.1 19.87zm6.207-6.318c-.168-.157-.392-.254-.642-.254h-11.13c-.248 0-.473.098-.64.256l6.205 5.377 6.207-5.38zm-12.707.57v7.8l4.488-3.855-4.488-3.95zm8.512 3.945l4.488 3.85v-7.8l-4.488 3.947z"/> </g> </svg> </a> <a class="share-full-icon share-full-icon--whatsapp" href="whatsapp://send?text=Jacht op Kosovaren stelt NAVO voor dilemma https://www.volkskrant.nl/ws-b34416f7" data-gtm="share / top / whatsapp"> <svg viewbox="0 0 36 36"> <g> <circle fill="#69C063" cx="18" cy="18" r="18"/> <path fill="#fff" d="M18.137 12C14.753 12 12 14.754 12 18.14c0 1.34.433 2.586 1.17 3.597l-.766 2.28 2.36-.753c.97.642 2.128 1.014 3.377 1.014 3.39 0 6.14-2.754 6.14-6.14 0-3.384-2.75-6.138-6.132-6.138h-.003zm-1.714 3.118c-.12-.285-.21-.296-.39-.303-.06-.004-.13-.007-.205-.007-.235 0-.48.068-.628.22-.18.184-.628.613-.628 1.494 0 .88.643 1.732.73 1.852.09.12 1.25 1.952 3.055 2.7 1.41.584 1.83.53 2.15.46.47-.1 1.058-.447 1.206-.865.148-.42.148-.78.104-.86-.043-.076-.162-.12-.343-.21-.18-.09-1.057-.523-1.223-.58-.16-.06-.31-.04-.44.133-.17.24-.34.487-.48.634-.1.116-.28.13-.43.07-.2-.084-.75-.28-1.44-.89-.53-.47-.89-1.06-.99-1.237-.105-.18-.01-.284.07-.38.09-.11.178-.19.268-.296.09-.106.14-.16.2-.283.06-.12.017-.24-.027-.33-.044-.09-.405-.97-.553-1.326z"/> </g> </svg> </a> </div> <p class="share-modal__copy"> <input type="hidden" value="https://www.volkskrant.nl/cs-b34416f7" class="share-modal__copy-input fjs-share-modal-copy-input"> <a href="https://www.volkskrant.nl/cs-b34416f7" class="share-modal__copy-link fjs-share-modal-copy-link" data-gtm="share / top / copylink"> <svg width="20" height="10" class="share-modal__copy-icon"> <path fill="#000" fill-rule="nonzero" d="M1.9 5c0-1.7 1.4-3.1 3.1-3.1h4V0H5a5 5 0 0 0 0 10h4V8.1H5A3.1 3.1 0 0 1 1.9 5zM6 6h8V4H6v2zm9-6h-4v1.9h4a3.1 3.1 0 0 1 0 6.2h-4V10h4a5 5 0 0 0 0-10z"/> </svg> <span class="share-modal-copy-link-text fjs-share-modal-copy-link-text" data-click-value="Link gekopieerd"> Kopieer de link van het artikel </span> </a> </p> </div> <div class="menu__share-inline"> <a class="share-icon share-icon--facebook" target="_blank" rel="noopener noreferrer" href="https://www.facebook.com/sharer/sharer.php?u=https://www.volkskrant.nl/fs-b34416f7" data-gtm="share / top / facebook"> <svg width="9" height="22" viewBox="0 0 9 22"> <path d="M1.943 3.672v2.61H.031v3.193h1.912v9.486h3.93V9.475h2.635s.247-1.53.367-3.204H5.887V4.088c0-.326.428-.765.852-.765h2.14V0H5.97C1.846 0 1.943 3.195 1.943 3.672z"/> </svg> </a> <a class="share-icon share-icon--twitter" target="_blank" rel="noopener noreferrer" href="https://twitter.com/intent/tweet/?url=https://www.volkskrant.nl/ts-b34416f7&amp;via=volkskrant&amp;text=Jacht op Kosovaren stelt NAVO voor dilemma" data-gtm="share / top / twitter"> <svg width="20" height="22" viewBox="0 0 20 16"> <path d="M17.64 3.94c.007.17.012.342.012.514 0 5.236-4.075 11.27-11.53 11.27-2.288 0-4.418-.653-6.211-1.778.317.037.64.055.966.055a8.25 8.25 0 0 0 5.033-1.695c-1.773-.032-3.27-1.178-3.785-2.752a4.174 4.174 0 0 0 1.83-.067C2.1 9.124.704 7.523.704 5.602v-.05c.546.297 1.17.475 1.835.496A3.937 3.937 0 0 1 .738 2.75c0-.727.199-1.408.548-1.993a11.594 11.594 0 0 0 8.351 4.14 3.873 3.873 0 0 1-.104-.903c0-2.187 1.814-3.961 4.051-3.961 1.166 0 2.218.48 2.958 1.25a8.2 8.2 0 0 0 2.574-.961 3.997 3.997 0 0 1-1.784 2.192 8.2 8.2 0 0 0 2.327-.623 8.052 8.052 0 0 1-2.02 2.048z"/> </svg> </a> <a class="share-icon share-icon--linkedin" target="_blank" rel="noopener noreferrer" href="https://www.linkedin.com/shareArticle?mini=true&amp;url=https://www.volkskrant.nl/ls-b34416f7" data-gtm="share / top / linkedin"> <svg width="20" height="20" viewBox="0 0 20 20"> <path d="M19.835 11.646v7.334h-4.25v-6.842c0-1.72-.616-2.892-2.155-2.892-1.174 0-1.873.79-2.181 1.555-.112.273-.141.654-.141 1.037v7.142H6.855s.057-11.589 0-12.79h4.253v1.813l-.028.042h.028v-.042c.565-.87 1.573-2.113 3.832-2.113 2.798 0 4.895 1.828 4.895 5.756zM2.406.026C.952.026 0 .98 0 2.235c0 1.228.924 2.21 2.35 2.21h.028c1.484 0 2.406-.982 2.406-2.21C4.756.98 3.862.025 2.406.025zM.253 18.98h4.25V6.19H.254v12.79z"/> </svg> </a> <a class="share-icon share-icon--article fjs-share-modal-desktop-copy-link" data-gtm="share / top / copylink" href="https://www.volkskrant.nl/cs-b34416f7"> <input type="hidden" value="https://www.volkskrant.nl/cs-b34416f7" class="share-modal__copy-input fjs-share-modal-desktop-copy-input"> <svg width="20" height="24" viewBox="0 0 20 10"> <path d="M1.9 5c0-1.7 1.4-3.1 3.1-3.1h4V0H5a5 5 0 0 0 0 10h4V8.1H5A3.1 3.1 0 0 1 1.9 5zM6 6h8V4H6v2zm9-6h-4v1.9h4a3.1 3.1 0 0 1 0 6.2h-4V10h4a5 5 0 0 0 0-10z"/> </svg> </a> <a class="share-icon share-icon--mail" target="_blank" rel="noopener noreferrer" href="mailto:?subject=Jacht op Kosovaren stelt NAVO voor dilemma&amp;body=https://www.volkskrant.nl/es-b34416f7" data-gtm="share / top / mail"> <svg width="20" height="20" viewBox="0 0 20 10"> <path d="M9.819 9.775l-2.43-2.117-6.946 5.927c.253.233.594.379.97.379h16.812c.374 0 .714-.146.965-.379l-6.942-5.927-2.43 2.117zM19.194.378a1.416 1.416 0 0 0-.97-.378H1.413a1.41 1.41 0 0 0-.966.381l9.373 7.997 9.375-8zM0 1.226v11.601l6.779-5.735L0 1.226zm12.858 5.866l6.78 5.735V1.222l-6.78 5.87z"/> </svg> </a>  </div>  </div> </div> </header>  <div class="app-navigation-anchor"></div> <nav class="app-navigation fjs-app-navigation"> <div class="app-navigation__sections-wrapper fjs-app-navigation-wrapper"> <ul class="app-navigation__sections fjs-app-navigation-sections"> <li class="app-navigation__section fjs-app-navigation-section   voorpagina" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/#voorpagina" data-gtm="navigation / voorpagina">Voorpagina</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section  app-navigation__section--active nieuws-achtergrond" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/nieuws-achtergrond" data-gtm="navigation / nieuws-achtergrond">Nieuws &amp; Achtergrond</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   columns-opinie" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/columns-opinie" data-gtm="navigation / columns-opinie">Columns &amp; Opinie</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   uitgelicht" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/uitgelicht" data-gtm="navigation / uitgelicht">Uitgelicht</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   wetenschap" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/wetenschap" data-gtm="navigation / wetenschap">Wetenschap</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   mensen" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/mensen" data-gtm="navigation / mensen">Mensen</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   beter-leven" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/beter-leven" data-gtm="navigation / beter-leven">Beter Leven</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   cultuur-media" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/cultuur-media" data-gtm="navigation / cultuur-media">Cultuur &amp; Media</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   foto" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/foto" data-gtm="navigation / foto">Foto</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   economie" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/economie" data-gtm="navigation / economie">Economie</a> </div> </li> <li class="app-navigation__section fjs-app-navigation-section   sport" style="border-color:#000000;"> <div class="app-navigation__section-wrapper"> <a class="app-navigation__link fjs-app-navigation-link" href="/sport" data-gtm="navigation / sport">Sport</a> </div> </li> </ul> <div class="app-navigation__button app-navigation__button--previous fjs-button-previous"> <svg xmlns="http://www.w3.org/2000/svg" width="7" height="12" class="app-navigation__button__arrow"> <path fill="#9B9B9B" d="M5.676 12L7 10.6 2.65 6 7 1.4 5.676 0 0 6l3.75 3.966z"/> </svg> </div> <div class="app-navigation__button app-navigation__button--next fjs-button-next"> <svg xmlns="http://www.w3.org/2000/svg" width="7" height="12" class="app-navigation__button__arrow"> <path fill="#9B9B9B" d="M1.824.5L.5 1.9l4.35 4.6L.5 11.1l1.324 1.4 5.676-6-3.75-3.966z"/> </svg> </div> </div> </nav>   <main id="main-content"> <article class="artstyle artstyle--type-default fu--type-default artstyle--padding-top-s-l fu--padding-top-s-l" data-gtm="article"><header class="artstyle__header "  data-element-id="article-element-header"><h1 class="artstyle__header-title " data-element-id="article-element-headertitle">\n    Jacht op Kosovaren stelt NAVO voor dilemma\n  </h1><p class="artstyle__intro " data-element-id="834821">Servi\xc3\xab graaft zich in. In Kosovo woedt een moorddadige terreur tegen de Albanese burgers. Wat kan de NAVO doen?..\n  </p><section class="artstyle__production artstyle__production--type-plus fu--type-plus"    data-element-id="article-element-authors"><span class="artstyle__production__authors"><a class="artstyle__production__author" href="/auteur/Van onze buitenlandredactie Anet Bleich">Van onze buitenlandredactie Anet Bleich</a></span><time class="artstyle__production__datetime" datetime="29 maart 1999"><span class="artstyle__production__date">29 maart 1999</span><span class="artstyle__production__time">, 00:00</span></time><span class="artstyle__production-decoration"><svg width="184" height="11" viewBox="0 0 184 11" fill="none" xmlns="http://www.w3.org/2000/svg"><mask id="path-1-inside-1" fill="white"><path fill-rule="evenodd" clip-rule="evenodd" d="M86.9133 5.08154L81.8318 0L76.7745 5.05725H0V5.65725H76.7288L81.8318 10.7603L86.9133 5.67871L91.9949 10.7603L96.9948 5.76038L101.995 10.7603L107.098 5.65725H184V5.05725H107.052L101.995 0L96.9948 4.99988L91.9949 0L86.9133 5.08154Z"/></mask><path fill-rule="evenodd" clip-rule="evenodd" d="M86.9133 5.08154L81.8318 0L76.7745 5.05725H0V5.65725H76.7288L81.8318 10.7603L86.9133 5.67871L91.9949 10.7603L96.9948 5.76038L101.995 10.7603L107.098 5.65725H184V5.05725H107.052L101.995 0L96.9948 4.99988L91.9949 0L86.9133 5.08154Z" fill="white"/><path d="M81.8318 0L82.2122 -0.380432L81.8318 -0.760863L81.4514 -0.380432L81.8318 0ZM86.9133 5.08154L86.5329 5.46197L86.9133 5.84241L87.2938 5.46197L86.9133 5.08154ZM76.7745 5.05725V5.59526H76.9974L77.155 5.43768L76.7745 5.05725ZM0 5.05725V4.51924H-0.538012V5.05725H0ZM0 5.65725H-0.538012V6.19526H0V5.65725ZM76.7288 5.65725L77.1092 5.27682L76.9516 5.11924H76.7288V5.65725ZM81.8318 10.7603L81.4514 11.1407L81.8318 11.5211L82.2122 11.1407L81.8318 10.7603ZM86.9133 5.67871L87.2938 5.29828L86.9133 4.91785L86.5329 5.29828L86.9133 5.67871ZM91.9949 10.7603L91.6144 11.1407L91.9949 11.5211L92.3753 11.1407L91.9949 10.7603ZM96.9948 5.76038L97.3752 5.37995L96.9948 4.99951L96.6143 5.37995L96.9948 5.76038ZM101.995 10.7603L101.614 11.1407L101.995 11.5211L102.375 11.1407L101.995 10.7603ZM107.098 5.65725V5.11924H106.875L106.717 5.27682L107.098 5.65725ZM184 5.65725V6.19526H184.538V5.65725H184ZM184 5.05725H184.538V4.51924H184V5.05725ZM107.052 5.05725L106.671 5.43768L106.829 5.59526H107.052V5.05725ZM101.995 0L102.375 -0.380432L101.995 -0.760863L101.614 -0.380432L101.995 0ZM96.9948 4.99988L96.6143 5.38031L96.9948 5.76074L97.3752 5.38031L96.9948 4.99988ZM91.9949 0L92.3753 -0.380432L91.9949 -0.760863L91.6144 -0.380432L91.9949 0ZM81.4514 0.380432L86.5329 5.46197L87.2938 4.70111L82.2122 -0.380432L81.4514 0.380432ZM77.155 5.43768L82.2122 0.380432L81.4514 -0.380432L76.3941 4.67682L77.155 5.43768ZM76.7745 4.51924H0V5.59526H76.7745V4.51924ZM-0.538012 5.05725V5.65725H0.538012V5.05725H-0.538012ZM0 6.19526H76.7288V5.11924H0V6.19526ZM82.2122 10.3798L77.1092 5.27682L76.3484 6.03768L81.4514 11.1407L82.2122 10.3798ZM86.5329 5.29828L81.4514 10.3798L82.2122 11.1407L87.2938 6.05914L86.5329 5.29828ZM92.3753 10.3798L87.2938 5.29828L86.5329 6.05914L91.6144 11.1407L92.3753 10.3798ZM96.6143 5.37995L91.6144 10.3798L92.3753 11.1407L97.3752 6.14081L96.6143 5.37995ZM102.375 10.3798L97.3752 5.37995L96.6143 6.14081L101.614 11.1407L102.375 10.3798ZM106.717 5.27682L101.614 10.3798L102.375 11.1407L107.478 6.03768L106.717 5.27682ZM107.098 6.19526H184V5.11924H107.098V6.19526ZM184.538 5.65725V5.05725H183.462V5.65725H184.538ZM184 4.51924H107.052V5.59526H184V4.51924ZM101.614 0.380432L106.671 5.43768L107.432 4.67682L102.375 -0.380432L101.614 0.380432ZM97.3752 5.38031L102.375 0.380432L101.614 -0.380432L96.6143 4.61945L97.3752 5.38031ZM91.6144 0.380432L96.6143 5.38031L97.3752 4.61945L92.3753 -0.380432L91.6144 0.380432ZM87.2938 5.46197L92.3753 0.380432L91.6144 -0.380432L86.5329 4.70111L87.2938 5.46197Z" fill="black" mask="url(#path-1-inside-1)"/></svg></span></section></header><section class="artstyle__main"><section class="artstyle__main--container"><div class="artstyle__freecontentdelimiter" data-element-id="article-element-freecontentdelimiter"></div><div data-temptation-position="ARTICLE_INLINE"></div><p class="artstyle__paragraph " data-element-id="834822">\n      <p>Technisch verlopen de NAVO-bombardementen vrijwel helemaal naar wens. De luchtaanvallen hebben het militaire potentieel van Joegoslavi\xc3\xab zware klappen toegebracht. Het oogmerk is echter bij uitstek politiek: de bommen en raketten dienen als een laatste, welhaast onontkoombaar argument, dat Milosevic moet dwingen in te stemmen met een compromis voor Kosovo.</p><p>Juist op dit politieke front doen zich twee uitermate zorgwekkende ontwikkelingen voor, die een ernstige schaduw werpen op het militaire succes van de NAVO-operatie. In de eerste plaats ontwikkelt de toestand in Kosovo zelf zich catastrofaal. De Albanezen zijn ten prooi aan totale willekeur en worden met verdrijving, mishandeling of de dood geconfronteerd.</p><p>Het is een wrange paradox dat het NAVO-optreden, dat beoogt de Kosovaren te beschermen tegen een r\xc3\xbccksichtsloze anti-guerrillacampagne, gepaard met incidentele massamoorden, nu als effect heeft dat tegen die Kosovaren de gruwelijke methoden van systematische etnische zuivering - bekend uit de oorlog in Bosni\xc3\xab - worden ge\xc3\xafntroduceerd. Te denken valt aan het loslaten van paramilitaire bandieten die huizen doorzoeken, mensen verjagen, vrouwen en mannen van elkaar scheiden en de laatsten vasthouden (wie denkt hierbij niet aan Srebrenica?)</p><p>Het is een logische reactie van de NAVO om zich versneld te gaan richten op het uitschakelen van het militaire materieel van de Servi\xc3\xabrs in Kosovo. Daarmee kan mogelijk inderdaad ernstige schade worden toegebracht aan de Servische tanks die Albanese dorpen in puin schieten en logistieke steun verlenen aan de paramilitaire bendes. Hun terreur kan echter niet met bombardementen worden gestopt: dat kan alleen door een opdracht uit Belgrado.</p><p>Juist hier wordt de tweede zwakke plek van de NAVO-operatie zichtbaar. Want de conclusie lijkt ononontkoombaar dat de bombardementen leiden tot solidariteit onder de Servi\xc3\xabrs. Zij keren zich niet tegen de leider die dit bommentapijt over Servi\xc3\xab heeft afgeroepen. Ze scharen zich integendeel \xc3\xa1chter Milosevic, omdat ze zich diep gekrenkt voelen in hun nationale trots.</p><p>Moet uit deze twee zware tegenslagen worden afgeleid dat de militair succesvolle NAVO-campagne op een politiek debacle dreigt uit te lopen? Zijn er alternatieve scenario\'s, en welke dan?</p><p>E\xc3\xa9n theoretische mogelijkheid zou zijn om een volgende stap op de escalatieladder te zetten en een actie met grondtroepen voor te bereiden. De verontwaardiging over Milosevic\' uitdagende optreden en vooral over het inferno in Kosovo maakt deze optie op het eerste gezicht aantrekkelijk. Maar er kleven ernstige bezwaren aan. Om te beginnen zijn de politici en de publieke opinie in Amerika en Europa absoluut niet rijp voor het sturen van tien- of honderdduizenden militairen om een landoorlog te voeren, met alle fysieke risico\'s vandien.</p><p>In de tweede plaats zou een full-scale oorlog naar alle waarschijnlijkheid leiden tot een complete breuk met Moskou, dat zich tot nu toe beperkt tot een verbaal protest. Een nieuwe koude oorlog en een machtsovername in Rusland door communisten en nationalisten komen dan in het vizier.</p><p>Bovendien zou zo\'n grondoorlog de problemen die door de luchtacties zijn blootgelegd, niet oplossen, zeker niet op korte termijn. In de weken tot aan het in actie komen van de grondtroepen, zou de dreigende genocide in Kosovo verder haar beslag kunnen krijgen. En het Servische nationalisme zou nog meer worden aangewakkerd. Misschien zou binnen een maand of twee Belgrado kunnen worden veroverd. Maar de prijs - voor Kosovaren, Servi\xc3\xabrs, geallieerde militairen en de toekomstige stabiliteit in Europa - wordt astronomisch hoog.</p><p>Een andere optie is het staken van de luchtacties en het hervatten van de diplomatie. Dat heeft het niet te onderschatten voordeel dat in Kosovo wellicht een adempauze ontstaat. De andere kant van de medaille liegt er echter ook niet om. Milosevic zal victorie kraaien, de NAVO lijdt een smadelijk gezichtsverlies, dat op termijn zelfs de cohesie van de alliantie kan bedreigen. Hoewel de genocide voor het moment wordt afgewend, slinken de kansen op een politieke oplossing ongeveer tot het nulpunt. Op deze wijze komt namelijk vast te staan dat Milosevic in Kosovo de vrije hand heeft. De Albanezen zullen zich daarbij niet neerleggen, de guerrilla zal verder escaleren.</p><p>De NAVO zal derhalve tussen de geschetste uitersten moeten laveren. Een scenario dat denkbaar is, behelst een driesporen-strategie. Ten eerste het voortzetten van de luchtaanvallen. Ten tweede moet Milosevic te horen krijgen dat hij grote kans maakt door het Joegoslavi\xc3\xab-tribunaal te worden aangeklaagd. Ten derde moet hem worden duidelijk gemaakt dat er een begaanbare uitweg is: een wapenstilstand. Stopzetting van de Servische acties in Kosovo en van de bombardementen, gekoppeld aan de openlijk uitgesproken Joegoslavische toezegging het Rambouillet-overleg te goeder trouw te hervatten.</p><p>Om Milosevic ertoe te brengen uit zijn schuttersput te komen, kan de diplomatieke missie van de Russische liberalen Gajdar, Nemtsov en Fjodorov van grote betekenis zijn.</p>\n    </p><div data-temptation-position="ARTICLE_BOTTOM"></div><div class="artstyle__footer" data-footer></div></section></section><template id="article-no-consent-notification">\n      <div class="artstyle__disabled-embed">\n        <h4 class="artstyle__disabled-embed__title">Hier zou content moeten staan van bijv. Twitter, Facebook of Instagram</h4>\n        <p class="artstyle__disabled-embed__text">Om u deze content te kunnen laten zien, hebben wij uw toestemming nodig om cookies te plaatsen.\n  <a href="#" onclick="window._privacy.openModal();">Open uw cookie-instellingen</a> om te kiezen welke cookies u wilt accepteren.\n  Voor een optimale gebruikservaring van onze site selecteert u "Accepteer alles".\n  U kunt ook alleen de sociale content aanzetten: vink hiervoor "Cookies accepteren van sociale media" aan.</p>\n      </div>\n    </template><template id="article-loader">\n    <div class="loading artstyle__loader">\n      <svg class="spinner" viewBox="0 0 50 50">\n        <circle class="path" cx="25" cy="25" r="20" fill="none" stroke-width="5">\n          <title>Loading</title>\n        </circle>\n      </svg>\n    </div>\n  </template><div class="pswp" tabindex="-1" role="dialog" aria-hidden="true">\n    <div class="pswp__bg"></div>\n    <div class="pswp__scroll-wrap">\n      <div class="pswp__container">\n        <div class="pswp__item"></div>\n        <div class="pswp__item"></div>\n        <div class="pswp__item"></div>\n      </div>\n      <div class="pswp__ui pswp__ui--hidden">\n        <div class="pswp__top-bar">\n          <div class="pswp__counter"></div>\n          <button class="pswp__button pswp__button--close" title="Close (Esc)"></button>\n          <button class="pswp__button pswp__button--fs" title="Toggle fullscreen"></button>\n          <button class="pswp__button pswp__button--zoom" title="Zoom in/out"></button>\n          <div class="pswp__preloader"></div>\n        </div>\n        <div class="pswp__share-modal pswp__share-modal--hidden pswp__single-tap">\n          <div class="pswp__share-tooltip"></div>\n        </div>\n        <button class="pswp__button pswp__button--arrow--left" title="Previous (arrow left)"></button>\n        <button class="pswp__button pswp__button--arrow--right" title="Next (arrow right)"></button>\n        <div class="pswp__caption">\n          <div class="pswp__caption__center"></div>\n        </div>\n      </div>\n    </div>\n  </div></article> <div class="article-footer__sharing"> <a class="share-full-icon share-full-icon--facebook" target="_blank" rel="noopener noreferrer" href="https://www.facebook.com/sharer/sharer.php?u=https://www.volkskrant.nl/fs-b34416f7" data-gtm="share / footer / facebook"> <svg viewbox="0 0 36 36"> <g> <circle fill="#45619E" cx="18" cy="18" r="18"/> <path fill="#fff" d="M15 15.3h1.392v-1.307c0-.575.016-1.463.45-2.014.456-.587 1.083-.98 2.16-.98 1.758 0 2.498.24 2.498.24l-.348 1.99s-.58-.16-1.122-.16c-.542 0-1.027.19-1.027.71v1.52h2.22l-.154 1.942H19V24h-2.61v-6.756H15V15.3"/> </g> </svg> </a> <a class="share-full-icon share-full-icon--twitter" target="_blank" rel="noopener noreferrer" href="https://twitter.com/intent/tweet/?url=https://www.volkskrant.nl/ts-b34416f7&amp;via=volkskrant&amp;text=Jacht op Kosovaren stelt NAVO voor dilemma" data-gtm="share / footer / twitter"> <svg viewbox="0 0 36 36"> <defs> <path id="a" d="M.003.04h12.624v11.027H.003"/> </defs> <g> <circle fill="#1BABE9" cx="18" cy="18" r="18"/> <g transform="translate(12 13)"> <path fill="#fff" d="M12.627 1.346c-.464.222-.963.372-1.488.44.53-.346.94-.89 1.14-1.542-.507.32-1.06.55-1.65.676C10.16.38 9.49.04 8.74.04c-1.43 0-2.59 1.247-2.59 2.785 0 .218.023.43.067.635C4.062 3.343 2.153 2.235.88.55c-.224.412-.352.89-.352 1.4 0 .966.458 1.818 1.153 2.318-.42-.015-.82-.14-1.17-.35v.036c0 1.35.895 2.475 2.08 2.73-.217.064-.446.098-.682.098-.167 0-.33-.017-.488-.05.33 1.106 1.286 1.91 2.42 1.934-.89.747-2.007 1.192-3.22 1.192-.21 0-.416-.013-.62-.04 1.147.79 2.51 1.252 3.97 1.252 4.767 0 7.373-4.243 7.373-7.922 0-.12-.002-.24-.007-.36.506-.393.945-.883 1.292-1.442" mask="url(#b)"/> </g> </g> </svg> </a> <a class="share-full-icon share-full-icon--mail" href="mailto:?subject=Jacht op Kosovaren stelt NAVO voor dilemma&amp;body=https://www.volkskrant.nl/es-b34416f7" data-gtm="share / footer / mail"> <svg viewbox="0 0 36 36"> <g> <path d="M18 36c9.94 0 18-8.06 18-18S27.94 0 18 0 0 8.06 0 18s8.06 18 18 18z"/> <path fill="#fff" d="M18.1 19.872l-1.608-1.423-4.6 3.98c.17.156.394.254.643.254h11.13c.248 0 .473-.1.64-.256l-4.597-3.985L18.1 19.87zm6.207-6.318c-.168-.157-.392-.254-.642-.254h-11.13c-.248 0-.473.098-.64.256l6.205 5.377 6.207-5.38zm-12.707.57v7.8l4.488-3.855-4.488-3.95zm8.512 3.945l4.488 3.85v-7.8l-4.488 3.947z"/> </g> </svg> </a> <a class="share-full-icon share-full-icon--whatsapp" href="whatsapp://send?text=Jacht op Kosovaren stelt NAVO voor dilemma https://www.volkskrant.nl/ws-b34416f7" data-gtm="share / footer / whatsapp"> <svg viewbox="0 0 36 36"> <g> <circle fill="#69C063" cx="18" cy="18" r="18"/> <path fill="#fff" d="M18.137 12C14.753 12 12 14.754 12 18.14c0 1.34.433 2.586 1.17 3.597l-.766 2.28 2.36-.753c.97.642 2.128 1.014 3.377 1.014 3.39 0 6.14-2.754 6.14-6.14 0-3.384-2.75-6.138-6.132-6.138h-.003zm-1.714 3.118c-.12-.285-.21-.296-.39-.303-.06-.004-.13-.007-.205-.007-.235 0-.48.068-.628.22-.18.184-.628.613-.628 1.494 0 .88.643 1.732.73 1.852.09.12 1.25 1.952 3.055 2.7 1.41.584 1.83.53 2.15.46.47-.1 1.058-.447 1.206-.865.148-.42.148-.78.104-.86-.043-.076-.162-.12-.343-.21-.18-.09-1.057-.523-1.223-.58-.16-.06-.31-.04-.44.133-.17.24-.34.487-.48.634-.1.116-.28.13-.43.07-.2-.084-.75-.28-1.44-.89-.53-.47-.89-1.06-.99-1.237-.105-.18-.01-.284.07-.38.09-.11.178-.19.268-.296.09-.106.14-.16.2-.283.06-.12.017-.24-.027-.33-.044-.09-.405-.97-.553-1.326z"/> </g> </svg> </a> </div>  <div class="article-footer__tags"> <span class="tags-prefix">Meer over</span> <ul class="tags-list"> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=politiek, href=/tag/politiek)" href="/tag/politiek">politiek</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=internationale betrekkingen, href=/tag/internationale-betrekkingen)" href="/tag/internationale-betrekkingen">internationale betrekkingen</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=conflicten, oorlog en vrede, href=/tag/conflicten-oorlog-en-vrede)" href="/tag/conflicten-oorlog-en-vrede">conflicten, oorlog en vrede</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=internationale organisaties, href=/tag/internationale-organisaties)" href="/tag/internationale-organisaties">internationale organisaties</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=Amerika, href=/tag/amerika)" href="/tag/amerika">Amerika</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=Belgrado, href=/tag/belgrado)" href="/tag/belgrado">Belgrado</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=Bosni\xc3\xab, href=/tag/bosnie)" href="/tag/bosnie">Bosni\xc3\xab</a> </li> <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / tag / Link(label=Europa, href=/tag/europa)" href="/tag/europa">Europa</a> </li>  <li class="tags-list__item"> <a class="tags-list__link" data-gtm="more / author / Link(label=Van onze buitenlandredactie Anet Bleich, href=/auteur/van-onze-buitenlandredactie-anet-bleich)" href="/auteur/van-onze-buitenlandredactie-anet-bleich">Van onze buitenlandredactie Anet Bleich</a> </li> </ul> </div>  <div class="ad--mid" id="mid1" data-id="mid1"> <script>"function"==typeof window.loadAdvertSlot&&window.loadAdvertSlot("mid1")</script> </div>   <section class="section related-content"> <div class="section__anchor fjs-section-anchor" id="related"></div> <div class="section__header"> <h2 class="section-title" id="section-title-nieuws-achtergrond" style="--section-title-link:#000000;"> <span class="section-title__wrapper"> <a class="section-title__link" href="/nieuws-achtergrond" data-gtm="sectie / titel / Nieuws &amp; Achtergrond">Nieuws <span class=section-title__connector>&</span> Achtergrond</a> </span> </h2>  </div> \n    <div\n    class="teaser-block  teaser-block--related-article-grid-nl\n                            \n                            teaser-block--last grid--last fjs-last-grid fjs-last-teaser-block "\n>\n    \n    \n        \n            \n                \n    <!-- section = ${section != null ? section.name : \'\'} / grid = ${grid != null ? grid.type : \'\'} / teaser_size = ${tile.position.size} / position_grid = ${tile.position.index + 1} -->\n    <article\n        class="wl-tile   \n                                     teaser--position-0\n                                     teaser--size-s\n                                     \n                                        "\n        data-gtm="teaser_click"\n        data-position="1"\n        data-topic="Nieuws &amp; Achtergrond"\n        data-content_id="50a0696"\n        data-element=" text, image\n                                \n                                 "\n        data-teaser-size="S"\n        data-type="news feed"\n        data-teaser-grid-type="related-article-grid-nl"\n        data-personalisation="null"\n    >\n        \n    <div class="line1"></div>\n\n    <div class="line2"></div>\n\n\n        \n        \n        \n        \n        \n            <a\n    class="wl-teaser--basic  teaser--header-alignment-left teaser--element-direction-vertical   teaser--typographyset-basic   teaser--header-position-horizontal-left teaser--header-position-vertical-top teaser--has-short-title \n                           \n                           teaser--image-position-before_title"\n    href="/nieuws-achtergrond/de-politieke-schuldvraag-over-de-corona-aanpak-komt-later-pas-maar-nu-al-moet-minister-de-jonge-het-ontgelden~b50a0696/"\n    style=" \n                    border-color: #000000;\n                     "\n    aria-label="&lt;strong&gt;De schuldvraag over de corona-aanpak komt later pas, maar De Jonge moet het nu al ontgelden&lt;/strong&gt;"\n>\n    \n        \n            \n            \n            \n                \n    \n    \n        <picture class="teaser__image">\n            \n            \n            \n                <source\n                    data-srcset="https://images0.persgroep.net/rcs/IHkGawzhfTn8-py5LXcuQa-SXdw/diocontent/212791582/_crop/0/594/5855/2927/_fill/280/140?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85 280w"\n                    data-sizes="280px"\n                    data-media="(min-width: 640px)"\n                />\n                \n                <img\n                    class="teaser__image__thumbnail lazy"\n                    data-src="https://images0.persgroep.net/rcs/uOXMEtqtuH6Unt8Z6slspf_3RIM/diocontent/212791582/_crop/0/594/5855/2927/_fill/560/280?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n                    alt=""\n                    loading="lazy"\n                    width="280"\n                    height="140"\n                />\n                \n            \n        </picture>\n    \n\n\n            \n        \n    \n        \n            \n                \n    \n        \n            <header\n                class="teaser__header"\n                style="background-image: linear-gradient(to bottom,rgba(255, 255, 255, 0),rgba(255, 255, 255, 0), rgba(0, 0, 0, 0));"\n            >\n                <div\n    class="teaser__header__content-wrapper"\n    style="width:100%;"\n>\n    <h4\n        class="teaser__subtitle"\n        style="  --teaser-subtitle:#000000; "\n    >\n        <span class="teaser__labels">\n            \n            <span\n                class="teaser__label"\n            >Nieuws</span>\n            <span\n                class="teaser__sublabel"\n                style="color:#000000;\n                                display: inline;"\n            ></span>\n        </span>\n    </h4>\n\n    <span class="teaser__title-wrapper">\n        \n        <h3 class="teaser__title">\n            <span class="teaser__title__value--long"><strong>De schuldvraag over de corona-aanpak komt later pas, maar De Jonge moet het nu al ontgelden</strong></span>\n            <span class="teaser__title__value--short"><strong>De schuldvraag over de corona-aanpak komt later pas, maar De Jonge moet het nu al ontgelden</strong></span>\n        </h3>\n    </span>\n</div>\n            </header>\n        \n        \n        \n        \n    \n\n            \n            \n            \n        \n    \n</a>\n\n        \n    </article>\n\n\n<hr\n    class="divider"\n/>\n\n            \n            \n            \n            \n            \n            \n            \n            \n            \n        \n    \n        \n            \n            \n                \n    <article\n        data-id="mid2"\n        id="dfp-mid2-507da2dd-e255-4ec6-9624-67ed0c587262"\n        data-gtm="section = Nieuws &amp; Achtergrond / grid = ADVERTISEMENT / teaser_size = H / position_grid = 2"\n        class="teaser teaser-advertisement  advertisement--position-1 teaser--size-h mid2 "\n    >\n        \n    <div class="line1"></div>\n\n\n        <script>\n            window.APP.helpers.runWhenInBreakpoint({ \'xlarge\': true, \'large\':\n            true, \'medium\': true, \'small\':\n            true }, function() { window.loadAdvertSlot("mid2",\n            "dfp-mid2-507da2dd-e255-4ec6-9624-67ed0c587262"); });\n        </script>\n        \n    </article>\n\n\n            \n            \n            \n            \n            \n            \n            \n            \n        \n    \n        \n            \n            \n            \n            \n            \n            \n            \n            \n                <article\n    class="wl-tile list--position-5 teaser--size-v"\n>\n    \n    <div class="line1"></div>\n\n    <div class="line2"></div>\n\n\n    <div class="wl-tile__teaser-list  teaser-subteasers-counter--hide teaser--heavy-ruler  ">\n        <header class="teaser-list__header">\n            <a\n                href="/nieuws"\n                class="teaser-list__header-link"\n                data-gtm="NIEUWS / title / more"\n                aria-label="Meer nieuws"\n            >\n                <h4\n                    class="teaser-list__title teaser-list--align-left"\n                    style="--teaser-list-title:#000000"\n                >NIEUWS</h4>\n                <span\n                    class="teaser-list__more"\n                    href="/nieuws"\n                >meer <svg width="6" height="10"><path d="M.385 10a.385.385 0 0 1-.272-.657L4.456 5 .113.657A.385.385 0 0 1 .657.113l4.615 4.615c.15.15.15.394 0 .544L.657 9.887A.383.383 0 0 1 .385 10z"></path></svg></span>\n            </a>\n            \n        </header>\n        <div class="teaser-list__items">\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="888f366"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    \n    <a\n        class="teaser__link"\n        href="/sport/goud-voor-takagi-op-de-1-000-meter-leerdam-pakt-zilver~b888f366/"\n        aria-label="Goud voor Takagi op de 1.000 meter, Leerdam pakt zilver"\n    >\n        <span class="teaser__counter--small">1</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/-rexQrwR9gSNUcWie19E0I1zd7U/diocontent/212828172/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Goud voor Takagi op de 1.000 meter, Leerdam pakt zilver</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="102ffa5"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/nederland-schuldig-aan-structureel-extreem-oorlogsgeweld-in-indonesie-politiek-den-haag-keek-weg~b102ffa5/"\n        aria-label="Nederland schuldig aan structureel geweld in Indonesi\xc3\xab, Den Haag keek weg"\n    >\n        <span class="teaser__counter--small">2</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/X0DOv37heexg5E7KyRVP6dvQRIw/diocontent/174264717/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Nederland schuldig aan structureel geweld in Indonesi\xc3\xab, Den Haag keek weg</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="4d4e32c"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/graanprijs-blijft-omhoog-gaan-prijs-van-pilsje-stijgt-vermoedelijk-mee-die-van-brood-misschien-niet~b4d4e32c/"\n        aria-label="De graanprijs blijft omhoog gaan. De prijs van een pilsje stijgt vermoedelijk mee, die van brood misschien niet"\n    >\n        <span class="teaser__counter--small">3</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/H4Yv6NIktLJIk2aU2R1YZhnGgG4/diocontent/203641194/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">De graanprijs blijft omhoog gaan. De prijs van een pilsje stijgt vermoedelijk mee, die van brood misschien niet</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="a7422af"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/de-navo-is-terug-maar-rusland-ook-en-landen-buiten-de-navo-betalen-de-prijs~ba7422af/"\n        aria-label="De Navo is \xe2\x80\x98terug\xe2\x80\x99, maar Rusland ook \xe2\x80\x93 en landen buiten de Navo betalen de prijs"\n    >\n        <span class="teaser__counter--small">4</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/44Cs7eab0wZmEutYuuFbhtjpvrQ/diocontent/212790332/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">De Navo is \xe2\x80\x98terug\xe2\x80\x99, maar Rusland ook \xe2\x80\x93 en landen buiten de Navo betalen de prijs</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="b7135fa"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/dudley-houdt-huis-in-heel-nederland-gewonden-door-omgevallen-boom-en-afgewaaide-dakplaten~bb7135fa/"\n        aria-label="Dudley houdt huis in heel Nederland: gewonden door omgevallen boom en afgewaaide dakplaten"\n    >\n        <span class="teaser__counter--small">5</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/JMufG1QXmtffYARDFwuYrlY5jNA/diocontent/212822621/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Dudley houdt huis in heel Nederland: gewonden door omgevallen boom en afgewaaide dakplaten</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="f208c6f"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/air-france-klm-klimt-langzaam-uit-coronadal-maar-blijft-miljarden-verliezen~bf208c6f/"\n        aria-label="Air France-KLM klimt langzaam uit coronadal, maar blijft miljarden verliezen"\n    >\n        <span class="teaser__counter--small">6</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/bRKNuuZQJAQwc6pagIUa7HCQb8k/diocontent/209338506/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Air France-KLM klimt langzaam uit coronadal, maar blijft miljarden verliezen</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="aecefe4"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/frankrijk-en-europese-bondgenoten-trekken-zich-terug-uit-mali~baecefe4/"\n        aria-label="Frankrijk en Europese bondgenoten trekken zich terug uit Mali"\n    >\n        <span class="teaser__counter--small">7</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/Gxi1ldTvdmPKkbxtPKUtFIixMXE/diocontent/207385700/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Frankrijk en Europese bondgenoten trekken zich terug uit Mali</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="35f1d6b"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/wat-heeft-biden-te-winnen-met-de-oorlogsretoriek-over-oekraine~b35f1d6b/"\n        aria-label="Wat heeft Biden te winnen met de oorlogsretoriek over Oekra\xc3\xafne?"\n    >\n        <span class="teaser__counter--small">8</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/rtOCDa401StrLhVNgaVPO9-NerM/diocontent/212611535/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Wat heeft Biden te winnen met de oorlogsretoriek over Oekra\xc3\xafne?</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="0cfaba5"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/economie/republikeinen-blokkeren-drie-nieuwe-topbenoemingen-bij-de-federal-reserve~b0cfaba5/"\n        aria-label="Republikeinen blokkeren drie nieuwe topbenoemingen bij de Federal Reserve"\n    >\n        <span class="teaser__counter--small">9</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/yAwTspBC66rUOWdDHeHZ6JM4Zys/diocontent/212240196/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Republikeinen blokkeren drie nieuwe topbenoemingen bij de Federal Reserve</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--left    "\n    data-gtm="teaser_click"\n    data-position="6"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="18f4613"\n    data-type="list nieuws"\n    data-teaser-size="V"\n    data-element=" text, image "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/de-belangrijkste-grafieken-en-kaarten-over-de-uitbraak-van-het-coronavirus-in-nederland~b18f4613/"\n        aria-label="Wat zijn de huidige besmettingscijfers en hoe is de situatie in de ziekenhuizen?"\n    >\n        <span class="teaser__counter--small">10</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/jZo6ljFZAU_B2QVPzzhXyxmifxo/diocontent/174596023/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Wat zijn de huidige besmettingscijfers en hoe is de situatie in de ziekenhuizen?</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n        </div>\n    </div>\n</article>\n\n            \n            \n        \n    \n        \n            \n                \n    <!-- section = ${section != null ? section.name : \'\'} / grid = ${grid != null ? grid.type : \'\'} / teaser_size = ${tile.position.size} / position_grid = ${tile.position.index + 1} -->\n    <article\n        class="wl-tile   \n                                     teaser--position-2\n                                     teaser--size-s\n                                     \n                                        "\n        data-gtm="teaser_click"\n        data-position="3"\n        data-topic="Nieuws &amp; Achtergrond"\n        data-content_id="bf1fcb9"\n        data-element=" text, image\n                                \n                                 "\n        data-teaser-size="S"\n        data-type="news feed"\n        data-teaser-grid-type="related-article-grid-nl"\n        data-personalisation="null"\n    >\n        \n    <div class="line1"></div>\n\n\n        \n        \n        \n        \n        \n            <a\n    class="wl-teaser--basic  teaser--header-alignment-left teaser--element-direction-vertical   teaser--typographyset-basic   teaser--header-position-horizontal-left teaser--header-position-vertical-top teaser--has-short-title \n                           \n                           teaser--image-position-before_title"\n    href="/nieuws-achtergrond/was-de-afghaan-die-jaren-onopgemerkt-in-kerkrade-woonde-ooit-commandant-van-een-foltergevangenis-nabij-kabul~bbf1fcb9/"\n    style=" \n                    border-color: #000000;\n                     "\n    aria-label="Was de bejaarde Afghaan die in Kerkrade woonde de gevreesde commandant van een martelgevangenis?"\n>\n    \n        \n            \n            \n            \n                \n    \n    \n        <picture class="teaser__image">\n            \n            \n            \n                <source\n                    data-srcset="https://images0.persgroep.net/rcs/A0aeuTNemY8lz5je_xsGQgZzTxU/diocontent/212779765/_crop/0/297/2048/1024/_fill/280/140?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85 280w"\n                    data-sizes="280px"\n                    data-media="(min-width: 640px)"\n                />\n                \n                <img\n                    class="teaser__image__thumbnail lazy"\n                    data-src="https://images0.persgroep.net/rcs/CW4-e5DAWbWrTTP_L0msPCoOFmU/diocontent/212779765/_crop/0/297/2048/1024/_fill/560/280?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n                    alt=""\n                    loading="lazy"\n                    width="280"\n                    height="140"\n                />\n                \n            \n        </picture>\n    \n\n\n            \n        \n    \n        \n            \n                \n    \n        \n            <header\n                class="teaser__header"\n                style="background-image: linear-gradient(to bottom,rgba(255, 255, 255, 0),rgba(255, 255, 255, 0), rgba(0, 0, 0, 0));"\n            >\n                <div\n    class="teaser__header__content-wrapper"\n    style="width:100%;"\n>\n    <h4\n        class="teaser__subtitle"\n        style="  --teaser-subtitle:#000000; "\n    >\n        <span class="teaser__labels">\n            \n            <span\n                class="teaser__label"\n            >Reporta&shy;ge</span>\n            <span\n                class="teaser__sublabel"\n                style="color:#000000;\n                                display: inline;"\n            ></span>\n        </span>\n    </h4>\n\n    <span class="teaser__title-wrapper">\n        \n        <h3 class="teaser__title">\n            <span class="teaser__title__value--long">Was de bejaarde Afghaan die in Kerkrade woonde de gevreesde commandant van een martelgevangenis?</span>\n            <span class="teaser__title__value--short">Was de bejaarde Afghaan die in Kerkrade woonde de gevreesde commandant van een martelgevangenis?</span>\n        </h3>\n    </span>\n</div>\n            </header>\n        \n        \n        \n        \n    \n\n            \n            \n            \n        \n    \n</a>\n\n        \n    </article>\n\n\n<hr\n    class="divider"\n/>\n\n            \n            \n            \n            \n            \n            \n            \n            \n            \n        \n    \n        \n            \n                \n    <!-- section = ${section != null ? section.name : \'\'} / grid = ${grid != null ? grid.type : \'\'} / teaser_size = ${tile.position.size} / position_grid = ${tile.position.index + 1} -->\n    <article\n        class="wl-tile   \n                                     teaser--position-3\n                                     teaser--size-s\n                                     \n                                        "\n        data-gtm="teaser_click"\n        data-position="4"\n        data-topic="Nieuws &amp; Achtergrond"\n        data-element=" text, image\n                                \n                                 "\n        data-teaser-size="S"\n        data-type="news feed"\n        data-teaser-grid-type="related-article-grid-nl"\n        data-personalisation="null"\n    >\n        \n    <div class="line1"></div>\n\n\n        \n        \n        \n        \n        \n            <a\n    class="wl-teaser--basic  teaser--header-alignment-left teaser--element-direction-vertical   teaser--typographyset-basic   teaser--header-position-horizontal-left teaser--header-position-vertical-top teaser--has-short-title \n                           \n                           teaser--image-position-before_title"\n    href="https://www.volkskrant.nl/kijkverder/v/2021/grijs-gebied~v464558/"\n    target="_blank"\n    rel="noopener noreferrer"\n    style=" \n                    border-color: #000000;\n                     "\n    aria-label="Nabestaanden vrezen dat de moord op Heidy wordt vergeten: \xe2\x80\x98Mijn zus is dood\xe2\x80\x99"\n>\n    \n        \n            \n            \n            \n                \n    \n    \n        <picture class="teaser__image">\n            \n            \n            \n                <source\n                    data-srcset="https://images0.persgroep.net/rcs/ejKJyz8h8mNj2h5QsUU0Hx6unAg/diocontent/212800201/_focus/0.5/0.5/_fill/280/140?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85 280w"\n                    data-sizes="280px"\n                    data-media="(min-width: 640px)"\n                />\n                \n                <img\n                    class="teaser__image__thumbnail lazy"\n                    data-src="https://images0.persgroep.net/rcs/VevsN39-7HAGMcZlnrN8lPHcjDk/diocontent/212800201/_focus/0.5/0.5/_fill/560/280?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n                    alt=""\n                    loading="lazy"\n                    width="280"\n                    height="140"\n                />\n                \n            \n        </picture>\n    \n\n\n            \n        \n    \n        \n            \n                \n    \n        \n            <header\n                class="teaser__header"\n                style="background-image: linear-gradient(to bottom,rgba(255, 255, 255, 0),rgba(255, 255, 255, 0), rgba(0, 0, 0, 0));"\n            >\n                <div\n    class="teaser__header__content-wrapper"\n    style="width:100%;"\n>\n    <h4\n        class="teaser__subtitle"\n        style="  --teaser-subtitle:#000000; "\n    >\n        <span class="teaser__labels">\n            \n            <span\n                class="teaser__label"\n            >Podcast Grijs Gebied</span>\n            <span\n                class="teaser__sublabel"\n                style="color:#000000;\n                                display: inline;"\n            ></span>\n        </span>\n    </h4>\n\n    <span class="teaser__title-wrapper">\n        \n        <h3 class="teaser__title">\n            <span class="teaser__title__value--long">Nabestaanden vrezen dat de moord op Heidy wordt vergeten: \xe2\x80\x98Mijn zus is dood\xe2\x80\x99</span>\n            <span class="teaser__title__value--short">Nabestaanden vrezen dat de moord op Heidy wordt vergeten: \xe2\x80\x98Mijn zus is dood\xe2\x80\x99</span>\n        </h3>\n    </span>\n</div>\n            </header>\n        \n        \n        \n        \n    \n\n            \n            \n            \n        \n    \n</a>\n\n        \n    </article>\n\n\n<hr\n    class="divider"\n/>\n\n            \n            \n            \n            \n            \n            \n            \n            \n            \n        \n    \n        \n            \n            \n            \n            \n            \n            \n            \n            \n                <article\n    class="wl-tile list--position-4 teaser--size-s"\n>\n    \n    <div class="line1"></div>\n\n\n    <div class="wl-tile__teaser-list  teaser-subteasers-counter--show teaser--heavy-ruler  ">\n        <header class="teaser-list__header">\n            <a\n                href="/best-gelezen"\n                class="teaser-list__header-link"\n                data-gtm="BEST GELEZEN / title / more"\n                aria-label="Meer best gelezen"\n            >\n                <h4\n                    class="teaser-list__title teaser-list--align-left"\n                    style="--teaser-list-title:#000000"\n                >BEST GELEZEN</h4>\n                <span\n                    class="teaser-list__more"\n                    href="/best-gelezen"\n                >meer <svg width="6" height="10"><path d="M.385 10a.385.385 0 0 1-.272-.657L4.456 5 .113.657A.385.385 0 0 1 .657.113l4.615 4.615c.15.15.15.394 0 .544L.657 9.887A.383.383 0 0 1 .385 10z"></path></svg></span>\n            </a>\n            \n        </header>\n        <div class="teaser-list__items">\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--none    "\n    data-gtm="teaser_click"\n    data-position="5"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="6035fe3"\n    data-type="list best gelezen"\n    data-teaser-size="S"\n    data-element=" text "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    \n    <a\n        class="teaser__link"\n        href="/cultuur-media/hoe-zelfbenoemd-zevenvinker-joris-luyendijk-het-middelpunt-werd-van-een-mediastorm~b6035fe3/"\n        aria-label="Hoe zelfbenoemd \xe2\x80\x98zevenvinker\xe2\x80\x99 Joris Luyendijk het middelpunt werd van een mediastorm"\n    >\n        <span class="teaser__counter--small">1</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/58NM3sKr8kwQ8DS5n7jdqVkXDRE/diocontent/212780432/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Hoe zelfbenoemd \xe2\x80\x98zevenvinker\xe2\x80\x99 Joris Luyendijk het middelpunt werd van een mediastorm</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--none    "\n    data-gtm="teaser_click"\n    data-position="5"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="a7422af"\n    data-type="list best gelezen"\n    data-teaser-size="S"\n    data-element=" text "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/nieuws-achtergrond/de-navo-is-terug-maar-rusland-ook-en-landen-buiten-de-navo-betalen-de-prijs~ba7422af/"\n        aria-label="De Navo is \xe2\x80\x98terug\xe2\x80\x99, maar Rusland ook \xe2\x80\x93 en landen buiten de Navo betalen de prijs"\n    >\n        <span class="teaser__counter--small">2</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/44Cs7eab0wZmEutYuuFbhtjpvrQ/diocontent/212790332/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">De Navo is \xe2\x80\x98terug\xe2\x80\x99, maar Rusland ook \xe2\x80\x93 en landen buiten de Navo betalen de prijs</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n                \n<article\n    class="teaser--small  image-position-horizontal--none    "\n    data-gtm="teaser_click"\n    data-position="5"\n    data-topic="Nieuws &amp; Achtergrond"\n    data-content_id="24b98f6"\n    data-type="list best gelezen"\n    data-teaser-size="S"\n    data-element=" text "\n    data-teaser-grid-type="related-article-grid-nl"\n>\n    <hr class="divider--small" />\n    <a\n        class="teaser__link"\n        href="/mensen/ik-leer-veel-van-jonge-activisten~b24b98f6/"\n        aria-label="Kristijaan (71): \xe2\x80\x98Ik ben 100 procent queer. Queer is een middelvinger naar die hele alfabetsoep\xe2\x80\x99"\n    >\n        <span class="teaser__counter--small">3</span>\n        \n        <img\n            class="wl-teaser__image teaser__image--small fjs-teaser-image lazy  "\n            onerror="if (!this.getAttribute(\'data-has-retried\')) { this.src=this.getAttribute(\'data-placeholder-image-source\'); this.setAttribute(\'data-has-retried\', true);}"\n            data-placeholder-image-source="https://sel-static.persgroep.net/shared/images/vk-placeholder.png"\n            data-src="https://images0.persgroep.net/rcs/Jw839Jbvwrz4XA5erIq_j1AIfMA/diocontent/211615720/_focus/0.5/0.5/_fill/70/70?appId=93a17a8fd81db0de025c8abd1cca1279&amp;quality=0.85"\n            style=" "\n            alt=""\n            loading="lazy"\n            width="65"\n            height="65"\n        />\n        <header class="teaser__header--small teaser__header--align-left">\n            <h4\n                class="wl-teaser__labels teaser__subtitle teaser__subtitle--small teaser__labels--align-left"\n            >\n                \n                <span class="teaser__sublabel teaser__sublabel--small"></span>\n                \n            </h4>\n            <span class="teaser__title-wrapper"\n                >\n                <span class="teaser__title--small fjs-teaser-title">Kristijaan (71): \xe2\x80\x98Ik ben 100 procent queer. Queer is een middelvinger naar die hele alfabetsoep\xe2\x80\x99</span\n                >\n            </span>\n        </header>\n    </a>\n</article>\n\n            \n        </div>\n    </div>\n</article>\n\n            \n            \n        \n    \n\n</div>\n\n <div class="read-more"> <div class="read-more__wrapper"> <a class="read-more__link fjs-read-more-link" data-gtm="more / nieuws-achtergrond / 3" data-href="/leesmeer/VKGO_NIEUWS_ACHTERGROND/3" href="/nieuws-achtergrond">Meer Nieuws &amp; Achtergrond</a>  </div> </div>    </section>  <div class="fjs-paywall"></div>  <div class="ad--section fjs-ad-hide-empty" id="branded2" data-id="branded2"></div> <script>window.APP.helpers.runWhenInBreakpoint({xlarge:!0,large:!0,medium:!0,small:!1},function(){"function"==typeof window.loadAdvertSlot&&window.loadAdvertSlot("branded2")})</script>  <script src="/article-static/article-a30be65658.js" async="async"></script>  </main>  <nav class="app-bottom-navigation fjs-app-bottom-navigation"> <ul class="app-bottom-navigation__item-list"> <li class="app-bottom-navigation__item"> <a class="app-bottom-navigation__link" href="/"> <svg class="app-bottom-navigation__item-icon" width="22" height="22" xmlns="http://www.w3.org/2000/svg"> <g fill="none" fill-rule="evenodd"> <path fill="none" d="M0 0h22v22H0z"/> <path d="M10.752 19.865L3.447 3.996a4.208 4.208 0 00-.479-.851c-.23-.266-.531-.355-1.382-.355H.983v-.655h9.61v.655h-.337a6.061 6.061 0 00-1.082.053.682.682 0 00-.656.692c.08.573.267 1.126.55 1.631l4.077 9.148L16.94 5.75a7.218 7.218 0 00.71-2.11c0-.655-.426-.85-2.021-.85h-.143v-.656h5.532v.655h-.177c-.35-.02-.702.004-1.047.071-.371.106-.549.337-.904 1.135l-7.198 15.87h-.94z" fill="currentColor" fill-rule="nonzero"/> </g> </svg>  <span class="app-bottom-navigation__link-text">Live</span> </a> </li> <li class="app-bottom-navigation__item" data-gtm="navigation / edition"> <a class="app-bottom-navigation__link" href="/editie/vandaag"> <svg class="app-bottom-navigation__item-icon" width="22" height="22" fill="none" xmlns="http://www.w3.org/2000/svg"> <path fill-rule="nonzero" d="M23.5 9h2a.5.5 0 01.5.5v7a2.503 2.503 0 01-2.5 2.5h-21A2.503 2.503 0 010 16.5V.5A.5.5 0 01.5 0h21a.5.5 0 01.5.5v16a1.5 1.5 0 003 0V10h-1.5a.5.5 0 010-1zM1 16.5c0 .828.672 1.5 1.5 1.5h19.001A2.486 2.486 0 0121 16.5V1H1v15.5z" fill="currentColor"/> <path fill-rule="nonzero" d="M10.5 13h-7a.5.5 0 01-.5-.5v-3a.5.5 0 01.5-.5h7a.5.5 0 01.5.5v3a.5.5 0 01-.5.5zm-.5-3H4v2h6v-2z" fill="currentColor"/> <path d="M3.596 4h14.808a.5.5 0 000-1H3.596a.5.5 0 000 1zM3.596 7h14.808a.5.5 0 000-1H3.596a.5.5 0 000 1zM18.505 9H13.5a.5.5 0 000 1h5.005a.5.5 0 000-1zM18.404 13a.5.5 0 000-1H13.5a.5.5 0 000 1h4.904zM18.5 15h-15a.5.5 0 000 1h15a.5.5 0 000-1z" fill="currentColor"/> </svg> <span class="app-bottom-navigation__link-text">Editie</span> </a> </li> <li class="app-bottom-navigation__item"> <a class="app-bottom-navigation__link" href="/best-gelezen"> <svg class="app-bottom-navigation__item-icon" width="22" height="22" xmlns="http://www.w3.org/2000/svg"> <g fill="none" fill-rule="evenodd"> <path fill="none" d="M0 0h22v22H0z"/> <path d="M20.617 4.063a.5.5 0 00-.371-.141H16.02a.5.5 0 100 1h3.044l-6.9 6.964-3.675-1.908a.5.5 0 00-.584.09l-7 7a.5.5 0 10.707.707l6.741-6.741 3.677 1.909a.5.5 0 00.586-.092l7.143-7.213v3.026a.5.5 0 101 0V4.437a.5.5 0 00-.142-.374z" fill-rule="nonzero" fill="currentColor"/> </g> </svg> <span class="app-bottom-navigation__link-text">Best gelezen</span> </a> </li>  <li class="app-bottom-navigation__item"> <a class="app-bottom-navigation__link" href="/search"> <svg class="app-bottom-navigation__item-icon" width="22" height="22" xmlns="http://www.w3.org/2000/svg"> <g fill="none" fill-rule="evenodd"> <path fill="none" d="M0 0h22v22H0z"/> <path d="M19.103 18.31l-4.696-4.858-.006-.004c1.738-2.019 2.17-4.903 1.107-7.372C14.444 3.607 12.08 2.013 9.468 2h-.034c-3.674-.01-6.66 3.064-6.668 6.864-.01 3.8 2.962 6.889 6.636 6.898h.034a6.475 6.475 0 004.413-1.744l.005.006 4.696 4.857a.382.382 0 00.548-.005.415.415 0 00.005-.567zm-9.697-3.358c-3.242-.008-5.863-2.733-5.856-6.086.008-3.353 2.642-6.065 5.884-6.056h.03c3.241.016 5.856 2.748 5.84 6.101-.016 3.353-2.657 6.058-5.898 6.041z" fill-rule="nonzero" fill="currentColor"/> </g> </svg> <span class="app-bottom-navigation__link-text">Zoeken</span> </a> </li> <li class="app-bottom-navigation__item"> <a class="app-bottom-navigation__link" href="/menu"> <svg class="app-bottom-navigation__item-icon" width="22" height="22" xmlns="http://www.w3.org/2000/svg"> <g fill="none" fill-rule="evenodd"> <path fill="none" d="M0 0h22v22H0z"/> <g fill-rule="nonzero"> <path d="M11 8.5a2.5 2.5 0 100 5 2.5 2.5 0 000-5zm0 4a1.5 1.5 0 110-3 1.5 1.5 0 010 3zM19 8.5a2.5 2.5 0 100 5 2.5 2.5 0 000-5zm0 4a1.5 1.5 0 110-3 1.5 1.5 0 010 3zM3 8.5a2.5 2.5 0 100 5 2.5 2.5 0 000-5zm0 4a1.5 1.5 0 110-3 1.5 1.5 0 010 3z" fill="currentColor"/> </g> </g> </svg> <span class="app-bottom-navigation__link-text">Service</span> </a> </li> </ul> </nav>  <div class="app-overlay js-is-hidden fjs-app-overlay"></div> <section id="menu" class="app-menu fjs-app-menu js-is-hidden"> <div class="menu-close"> <svg class="menu-close-icon fjs-close-menu"> <path d="M17.9.1a.5.5 0 0 0-.8 0L9 8.3 1.1.4a.5.5 0 0 0-.7.7L8.3 9 0 17.1a.5.5 0 1 0 .8.8L9 9.7l7.9 7.9a.5.5 0 0 0 .7-.7L9.7 9 18 .9a.5.5 0 0 0 0-.8z"/> </svg> </div> <h2 class="menu-title">Service</h2>    <ul class="menu-list">   <li class="menu-list__item fjs-hide-logged-in"> <a class="menu-list__item-link" href="/login?dpg_source=vk&amp;dpg_campaign=stdc_vk&amp;dpg_medium=menu&amp;redirect_url=https://www.volkskrant.nl" data-gtm="menu / Inloggen / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Inloggen</h4>  <p class="menu-item__text">Log in of maak een account aan</p>  </div> </a>  </li>     <li class="menu-list__item js-is-hidden fjs-show-logged-in"> <a class="menu-list__item-link" href="/logout" data-gtm="menu / Uitloggen / text"> <div class="menu-item__text-wrapper">  <h4 class="menu-item__title" onclick="window.APP.loginHelpers.logout()">Uitloggen</h4>  <p class="fjs-logout-text menu-item__text"></p> </div> </a>  </li>     <li class="menu-list__item js-is-hidden fjs-show-logged-in"> <a class="menu-list__item-link" href="https://myaccount.volkskrant.nl/" data-gtm="menu / Mijn account / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Mijn account</h4>  <p class="menu-item__text">Uw gegevens en de bezorg-, vakantie- en verhuisservice</p>  </div> </a> <a href="https://myaccount.volkskrant.nl/" target="_blank" rel="noopener noreferrer" data-gtm="menu / Mijn account / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/meldingen" data-gtm="menu / Meldingen / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Meldingen</h4>  <p class="menu-item__text">Hier kunt u meldingen uitschakelen</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/search" data-gtm="menu / Zoeken / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Zoeken</h4>  <p class="menu-item__text">Zoeken op auteur, titel of onderwerp</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/puzzels" data-gtm="menu / Puzzels / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Puzzels</h4>  <p class="menu-item__text">Test uw kennis met uitdagende puzzels</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/feedback-web" data-gtm="menu / Wat vindt u van deze website? / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Wat vindt u van deze website?</h4>  <p class="menu-item__text">We horen graag uw mening</p>  </div> </a> <a href="/feedback-web" target="_blank" rel="noopener noreferrer" data-gtm="menu / Wat vindt u van deze website? / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.volkskrant.nl/service" data-gtm="menu / Klantenservice / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Klantenservice</h4>  <p class="menu-item__text">Serviceomgeving voor al uw vragen</p>  </div> </a>  </li>      <div class="fjs-subscribe-menu-item"></div> <li data-temptation-position="MENU_ITEM" class="menu-list__item"></li>    <li class="menu-list__item fjs-hide-logged-in"> <a class="menu-list__item-link" href="/nieuwsbrief" data-gtm="menu / Nieuwsbrieven / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Nieuwsbrieven</h4>  <p class="menu-item__text">Kies en beheer uw favoriete nieuwsbrieven</p>  </div> </a>  </li>     <li class="menu-list__item js-is-hidden fjs-show-logged-in"> <a class="menu-list__item-link" href="/nieuwsbrief" data-gtm="menu / Mijn nieuwsbrieven / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Mijn nieuwsbrieven</h4>  <p class="menu-item__text">Kies en beheer uw favoriete nieuwsbrieven</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/editie/vandaag" data-gtm="menu / Editie / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Editie</h4>  <p class="menu-item__text">Klassieke en Digitale versie van de krant</p>  </div> </a> <a href="/editie/vandaag" target="_blank" rel="noopener noreferrer" data-gtm="menu / Editie / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.volkskrant.nl/familieberichten" data-gtm="menu / Familieberichten / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Familieberichten</h4>  <p class="menu-item__text">Uitleg over overlijdensberichten in de krant</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.topics.nl/" data-gtm="menu / Topics / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Topics</h4>  <p class="menu-item__text">Alle artikelen uit 13 kwaliteitskranten</p>  </div> </a> <a href="https://www.topics.nl/" target="_blank" rel="noopener noreferrer" data-gtm="menu / Topics / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://webwinkel.volkskrant.nl/" data-gtm="menu / Winkel / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Winkel</h4>  <p class="menu-item__text">Zorgvuldig geselecteerde artikelen</p>  </div> </a> <a href="https://webwinkel.volkskrant.nl/" target="_blank" rel="noopener noreferrer" data-gtm="menu / Winkel / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.volkskrant.nl/tipderedactie/" data-gtm="menu / Tip de redactie / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Tip de redactie</h4>  <p class="menu-item__text">Deel belangrijke info met de Volkskrantredactie</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.dpgmedia.nl/gebruiksvoorwaarden" data-gtm="menu / Gebruiksvoorwaarden / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Gebruiksvoorwaarden</h4>  <p class="menu-item__text">Uitleg over uw rechten en regels</p>  </div> </a> <a href="https://www.dpgmedia.nl/gebruiksvoorwaarden" target="_blank" rel="noopener noreferrer" data-gtm="menu / Gebruiksvoorwaarden / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="https://www.dpgmedia.nl/privacy" data-gtm="menu / Privacy / text" target="_blank" rel="noopener noreferrer"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Privacy</h4>  <p class="menu-item__text">Uitleg over wat wij met uw gegevens doen</p>  </div> </a> <a href="https://www.dpgmedia.nl/privacy" target="_blank" rel="noopener noreferrer" data-gtm="menu / Privacy / icon"> <div class="menu-item__icon-wrapper"> <svg class="menu-item__icon"> <path d="M12.5 0h-9a.5.5 0 0 0 0 1h7.8L0 12.1a.5.5 0 1 0 .8.8L12 1.7v7.8a.5.5 0 1 0 1 0v-9a.5.5 0 0 0-.5-.5z"/> </svg> </div> </a> </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="/cookiewall/info" data-gtm="menu / Cookiebeleid / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Cookiebeleid</h4>  <p class="menu-item__text">Uitleg over de gebruikte cookies op onze site en app</p>  </div> </a>  </li>     <li class="menu-list__item"> <a class="menu-list__item-link" href="javascript:showPrivacyModal();" data-gtm="menu / Cookie instellingen / text"> <div class="menu-item__text-wrapper"> <h4 class="menu-item__title">Cookie instellingen</h4>  <p class="menu-item__text">Beheer uw privacy instellingen</p>  </div> </a>  </li>    </ul>   </section>  <div class="app-editor-tip"> <h3 class="editor-tip-title">Wilt u belangrijke informatie delen met de Volkskrant?</h3> <p class="editor-tip-text"><a href="https://www.volkskrant.nl/tip/">Tip hier onze journalisten</a></p> </div>  <div class="app-footer"> <div class="app-footer__column"> <h4 class="app-footer__title">Algemeen</h4> <ul class="app-footer__list"> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/contact-redactie" data-gtm="footer / Contact met de redactie">Contact met de redactie</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/service/contact" data-gtm="footer / Contact met de klantenservice">Contact met de klantenservice</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.dpgmedia.nl/privacy" data-gtm="footer / Privacystatement">Privacystatement</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/service/sites/default/files/Abonnementsvoorwaarden_de_Volkskrant.pdf" data-gtm="footer / Abonnementsvoorwaarden">Abonnementsvoorwaarden</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.dpgmedia.nl/gebruiksvoorwaarden" data-gtm="footer / Gebruiksvoorwaarden"> Gebruiksvoorwaarden</a> </li>  <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/cookies" data-gtm="footer / Cookiebeleid">Cookiebeleid</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="#" onclick="showPrivacyModal()" data-gtm="footer / Cookieinstellingen">Cookie-instellingen</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/nieuws-achtergrond/copyright-auteursrechten-volkskrant~b602876a9/" data-gtm="footer / Auteursrecht"> Auteursrecht </a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/nieuws-achtergrond/het-colofon-van-de-volkskrant~bce3473d/" data-gtm="footer / Colofon"> Colofon</a> </li> </ul> </div> <div class="app-footer__column"> <h4 class="app-footer__title">Service</h4> <ul class="app-footer__list"> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/service" data-gtm="footer / klantenservice">Klantenservice</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://myaccount.volkskrant.nl" data-gtm="footer / Mijn account">Mijn account</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/service/vakantie" data-gtm="footer / Vakantieservice">Vakantieservice</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.dpgmedia.nl/merk/devolkskrant" data-gtm="footer / Adverteren">Adverteren</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.dpglosseverkoop.nl/tornl_j2ee/pag/pve_index.jsp" data-gtm="footer / Losse\n                verkoop">Losse verkoop</a> </li> </ul> </div> <div class="app-footer__column"> <h4 class="app-footer__title">Meer de Volkskrant</h4> <ul class="app-footer__list"> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/abonnementen/" data-gtm="footer / Abonneren">Abonneren</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/nieuwsbrief" data-gtm="footer / Nieuwsbrieven">Nieuwsbrieven</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://krant.volkskrant.nl/titles/volkskrant/7929/" data-gtm="footer / Digitale krant">Digitale krant</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://webwinkel.volkskrant.nl/" data-gtm="footer / Webwinkel">Webwinkel</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.volkskrant.nl/puzzels/uitgelicht" data-gtm="footer / Puzzels">Puzzels</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/rss/feeds" data-gtm="footer / RSS-feeds">RSS-feeds</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.facebook.com/volkskrant/" data-gtm="footer / Facebook">Facebook</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://www.twitter.com/volkskrant" data-gtm="footer / Twitter">Twitter</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://play.google.com/store/apps/details?id=be.persgroep.android.news.mobilevk&hl=nl" data-gtm="footer / Android apps">Android apps</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="https://itunes.apple.com/nl/app/de-volkskrant/id418873064?mt=8" data-gtm="footer / iOS apps">iOS apps</a> </li> </ul> </div> <div class="app-footer__column"> <h4 class="app-footer__title">Navigeer</h4> <ul class="app-footer__list"> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/columns" data-gtm="footer / Columnisten">Columnisten</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/recensies" data-gtm="footer / Recensies">Recensies</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/volkskeuken" data-gtm="footer / De Volkskeuken">De Volkskeuken</a> </li> <li class="app-footer__list-item"> <a class="app-footer__list-link" href="/archief" data-gtm="footer / Archief">Archief</a> </li> </ul> </div> </div>  <div class="app-disclaimer__wrapper"> <div class="app-disclaimer"> <img class="app-disclaimer__image lazy" data-src="/static/sigmund-footer.png" alt="" loading="lazy" width="110" height="88"> <hr class="app-disclaimer__line"> <p class="app-disclaimer__text"> Op alle verhalen van de Volkskrant rust uiteraard copyright.<br> Wil je tekst overnemen of een video(fragment), foto of illustratie gebruiken, mail dan naar <a href="mailto:copyright@volkskrant.nl">copyright @volkskrant.nl</a>. <br> \xc2\xa9 <span>2022</span> DPG Media B.V. - alle rechten voorbehouden </p> </div> </div>  <div class="ad--layer" id="layer1" data-id="layer1"> <script>"function"==typeof window.loadAdvertSlot&&window.loadAdvertSlot("layer1")</script> </div>  <noscript> <iframe src="https://t.volkskrant.nl/gtm.js?id=GTM-TPHZL7D" height="0" width="0" style="display:none;visibility:hidden"></iframe> </noscript>  <div data-temptation-position="PAGE_BOTTOM"></div>  </body> </html> ', 'in_cache_date': datetime.datetime(2022, 2, 17, 12, 18, 59, 373490), 'parsed': 0}
```

```python
def parse_article(r):
    # TODO: 
    # timestamp
    # title
    # publisherID
    # cleantext
    # category
    
    # (image)
    # (keywords)
    
    a = {}

    a['url']  = r['url']
    
    tree = html.fromstring(r['raw_html'])
    
    try:
        
        a['timestamp'] = tree.xpath('/html/body/main/article/header/section/time[@class="artstyle__production__datetime"]/@datetime')[0].strip()
    except: 
        a['timestamp'] = None
    try:
        a['title'] = tree.xpath('//h1/text()')[0].strip()
    except:
        a['title'] = None
    try:
        a['publisherID'] = tree.xpath('/html/body/main/article/header/section/span[1]/a/text()')[0].strip()
    except:
        a['publisherID'] = None
    try:
        a['cleantext'] = ''.join(tree.xpath('//html/body/main/article/section/section/*/text()')[1:])
    except:
        a['cleantext'] = None
    try:
        a['category'] = tree.xpath('/html/body/main/section/div[2]/h2/span/a/text()')[0].strip()
    except:
        a['category'] = None
    
    return a
```
